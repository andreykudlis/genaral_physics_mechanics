# Момент импульса и момент инерции абсолютно твёрдого тела

## Момент импульса

Будем называть момент импульса системы мат. точек:

$$\vec{L}\;=\;\sum_i \vec{L}_i\;=\;\sum_i \vec{r}_i \times \vec{p}_i,$$ 
$$\vec{L}\;=\; \int_V \vec{r} \times \vec{v} \mathrm{d}m.$$ 

Дальше необходимо детализировать рассмотрение: (рассмотрим частные случаи)

Например, можно рассмотреть момент импульса относительно оси:

<br>

\begin{align}
    &\vec{L}\;=\;\vec{e}_x L_x \;+\;\vec{e}_y L_y \;+\; \vec{e}_z L_z, \\
    &\vec{L}_z\;=\;L_z \vec{e}_z\;=\;\sum_i (\vec{r}_i \times \vec{p}_i)_z \vec{e}_z.
\end{align}

<br>

<br>


Рассмотрим i-ый вклад: $(\vec{r}_i \times \vec{p}_i)_z\;=$
$$=\;\left[ (\vec{r}_i^{||}\;+\; \vec{r}_i^{\perp})\;\times\; (\vec{p}_i^{||}\;+\; \vec{p}_i^{\perp}) \right]_z\;=\;$$
$$=\; (\vec{r}_i^{\perp}\;\times\;\vec{p}_i^{\perp})_z\;=\; (\vec{r}_i^{\perp}\;\times\;\vec{p}_i^{\perp})_z\;=\; m_i \omega r_i^{\perp 2}.$$ 

$$L_z\;=\;\sum_i m_i \omega r_i^{\perp 2}\;=\; \omega \int_V \mathrm{d}m r^{\perp 2}.$$

<br>

**Момент инерции относительно оси:** $I_z\;=\;\omega \int_V r^{\perp 2} \rho dx dy dz.$ 

Получим удобный инструмент вычислений моментов инерции относительно двух параллельных осей.

Пусть $\vec{R}$ - радиус-вектор ||-го переноса.
Пусть есть точка, ее вектор относительно оси 1 будет $\vec{r}_1$ связан с $\vec{r}_2$ (относительно второй оси):

<br>

<figure>
    <center>
    <img src="images/lecture_22_dve_osi.png" width="350" align="center"/>
        <figcaption>Две параллельные оси.</figcaption>
    </center>
</figure>

<br>


**Теорема Гюйгена-Штейнера**

$$\vec{r}_1 \;=\; \vec{r}_2 \;+\; \vec{R},$$

<br>

\begin{align}
    &I_1\;=\; \int_V \vec{r}_1^2 \mathrm{d}m\;=\;\int_V \vec{r}_1 \vec{r}_1 \mathrm{d}m\;=\;\int_V (\vec{r}_2\;+\;\vec{R})(\vec{r}_2\;+\;\vec{R}) \mathrm{d}m\;= \\
    &=\; \int_V \vec{r}_2^2 \mathrm{d}m \;+\;\vec{R}^2 \int \mathrm{d}m \;+\; 2\int \vec{r}_2 \vec{R} \mathrm{d}m\;= \\
    &=\; \int \vec{r}_2^2 \mathrm{d}m\;+\; \vec{R}^2 \int \mathrm{d}m\;+\; 2\vec{R} \int \vec{r}_2 \mathrm{d}m\;=\; I_2\;+\;R^2 \cdot M.
\end{align}

<br>

**Небольшое замечание:** Важно, что $I_2$  - момент инерции относительно оси, проходящей через центр масс тела параллельно исходной оси. Если это не будет так, то тогда теорема Гюйгенса-Штейнера не будет верна. Именно из-за того, что 2 -  система центра масс :

$\vec r_{2c}=0=\dfrac{\int \vec r_2 \mathrm{d}m}{M}$ (Радиус-вектор центра масс в системе центра масс равен 0)

Часто рассматривают момент инерции относительно точки

$$I_o\;=\; \int r^2 \mathrm{d}m$$
$r$ - расстояние от фиксированной точки до рассматриваемой точки тела.

<br>

\begin{align}
    &\int_V r^2 \mathrm{d}m\;=\; \int_V (x^2\;+\;y^2\;+\;z^2) \mathrm{d}m\;=\;\dfrac{1}{2} \int_V (2x^2\;+\;2y^2\;+\;2z^2) \mathrm{d}m\;= \\
    &=\;\dfrac{1}{2} \left[ \int_V (x^2\;+\;y^2) \mathrm{d}m \;+\; \int_V (x^2\;+\;z^2) \mathrm{d}m\;+\; \int_V (y^2\;+\;z^2) \mathrm{d}m \right].
\end{align}

<br>

$$2I_o\;=\;I_x\;+\;I_y\;+\;I_z.$$

Можно еще рассмотреть момент инерции относительно плоскости:

<br>

$$I_o\;=\int \vec r^2 \mathrm{d}m=\int (x^2+y^2+z^2)\mathrm{d}m=\int x^2\mathrm{d}m+\int y^2\mathrm{d}m+\int z^2\mathrm{d}m=\;I_{zy}\;+\;I_{zx}\;+\;I_{xy}.$$


## Примеры вычислений
### Тонкий стержень.

<br>

<br>


\begin{align}
    & I_z = \int x^2 \mathrm{d}m = \int x^2 \cdot dx \cdot \rho_{\lambda} = 2 \overset{\dfrac{l}{2}}{\underset{0}{\int}} x^2 \cdot dx \dfrac{m}{l} = \dfrac{2m}{l} \cdot \dfrac{x^3}{3} \; \overset{\dfrac{l}{2}}{\underset{0}{\bigg |}} = \newline
    & = \dfrac{2m}{3l} \cdot \dfrac{l^3}{8} = \dfrac{ml^2}{12}. 
    \\
    & \text{Относительно конца:} \dfrac{ml^2}{12} + m \bigg(\dfrac{l}{2} \bigg)^2 = \dfrac{4}{12} ml^2 = \dfrac{ml^2}{3}. 
\end{align}

### Пластина со сторонами a, b.

<br>

<br>
 
\begin{align}
    & I_z = \int x^2 \mathrm{d}m = \int x^2 \cdot \dfrac{m}{a\cdot b} \cdot dx dz = 2 \overset{\dfrac{a}{2}}{\underset{0}{\int}} x^2 dx \cdot \dfrac{m}{a\cdot b} \overset{b}{\underset{0}{\int}} dz = 2 \cdot \dfrac{(\dfrac{a}{2})^3}{3} \dfrac{m}{a\cdot b} = \dfrac{ma^2}{12}. 
\end{align}

<br>

А чему будет равен момент инерции через центр масс: 
\begin{align}
    & 2I_0 = I_x + I_y + I_z
    \\
    & \boxed{ I_y = I_x + I_z = \dfrac{m}{12} (a^2 + b^2)} 
\end{align}

### Цилиндр.

<br>

\begin{align}
    & I_z = \int(x^2 + y^2)\mathrm{d}m = \int \int \dfrac{M}{\pi R^2} \cdot (x^2 + y^2) dx dy = \int \dfrac{M}{\pi R^2} \cdot r^2 d\phi dr \cdot r = \dfrac{M}{\pi R^2} \cdot 2\pi \dfrac{R^4}{4} = \dfrac{MR^2}{2} 
    \\
    & \text{У цилиндра: } I_z = \dfrac{MR^2}{2}. 
\end{align}

<br>


### Шар (относительно оси).

<br>

\begin{align}
    & I_o = \int r^2 \mathrm{d}m = \int r^2 \dfrac{d\phi \cdot rd\omega \cdot r sin\omega dr M}{\dfrac{4}{3} \pi R^3} = 2\pi \cdot 2 \int r^4 dr \dfrac{m}{\dfrac{4}{3} \pi R^3} = \dfrac{3}{5} MR^2 \\
    & \Rightarrow 2I_o = I_x + I_y + I_z = 3I_z
    \\
    & I_z = \dfrac{2}{3}I_o = \dfrac{2}{5} MR^2. 
\end{align}

<br>

### Сферы:

<br>

\begin{align}
    I_o = MR^2 \; \Rightarrow \; I_z = \dfrac{2}{3}MR^2. 
\end{align}

Переходим к динамическим уравнениям. 

Для системы материальных точек мы уже получали:

Уравнение движения центра масс: $m\cdot\dfrac{d\vec{v_c}}{dt} = \sum \overset{ext}{F_i}$

Уравнение на моменты: $\dfrac{d\vec{L}}{dt}= \sum \overset{ext}{M_i}$. 


Про движение центра масс всё понятно! (Используем теорему о движении центра масс абсолютно твёрдого тела.)\\
Уравнение на моменты можно записывать относительно либо неподвижной точки (выбрали удобную ИСО), либо относительно центра масс. Но скорости всё ещё берём в ИСО!!!
Почему?

$\left(\vec{M}_i \text{ ― момент внешних сил, действующих на мат. i-тую точку}\right)$

Если писать основной закон динамики вращательного движения АТТ относительно центра масс, то и скорости можно брать относительно центра масс. Для того, чтобы это проверить, найдём связь между моментами импульса АТТ в СО центра масс и обычным моментом импульса АТТ относительно неподвижной точки : 

<br>

\begin{align}
    &\vec L=\sum_{i} [\vec r_i,\vec v_i]\mathrm{d}m, 
    &\vec v_i=\vec V_c+\vec v_{ic}, \\
    &\vec r_i=\vec R_c+\vec r_{ic}, \\
    &\vec L=\sum_{i} [(\vec R_c+\vec r_{ic}), (\vec V_c+\vec v_{ic})]\mathrm{d}m \nonumber \\
    &= \sum_{i} [\vec R_c, \vec V_c]\mathrm{d}m+\sum_{i} [\vec R_c, \vec v_{ic}]\mathrm{d}m \nonumber \\
    &+ \sum_{i} [\vec r_{ic}, \vec V_c]\mathrm{d}m+\sum_{i} [\vec r_{ic}, \vec v_{ic}]\mathrm{d}m, \\
    &\int \limits_M \vec v_{ic}\mathrm{d}m=M\vec{v'}_c=0,  \\
    &\int \limits_M \vec r_{ic}\mathrm{d}m=M\vec{r'}_c=0,
\end{align}

<br>

так как $M\vec{v'}_c$ ― скорость центра масс АТТ в СО центра масс, $M \vec{r'}_c$ ― радиус-вектор центра масс АТТ в СО центра масс. 
Таким образом,

<br>

\begin{align}   
    &\vec L=\vec L_c+M\vec R_c\times\vec V_c, \quad \dfrac{d\vec L}{dt}=\sum_{i} \vec M_i, \\
    &\sum_{i} \vec M_i=\sum_{i} [\vec r_i, \vec F_i]=\sum_{i} [(\vec R+\vec r_{ic}), \vec F_i]=\vec R\times \sum_{i} \vec F_i+\sum_{i}[\vec r_{ic}, \vec F_i], \\
    &\dfrac{d\vec L}{dt}=\dfrac{d\vec L_c}{dt}+M\vec V_c\times \vec V_c+\vec R_c\times \vec F_{sum}.
\end{align}

<br>

В итоге: 

<br>

\begin{align}
    &\dfrac{d\vec L}{dt}=\dfrac{d\vec L_c}{dt}+\vec R_c\times \vec F_{sum}=\sum_{i} \vec M_i=\vec R\times \sum_{i} \vec F_i+\sum_{i}[\vec r_{ic}, \vec F_i], \\
    &\dfrac{d\vec L_c}{dt}=\sum_{i}[\vec r_{ic}, \vec F_i]=\sum_{i}\vec M_{ic}.
\end{align}

<br>

\begin{align}
    &\text{Пусть } \vec{L}_i = \left[\vec{r}_i\;', m\vec{v}_i\right]\nonumber\\
    &\dfrac{\mathrm{d}\vec{L}_i}{\mathrm{d}t} = \left[\dfrac{\mathrm{d}\vec{r}_i\;'}{\mathrm{d}t}, m\vec{v}_i\right] + \left[\vec{r}_i\;', m\dfrac{\mathrm{d}\vec{v}_i}{\mathrm{d}t}\right] = \nonumber\\
    & = \left[\dfrac{\mathrm{d}}{\mathrm{d}t}\left(\vec{r}_i - \vec{R}\right), m\vec{v}_i\right]+\underbrace {\vec{r}_i\;' \times \vec{F}_i}_{\vec{M}_i} = \nonumber\\
    &=\vec{M}_i + (-1)\left[\vec{v}_0\times \vec{v}\right] \nonumber\\
    &\dfrac{\mathrm{d}\vec{L}_i}{\mathrm{d}t} = \vec{M}_i - \left[\vec{v}_0 \times m\vec{v}_i\right]\nonumber\\
    &\boxed{\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \sum_i \dfrac{\mathrm{d}\vec{L}_i}{\mathrm{d}t} = \sum_i \vec{M}_i + \vec{v}_0 \times {\sum_i\vec{p}_i}},\\
    &\text{Если в центре масс: } \implies \boxed{\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \vec{M}}.
\end{align}
