# Приливы и отливы. Принцип эквивалентности.

## Приливы
1) **Амплитуда прилива** – разность уровней большой и малой воды;

2) Время между двумя приливами $\sim$ 12:25 (половина периода видимого обращения Луны вокруг Земли);

3) Основной эффект от Луны.

Найдём эффект в нулевом приближении.

Приливное ускорение в точке находится путём вычитания гравитационного ускорения центра тела и гравитационного ускорения в точке рассмотрения.

<br>

\begin{align}
    &\vec{F} = -\gamma \dfrac{mM}{R^2} \vec{e}_{r} \nonumber\\
    &\implies \vec{a} = -\gamma\dfrac{M}{R^2}\vec{e}_r.
\end{align}

<br>

\begin{align}
    &\vec{a}_{B,A} = -\gamma \dfrac{M}{\left(R\pm\Delta r\right)^2}\vec{e}_r = -\gamma \dfrac{M}{R^2\left(1\pm\dfrac{\Delta r}{R}\right)^2}\vec{e}_r;\\
    &\vec{a}_{B} = -\gamma\dfrac{M}{R^2}\vec{e}_r\left(1+2\dfrac{\Delta r}{R}\right) + \ldots = -\gamma \dfrac{M}{R^2}\vec{e}_r - 2\gamma \dfrac{M}{R^2} \dfrac{\Delta r}{R}\vec{e}_r;\\
    &\vec{a}_{A} = -\gamma\dfrac{M}{R^2}\vec{e}_r\left(1-2\dfrac{\Delta r}{R}\right) + \ldots = -\gamma \dfrac{M}{R^2}\vec{e}_r + 2\gamma \dfrac{M}{R^2} \dfrac{\Delta r}{R}\vec{e}_r.
\end{align}

<br>

Замечание:

Прилив происходит не в кульминации  Луны (начальные скорость и координата). Аналогия с пружинкой в поле тяготения. Всё определяется поступательными силами инерции и силой тяготения Луны. Найдём формулу более строго:

<br>

    </center>
</figure>

<br>

\begin{align}
    \vec{g} = -\gamma\dfrac{M}{R^2}\vec{e}_r,
\end{align}

<br>

Проверим знак:

<br>

\begin{align}
    &\varphi = - \gamma\dfrac{M}{R},\nonumber \\
    &-\nabla \varphi = - \gamma \dfrac{M}{R^2}\vec{e}_r.
\end{align}

<br>

Запишем соотношение на потенциалы:

<br>

\begin{align}
    &\varphi = \varphi_\text{л} + \varphi_\text{ин},
\end{align}

<br>

Найдём $\varphi_\text{ин}$:

<br>

\begin{align}
    &\varphi_\text{ин} = a_0r\cos\theta,\nonumber \\
    &\vec{F}_\text{ин} = -m\vec{a}_0 = -ma_0\vec{e}_z,\nonumber \\
    &g = -a_0 \vec{e}_z = - \dfrac{d}{dz}\varphi_\text{ин}\vec{e}_z \nonumber \\
    &\Rightarrow \varphi_\text{ин} = a_0z = a_0r\cos\theta.
\end{align}

<br>

Найдём $\varphi_\text{л}$:

<br>

\begin{align}
&\varphi_\text{л} = - \gamma\dfrac{M_\text{л}}{R} = -\gamma\dfrac{M_\text{л}}{\sqrt{(R_\text{зл}-r\cos\theta)^2+(r\sin\theta)^2}} = -\gamma\dfrac{M_\text{л}}{\sqrt{R_\text{зл}^2+r^2-2R_\text{зл}r\cos\theta}} = \nonumber \\
&= -\gamma\dfrac{M_\text{л}}{R_\text{зл}}\left[1+ \dfrac{r\cos\theta}{R_\text{зл}} - \dfrac{1}{2}\dfrac{r^2}{R_\text{зл}^2} +\dfrac{3}{8}\left(\dfrac{2r\cos\theta}{R_\text{зл}}\right)^2+\dots \right].
\end{align}

<br>

Введём обозначение: 

<br>

\begin{align}
    x = \dfrac{-2R_\text{зл}r\cos\theta}{R_\text{зл}^2} + \dfrac{r^2}{R_\text{зл}^2}.
\end{align}

<br>

Тогда

<br>

\begin{align}
    (1+x)^p = 1 + px + \dfrac{p(p-1)x^2}{2!} + O(x^3) = 1 - \dfrac{1}{2}\left(\dfrac{r^2}{R_\text{зл}^2} - 2\dfrac{r\cos\theta}{R_\text{зл}}\right) + \dfrac{3}{8}\left(\dfrac{r^2}{R_\text{зл}^2} - 2\dfrac{r\cos\theta}{R_\text{зл}}\right)^2+ \dots.
\end{align}

<br>

Пусть $\varphi_0 = 0, \ \ -\gamma M_\text{л}/R_\text{зл}^2 = a_0$:

<br>

\begin{align}
    &\varphi = \varphi_0 - \gamma\dfrac{M_\text{л}}{R_\text{зл}}\cdot\dfrac{2r\cos\theta R_\text{зл}-r^2}{2R_\text{зл}^2} - \dfrac{3}{8}\gamma\dfrac{M_\text{л}}{R_\text{зл}}\left(\dfrac{2r\cos\theta}{R_\text{зл}}\right)^2 = \nonumber \\
    & = - \gamma \dfrac{M_\text{л}}{R_\text{зл}^2}r\cos\theta + \gamma\dfrac{M_\text{л}}{2R_\text{зл}^3}r^2 - \dfrac{3}{8}\left(\dfrac{r^2}{R_\text{зл}^2} - 2\dfrac{r\cos\theta}{R_\text{зл}}\right)^2 + \dots
\end{align}

<br>

\begin{align}
    &\varphi_\text{л} + \varphi_\text{ин} = a_0 r cos\theta - a_0r\cos\theta, \nonumber \\
    &\varphi = \varphi_\text{пр} = - \dfrac{3}{8} \gamma \dfrac{M_\text{л}}{R_\text{зл}^2} \dfrac{r^2}{R_\text{зл}}2\cdot(2\cos^2\theta - 1) = -\dfrac{3}{8} \gamma \dfrac{M_\text{л}}{R_\text{зл}^2}\cdot \dfrac{r^2}{R_\text{зл}}2\cos{2\theta}
\end{align}

<br>

## Гравитационная масса
Масса вводится как мера инерции: $m^i$.

Тела могут возбуждать гравитационное поле: $m^g$

Сила взаимодействия гравитационных масс обратно пропорциональная квадрату расстояния:

<br>

\begin{align}
    F = c \dfrac{m_1^g m_2^g}{r^2}.
\end{align}

<br>

Фиксируем $c \implies$ находим способом из лекции 9.

Закон физики, предложенный Ньютоном, заключается в пропорциональности $F$ инертным массам $\implies m^g \sim m^i$.

Выбором единиц мы можем добиться $m^g = m^i$.

Откуда это можно видеть?

<br>

\begin{align}
    &m^i\vec{a} = \vec{F} = m^g\vec{g} \implies\nonumber\\
    &\implies \vec{a} = \dfrac{m^g}{m^i} \vec{g}.
\end{align}

<br>

Наблюдаем, что $\vec{a} = \vec{g}$ (в одном и том же гравитационном поле все тела приобретают одинаковое ускорение). 

Это равенство проверяется и иначе: если верить Ньютону и Бесселю, период математического маятника (длинного стержня, на котором располагается тяжёлый груз, много массивнее стержня) удовлетворяет следующему выражению:

<br>

\begin{align}
    &T = 2\pi \sqrt{\dfrac{l m^i}{g m^g}}.
\end{align}

<br>

Меняя материал груза, мы замечаем, что период при этом никак не меняется: следовательно, с точностью примерно $10^{-5}\ \dfrac{\text{м}}{\text{с}^2}$ получаем $m^i = m^g$.

Этвеш Роланд в 1887 проверил равенство инерционной и гравитационной массы с точность до $10^{-9}\ \dfrac{\text{м}}{\text{с}^2}$ в опыте с крутильными весами: тело, ускоряющееся вблизи поверхности Земли, в НИСО, связанном с Землёй, ускоряется из-за двух сил: реальной силы гравитации и фиктивной силы инерции. Если бы инерционная и гравитационная массы не были бы равны, то с перемещением по поверхности Земли менялось бы направление общего ускорение свободно падающего тела.

<br>

    <figcaption>Рис. 15.5: Силы, действующие на тело вблизи поверхности Земли в НИСО Земли</figcaption>

<br>

Равновесие на крутильных весах достигается при следующем соотношении масс двух тел:

<br>

\begin{align}
    &m_1^g g - m_1^i \omega^2 r \cos \theta = m_2^g g - m_2^i \omega^2 r \cos \theta. \\
    &m_1^i \left(\underbrace{\dfrac{m_1^g}{m_1^i}}_{\alpha_1}g - \underbrace{\omega^2 r \cos \theta}_f \right) = m_2^i \left(\underbrace{\dfrac{m_2^g}{m_2^i}}_{\alpha_2}g - \underbrace{\omega^2 r \cos \theta}_f \right). \\
\end{align}

<br>

Если соотношение гравитационная масса не равна инерционной, то из $\alpha_1 \neq \alpha_2$ следует $m_1^i \neq m_2^i$. Положим $M = (m_1^i - m_2^i) \dfrac{l}{2} \omega^2 r \sin \theta$.

<br>

\begin{align}
    &f \varphi = M. \\
    &\varphi = \dfrac{M}{f}.
\end{align}

<br>

Но во всех случаях $\varphi = 0$!!! Значит, гравитационная масса равна инерционной массе.

## Принцип эквивалентности

**Формулировка :** Все физические явления в гравитационном поле происходят совершенно так же, как и в соответствующем поле сил инерции, если напряжённости обоих полей в соответствующих точках пространства совпадают, а начальные условия одинаковы для всех тел замкнутой системы.

Уже говорили, что все свободно падающие тела на земле приобретают одинаковое ускорение, следовательно то же самое выполняется и для ускорения относительно неинерциальной системы отсчёта. Рассмотрим пример с лифтом:
