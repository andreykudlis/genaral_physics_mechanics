# Законы Кеплера (часть 2)

Математика нам говорит, что для таких уравнений траекторией будет сечение конуса:

<br>

\begin{align}
    \rho \left(1 - \varepsilon \cos{\varphi}\right) = p,
\end{align}

<br>

где $\varepsilon \equiv e$, $\varepsilon$ -- эксцентриситет; $p$ -- фокальный параметр.
Подберём такие $p$ и $\varepsilon$, чтобы наша траектория удовлетворяла предыдущему уравнению.

<br>

<figure>
    <center>
    <img src="pics/8_dif_vectors.png" width="400" align="center"/>
    </center>
</figure>

<br>

$e > 1$ - гипербола;

$e = 1$ - парабола;

$e < 1$ - эллипс;

$e = 0$ - окружность.

<br>

\begin{align}
    &\dfrac{d}{dt}\left[\rho\left(1-e\cos\varphi\right)\right]=\dot{\rho}\left(1-e\cos\varphi\right)-\rho e (-\sin\varphi)\dot{\varphi} = 0, \nonumber \\
    &\dot{\rho}p + \dfrac{L}{m}e\sin\varphi = 0, \nonumber \\
    &\dfrac{d}{dt}\left[\dot{\rho}p + \dfrac{L}{m}e\sin\varphi\right]=\ddot{\rho}p+\dfrac{L}{m}e\cos{\varphi}\dot{\varphi}=0.
    \end{align}

<br>

Учитывая, что $e\cos{\varphi}=1-p / \rho$, имеем:

<br>

\begin{align}
    &\ddot{\rho}p+\dfrac{L}{m}\left[1-\dfrac{p}{\rho}\right]\dot{\varphi}=0 \nonumber\\
    & \Rightarrow p\ddot{\rho}-\dfrac{m\rho^2 \dot{\varphi}^2}{m}\cdot\dfrac{p}{\rho}\dot{\varphi}=-\dfrac{L}{m}\cdot \dot{\varphi}\cdot\dfrac{\rho^2m}{\rho^2m},\nonumber\\
    &p\ddot{\rho}-\rho \dot{\varphi}^2p =-\dfrac{1}{\rho^2}\left(\dfrac{L}{m}\right)^2, \nonumber \\
    &p\left[\ddot{\rho}-\rho\dot{\varphi}^2 \right] = -\dfrac{1}{\rho^2} \cdot \left(\dfrac{L}{m}\right), \text{где} \left[\ddot{\rho}-\rho\dot{\varphi}^2 \right] = -\gamma \dfrac{M}{\rho^2}, \left(\dfrac{L}{m}\right)^2 = const,\nonumber \\
    &\Rightarrow p\left(-\gamma\dfrac{M}{\rho^2}\right)=-\dfrac{1}{\rho^2}\cdot\left(\dfrac{L}{m}\right)^2 \nonumber \\
    &\Rightarrow \boxed{p=\dfrac{1}{\gamma M}\cdot\left(\dfrac{L}{m}\right)^2} .
\end{align}

<br>

Откуда найти $e$ ? 
Перепишем полную энергию:

<br>

\begin{align}
    &E=\dfrac{mv^2}{2}+U(\rho)=\dfrac{m}{2}\left[\dot{\rho}^2+\rho^2\dot{\varphi}^2\right]-\dfrac{\gamma m M}{\rho} \nonumber \\
    &\Rightarrow \dfrac{m}{2}\cdot \dot{\rho}^2 + \dfrac{L^2}{2m\rho^2} - \dfrac{\gamma m M}{\rho} = \dfrac{m}{2} \cdot \dot{\rho}^2+ U_\text{eff} = E.
\end{align}

<br>

$E_1$ - окружность ($\rho = const$);

$E=0 \Rightarrow$ парабола;

$E<0 \Rightarrow$ эллипс;

$E>0 \Rightarrow$ гипербола.

<br>

<figure>
    <center>
    <img src="pics/10_effective_energy.png" width="400" align="center"/>
    </center>
</figure>

<br>

В этих точках производная от $\rho$ по $t$ равна нулю:

<br>

\begin{align}
    &\dfrac{L^2}{2m\rho^2_\text{min,max}} - \dfrac{\gamma m M}{\rho_\text{min,max}} = E, \nonumber \\
    &\rho^2 + \dfrac{\gamma m M}{E}\rho - \dfrac{L^2}{2mE}=0.\nonumber \\
    &\rho_\text{min,max} = \dfrac{-\dfrac{\gamma m M}{E} \pm \dfrac{\gamma m M}{E} \sqrt{1+\dfrac{E^2 4 L^2}{2mE \gamma^2 m M^2}}}{2}, \\
    &\rho_\text{min, max} = -\dfrac{\gamma m M}{2E} \left( 1 \mp \sqrt{1 + \dfrac{2E}{m} \left( \dfrac{L}{\gamma m M} \right)^2} \right).
\end{align}

<br>

Выразим большую полуось через минимальное и максимальное расстояния:

<br>

\begin{align}
    &2a = \rho_\text{min} + \rho_\text{max} = - \dfrac{\gamma m M}{E}, \\
    &a = - \dfrac{\gamma m M}{2E}. \\
\end{align}

<br>

С другой стороны, из уравнения эллипса

<br>

\begin{align}
    &\rho = \dfrac{p}{1 - e \cos \varphi}.
    \\ \nonumber
\end{align}

<br>

Минимум достигается при $\cos \varphi = - 1$, максимум ― при $\cos \varphi = 1$:

<br>

\begin{align}
    &\rho_\text{min} = \dfrac{p}{1 + e} = \dfrac{p}{1 - e^2} (1 - e), \\
    &\rho_\text{max} = \dfrac{p}{1 - e} = \dfrac{p}{1 - e^2} (1 + e).
\end{align}

<br>

То есть,

<br>

\begin{align}
    &\dfrac{2p}{1-e^2} = \rho_\text{min} + \rho_\text{max} = 2a, \\
    &\dfrac{p}{a} = 1 - e^2, \\
    &e = \sqrt{1 - \dfrac{p}{a}} = \sqrt{1 + \dfrac{1}{\gamma M} \left( \dfrac{L}{m} \right)^2 \dfrac{2 E}{ \gamma m M}} \nonumber \\
    &\phantom{e } = \sqrt{1 + \dfrac{2E}{m} \left( \dfrac{L}{\gamma m M} \right)}.
\end{align}

<br>

Найдём и малую полуось эллипса $b$, применив немного геометрической смекалки.

<br>

\begin{align}
    &b^2 + c^2 = a^2. \\
    &\text{При этом, } c = ea. \\
    &b^2 + e^2 a^2 = a^2, \\
    &b = a\sqrt{1 - e^2}. \\
\end{align}

<br>

Подставим полученное чуть раньше выражение для эксцентриситета.
$$\require{enclose}$$

\begin{align}
    b = - \dfrac{\enclose{updiagonalstrike}{\gamma mM}}{2E} \sqrt{- \dfrac{2E}{m} \left( \dfrac{L}{\enclose{updiagonalstrike}{\gamma m M}} \right)^2 } = \dfrac{L}{\sqrt{-2 m E}}.
\end{align}

<br>

Если движение финитно, то довольно очевиден период обращения,

<br>

\begin{align}
    T = 2 \sqrt{\dfrac{m}{2}} \int \limits^{\rho_\text{max}}_{\rho_\text{min}} \dfrac{dr}{\sqrt{E - U_\text{eff}}}.
\end{align}

<br>

Однако, брать этот интеграл очень тоскливо и неприятно, гораздо легче воспользоваться постоянством секториальной скорости.

<br>

\begin{align}
    &|\vec{\sigma}| = \text{const} = \dfrac{S_\text{эллипс}}{T} = \dfrac{L}{2m}. \\
    &S_\text{эллипс} = \pi ab. \\
    &\dfrac{\pi ab}{T} = \dfrac{L}{2m}, \\
    &T = 2\pi \dfrac{mab}{L}. \\
\end{align}

<br>

Пока что мы рассматриваем частный случай некого тела, которое вращается вокруг массивного объекта ― попробуем вывести какую-то характеристику, которая будет одинакова для всех таких тел, которые вращаются вокруг одного и того же объекта.
$$\require{enclose}$$

<br>

\begin{align}
    &b = a \sqrt{1 - e^2} = a \sqrt{- \dfrac{2E}{m} \left( \dfrac{L}{\gamma m M} \right)^2}, \\
    &b^2 = a^2 \cdot \left( - \dfrac{2E}{\gamma m M} \right) \cdot \dfrac{L^2}{ \gamma m^2 M} = a^{\enclose{updiagonalstrike}{2}} \cdot \dfrac{1}{\enclose{updiagonalstrike}{a}} \cdot \dfrac{L^2}{\gamma m^2 M}. \\
    &T^2 = 4 \pi^2 \dfrac{m^2 a^2 b^2}{L^2} = 4 \pi^2 \dfrac{\enclose{updiagonalstrike}{m^2} a^2}{\enclose{updiagonalstrike}{L^2}} \cdot a \dfrac{\enclose{updiagonalstrike}{L^2}}{\gamma \enclose{updiagonalstrike}{m^2} M}. \\
    &\dfrac{T^2}{a^3} = \dfrac{4 \pi^2}{\gamma M} = K \text{ ― постоянная Кеплера.}
\end{align}

<br>

В своё время Кеплер на основании экспериментальных данных (очень сильно ему помогли длительные наблюдения Тихо Браге за движением планет по небу) постулировал следующие три закона (мы только что вывели их теоретически):


   1. Все планеты движутся по эллиптическим траекториям, в одном из фокусов которых непременно находится Солнце. Траектории движения всех планет лежат в одной плоскости.
    
   2. Секториальная скорость постоянна: за равные промежутки времени каждая планета «заметает» равные площади секторов эллипса, отложенных через фокус, содержащий Солнце.
   
   3. Квадраты периодов обращения планет относятся так же, как кубы больших полуосей.
