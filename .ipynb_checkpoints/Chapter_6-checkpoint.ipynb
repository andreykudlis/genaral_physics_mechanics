# Потенциальная энергия. Приложение

Итак, если силы консервативны (т.е. работа не зависит от траектории), то работа таких сил может быть выражена через изменение некой функции - потенциальной энергии.

<br>

\begin{align}
    \delta A\;=\; \vec{F} \cdot d\vec{r} \;=\; F_x \cdot dx\;+\;F_y \cdot dy\;+\;F_z \cdot dz \;=\; -dU(x,y,z).
\end{align}

<br>

\begin{align}
    F_x\;=\;-\frac{\partial U}{\partial x};\;F_y\;=\;-\frac{\partial U}{\partial y};\; F_z\;=\;-\frac{\partial U}{\partial z} \nonumber \;\text{ - частные производные}.
\end{align}

<br>

\begin{align}
    \text{Короткое обозначение:}\; \vec{F}\;=\;-\nabla U \nonumber
\end{align}

<br>

Оператор "набла" или "градиент" показывает направление роста (наискорейшего) у функции.

На самом деле, градиент и набла - не одно и то же. Градиент - произведение вектора набла на какое-то скалярное поле, например $\vec E=-grad \varphi=-\nabla \varphi$. Набла - более общее понятие, чем градиент, наблу можно скалярно умножать на векторное поле, тогда получается дифференциальный оператор дивергенция, например $div \vec E=\nabla \vec E=\dfrac{\rho}{\varepsilon_{0}}$ - теорема Гаусса в СИ в вакууме в дифференциальной форме. Более того, наблу можно векторно умножать на какое-то векторное поле, тогда получается дифференциальный оператор ротор, например $rot \vec E=\nabla \times \vec E=-\dfrac{\partial \vec B}{\partial t}$ - закон электромагнитной индукции в дифференциальной форме в СИ
<br>

**Пример.**

Сферически симметрический потенциал: $u\;=\; \alpha \cdot r^k$

<br>

\begin{align}
    \Rightarrow \; \vec F=-\nabla u \; =- \; \alpha \cdot k \cdot  r^{k-2} (x \cdot \vec{e}_x \; + \; y \vec{e}_y \; + \; z \vec{e}_z). \nonumber
\end{align}

<br>

\begin{align}
    -F_{x}=\frac{\partial}{\partial x} (\alpha r^k)\;=\;\frac{\partial}{\partial x} \alpha (x^2 + y^2 + z^2)^{\frac{k}{2}}\;=\; \alpha \cdot k \cdot r^{k - 1} \frac{\partial r}{\partial x} \;=\; \alpha k\cdot r^{k-1} \cdot \frac{x}{r}= \nonumber
\end{align}

<br>

\begin{align}
    = \alpha k \cdot x r^{k-2}. \nonumber
\end{align}

<br>

\begin{align}
    \vec{F}\;=\; -\nabla u\;=\;-\alpha kr^{k-2}\vec{r}\;=\;-\alpha kr^{k-1}\frac{\vec{r}}{r}\;=\;-\alpha kr^{k-1} \vec{e_r}. \nonumber
    \\ \nonumber
\end{align}

<br>

## Интегрирование уравнение движения в случае потенциальных сил

В некоторых случаях можно извлечь полезную информацию из уравнения движения без его полного решения.

Пусть $U(\vec r)$ $\;-\;$ достаточно гладкая.

\begin{align}
    U(\vec r) \Rightarrow m\ddot{r}\;=\;-\nabla U \;\Rightarrow m\ddot{x}\;=\;-\frac{\partial U}{\partial x}.
\end{align}

<br>

( Мы решили упростить себе задачу и вместо того, чтобы рассматривать трёхмерный случай, рассматриваем только одномерный, поэтому можем заменить наблу на частную производную потенциальной энергии по х)

В окрестности ($\cdot$) $x_0$:

<br>

\begin{align}
    U(x)\;=\;U(x_0)\;+\;\frac{dU}{dx}\bigg|_{x=x_0}(x\;-\;x_0)\;+\;\frac{d^{2}U}{2! \cdot dx^2} \bigg|_{x=x_{0}} (x\;-\;x_0)^2 + O((x\;-\;x_0)^3).
\end{align}

<br>

(Используется разложение функции в ряд Тейлора в окрестности $x_{0}$) 

Если мы рассмотрим систему в окрестности минимума, то:
$\frac{dU}{dx} |_{x=x_0} = 0 \;\Rightarrow$

<br>

\begin{align}
    U(x)= \dfrac{1}{2!}\cdot       \dfrac{d^2U}{dx^2}|_{x=x_0} (x-x_0)^2 \;\Rightarrow \nonumber\\
    U(x) \approx \dfrac{1}{2} k(x-x_0)^2 \Rightarrow m\ddot{x} = -k(x-x_0).
\end{align}

\begin{align}

В системе задана полная механическая энергия. Пусть у материальной точки будет энергия $E$, а в системе действуют только консервативные силы, то есть $E=T+U(x)$. 
Так как $T\geq 0 \Rightarrow E \geq U(x)$ - получаем неравенство на $x$. 

Если $U(x)\leq E$ для любого $x$, то движение будет инфинитным;
<br>
Если же есть какой-то $[a,b]$, то движение будет финитным. 


<br>

<figure>
    <center>
    <img src="pics/6_energy.png" width="400" align="center"/>

    </center>

<br>

$E_1$ $\;-\;$ инфинитное, $E_2$ $\;-\;$ финитное.
<br>
Рассмотрим уравнение движения:

<br>

\begin{align}
    m\frac{d}{dt}v_x = -\frac{\partial  U}{\partial  x},
\end{align}

<br>

однако, можем использовать интеграл движения:

<br>

\begin{align}
    & E = U(x) + T.\\
    & m\frac{\dot{x}^2}{2} + U(x) = E,\\
    & \frac{m}{2}\left(\frac{dx}{dt}\right)^2 = E - U(x) \Rightarrow \nonumber\\
    & \Rightarrow \frac{dx}{dt} = \sqrt{\frac 2m \left(E-U(x)\right)} \nonumber \Rightarrow\\
    & \Rightarrow \int^{x_2}_{x_1} \frac{dx}{\sqrt{\frac 2m \left(E-U(x)\right)}} = \int^{T/2}_0dt,
\end{align}

<br>

$x_2, x_1$ $\;-\;$ точки поворота $U(x_{1,2}) = E.$
<br>
Либо для произвольного пр-ка:

<br>

\begin{align}
    \int^{x_2}_{x_1} \frac{dx}{\sqrt{\frac 2m \left(E-U(x)\right)}} = t_2 - t_1.
\end{align}

<br>

Таким образом, период финитного движения:

<br>

\begin{align}
    T = 2\sqrt{\frac{m}{2}} \int^{x_2}_{x_1}\frac{dx}{\sqrt{E-U(x)}}.
\end{align}

<br>

Самый простой пример: 

<br>

<figure>
    <center>
    <img src="pics/6_u(x).png" width="400" align="center"/>
        <figcaption>Рис. 6.5: Зависимость энергии от координаты.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &E=\dfrac{kx_{\text{max}}^2}{2},\;E=-U(x)=\dfrac{k}{2}(x_{\text{max}}^2-x^2),\nonumber\\
    &\Rightarrow T=2\sqrt{\dfrac{m}{2}}\int \limits  _{-x_{\text{max}}} ^{x_{\text{max}}} \dfrac{dx}{\sqrt{x^2_{\text{max}}-x^2}}\sqrt{\dfrac{2}{k}} 
    \end{align}

<br>

Пусть $x=x_{\text{max}}\cos\alpha \Rightarrow dx=-x_{\text{max}} \sin \alpha d\alpha$:

<br>

\begin{align}
    2\sqrt{\dfrac{m}{2}}\int \limits  _{-x_{\text{max}}} ^{x_{\text{max}}} \dfrac{dx}{\sqrt{x_{\text{max}}^2-x^2}}\sqrt{\dfrac{2}{k}} = 2 \sqrt{\dfrac{m}{k}}\left(-\int \limits _{\pi}^0\dfrac{x_{\text{max}}\sin{\alpha}d\alpha}{x_{\text{max}} \sin{\alpha}}\right)=2\pi\sqrt{\dfrac{m}{k}}.
    \\ \nonumber
\end{align}

## Столкновение частиц
Бывают упругие и неупругие столкновения. При упругом столкновении сохраняется механическая энергия. При неупругих - не сохраняется, переходя в другие виды энергии, например, в тепловую (внутреннюю). В процессе столкновения не рассматривают изменение потенциальной энергии внешних полей, более того, если мы не сталкиваемся с аномальными поведениями силы, то можно считать, что импульс системы сохраняется.

### Упругое столкновение двух материальных точек
Пусть скорости $v_{11}, v_{21}$ - скорости до столкновения, а скорости $v_{12}, v_{22}$ - скорости после столкновения. Запишем закон сохранения импульса и закон сохранения энергии:
    
<br>

\begin{align}
    \begin{cases}
    m_1v_{11}+m_2v_{21} = m_1v_{12}+ m_2v_{22}, \\ 
    m_1v_{11}^2 + m_2v_{21}^2= m_1v_{12}^2+m_2v_{22}^2.
    \end{cases}
\end{align}

<br>

Постановка задачи: пусть нам известны $v_{11}, v_{21}$, следовательно должны быть две пары решений, одно из которых тривиальное: $v_{12}=v_{11}$ и $v_{22}=v_{21}$. Теперь находим другое (физическое):

<br>

\begin{align}
& \begin{cases}
        m_1(v_{11}-v_{12})=m_2(v_{22}-v_{21})\\
        m_1(v_{11}^2-v_{12}^2)=m_2(v_{22}^2-v_{21}^2)
\end{cases}
        \Rightarrow \nonumber\\
       & \Rightarrow m_1(v_{11}-v_{12})(v_{11}+v_{12})=m_2(v_{22}-v_{21})(v_{22}+v_{21}) \Rightarrow \nonumber \\ 
        & \Rightarrow v_{11}+v_{12}= v_{22}+v_{21}. 
    \end{align}

<br>

Отсюда:

<br>

\begin{align}
        v_{21}=v_{11}+v_{12}-v_{22} \  \text{или} \  v_{22}=(v_{11}-v_{21})+v_{12}. 
    \end{align}

<br>

Подставим $v_{22}$ в закон сохранения импульса:

<br>

\begin{align}
    &m_1v_{11}+m_2v_{21}=m_1v_{12}+m_2[(v_{11}-v_{21})+v_{12}] \nonumber \\ 
    & v_{12}=\dfrac{m_1v_{11}+2m_2v_{21}-m_2v_{11}}{m_1+m_2} = 2V_{\text{c.m.}}-v_{11}. \\
    &v_{12} = 2 V_{\text{c.m}} -  v_{11} \\
    &v_{22} = 2 V_{\text{c.m}} -  v_{21}. \\
    \end{align}

<br>

Если бы всё это время мы работали в системе отсчёта, связанной с центром масс, то, очевидно, выполнялось бы

<br>

\begin{align}
        &v'_{12} = - v'_{11} \nonumber\\
        &v'_{22} = - v'_{21}. \nonumber
\end{align}

<br>

Возьмём наше решение для общего случая

<br>

\begin{align}
        &v_{12} = 2\ \dfrac{m_1 v_{11} + m_2 v_{21}}{m_1 + m_2} - v_{11} \\
        &v_{22} = 2\ \dfrac{m_1 v_{11} + m_2 v_{21}}{m_1 + m_2} - v_{21},
    \end{align}

<br>

После чего рассмотрим некоторые частные ситуации:

<br>

1. Рассмотрим столкновение тела с неподвижной стенкой очень большой массы.

<br>

\begin{align}
        &v_{21} = 0 \nonumber\\
        &m_2 \to \infty. \nonumber\\ 
        &V_\text{c.m} \approx 0 \nonumber\\
        &v_{12} = -v_{11} \nonumber\\
        &v_{22} = 0.\nonumber
\end{align}

<br>

И правда ― после столкновения сама стенка не сдвинулась, а вот небольшое тело отразилось с той же по величине скоростью.

2. Рассмотрим столкновение одинаковых частиц.

<br>

\begin{align}
        &m_1 = m_2 \nonumber\\
        &v_{11} = -v_{21}. \nonumber\\
        &V_\text{c.m} = 0 \nonumber\\
        &v_{21} = -v_{11} \nonumber\\
        &v_{22} = -v_{21}. \nonumber\\
\end{align}


### Упругое рассеяние
Если речь о рассеянии двух частиц, то задачу достаточно рассмотреть на плоскости. Для лучшего понимания физического смысла воспользуемся системой отсчёта, связанной с центром масс системы. Её общий импульс в такой системе будет нулевой, следовательно

<br>

\begin{align}
    &\vec{p}_{1c} + \vec{p}_{2c} = 0 \nonumber\\
    &\vec{p}_{11c} = -\vec{p}_{21c} \nonumber\\
    &\vec{p}_{12c} = -\vec{p}_{22c}. \nonumber
\end{align}

<br>

В отсутствии внешних сил общая кинетическая энергия системы сохранится.

<br>

\begin{align}
    &T_{11c} + T_{21c} = T_{12c} + T_{22c} \\
    &\dfrac{\vec{p}_{11c}^{\ 2}}{2m_1} + \dfrac{\vec{p}_{21c}^{\ 2}}{2m_2} = \dfrac{\vec{p}_{12c}^{\ 2}}{2m_1} + \dfrac{\vec{p}_{22c}^{\ 2}}{2m_2} \\
    &\dfrac{\vec{p}_{11c}^{\ 2}}{2m_1} + \dfrac{(-\vec{p}_{11c})^{\ 2}}{2m_2} = \dfrac{\vec{p}_{12c}^{\ 2}}{2m_1} + \dfrac{(-\vec{p}_{12c})^{\ 2}}{2m_2} \\
    &\vec{p}_{11c}^{\ 2} \left( \dfrac{1}{2m_1} + \dfrac{1}{2m_2} \right) = \vec{p}_{12c}^{\ 2} \left( \dfrac{1}{2m_1} + \dfrac{1}{2m_2}\right) \\
    &\vec{p}_{11c}^{\ 2} = \vec{p}_{12c}^{\ 2}.
\end{align}

<br>

Из чего, довольно очевидно,

<br>

\begin{align}
    &|\vec{p}_{11c}| = |\vec{p}_{12c}| \nonumber\\
    &|\vec{p}_{21c}| = |\vec{p}_{22c}|. \nonumber\\
    &|\vec{v}_{11c}| = |\vec{v}_{12c}| \nonumber\\
    &|\vec{v}_{21c}| = |\vec{v}_{22c}|. \nonumber
\end{align}

<br>

Рассмотрим $m_1 < m_2$:

<br>

<br>

\begin{align}
    &\vec{v}_c = \dfrac{m_1}{m_1 + m_2} \vec{v}_{11} \nonumber\\
    &\vec{v}_{11} = \vec{v}_c + \vec{v}_{11c}.
\end{align}

<br>

Отсюда выводим, что

<br>

\begin{align}
    &\vec{v}_{11c} = \dfrac{m_2}{m_1 + m_2} \vec{v}_{11} \nonumber\\
    &\vec{v}_{21c} = - \dfrac{m_1}{m_1 + m_2} \vec{v}_{11} \nonumber
\end{align}

<br>

При этом рассеяние может происходить на любой угол! Чтобы рассмотреть различные ситуации, достаточно «мысленно повертеть» вектор и убедиться, что все такие случаи удовлетворяют записанным уравнениям.

**Неупругие столкновения.**

Рассмотрим пример слипания:

<br>

\begin{align}
    m_1\cdot \vec{v}_1\;+\;m_2\cdot \vec{v}_2\;=\;(m_1\;+\;m_2)\cdot \vec{u}\; \Rightarrow \; \vec{u}\;=\;\frac{m_1\cdot \vec{v}_1\;+\;m_2\cdot \vec{v}_2}{m_1\;+\;m_2}. \nonumber
\end{align}

<br>

т.к. $T\;=\;T_{c.m.}\;+\;\frac{MV^2}{2}$, где $V\;=\;u$.

т.е. будет двигаться со скоростью центра масс.

<br>

\begin{align}
    \Delta T\;=\;\frac{m_1\cdot V_1^2}{2}\;+\;\frac{m_2\cdot V_2^2}{2}\;-\;\frac{(m_1\;+\;m_2)\cdot u^2}{2}\;=\;\frac{1}{2}\cdot \bigg (\frac{m_1m_2v_1^2\;+\;m_1m_2v_2^2\;-\;2m_1m_2\vec{v}_1 \vec{v}_2}{m_1\;+\;m_2} \bigg ) \nonumber
\end{align}

<br>

\begin{align}
    =\; \frac{1}{2}\cdot \mu \cdot (\vec{v}_1\;-\;\vec{v}_2)^2. \nonumber
\end{align}

<br>

А чему равна эта величина? С чем она совпадает?

До удара: $T_0\;=\;T_{c.m.0} \;+\;\frac{Mu^2}{2} \;(неравные \ нулю)$

После: $T\;=\;T_{c.m.} \;+\;\frac{Mu^2}{2}$, где $T_{c.m.}\;=\;0$.

Отсюда видна энергия, которая может перейти в тепло.

<br>

\begin{align}
    & \vec{v}=\vec{v'}+\vec{u},  \nonumber
    \\
    & T_{c.m}= \dfrac{m_1 \vec{v'_1} ^2}{2} + \dfrac{m_2 \vec{v'_2} ^2}{2} = \dfrac{m_1}{2}(\vec{v_1}-\vec{u})^2 + \dfrac{m_2}{2}(\vec{v_2}-\vec{u})^2 = 
    \nonumber \\ \nonumber
   & = \dfrac{1}{2} \left[m_1\left(\dfrac{m_1 \vec{v_1}+m_2 \vec{v_1}-m_1 \vec{v_1}-m_2 \vec{v_2}}{m_1+m_2}\right)^2 + m_2\left(\dfrac{m_1 \vec{v_2} + m_2 \vec{v_2} - m_1 \vec{v_1} - m_2 \vec{v_2}}{m_1 + m_2}\right)^2 \right] =
   \\ \nonumber
   & = \dfrac{1}{2} \left[m_1\left(\dfrac{m_2(\vec{v_1}-\vec{v_2})}{m_1+m_2}\right)^2 + m_2\left(\dfrac{m_1 (\vec{v_1} - \vec{v_2})}{m_1 + m_2}\right)^2 \right] = \dfrac{1}{2} \dfrac{m_1 m_2}{(m_1 + m_2)^2} \cdot (m_2 + m_1)(\vec{v_1}-\vec{v_2})^2 =
   \\ \nonumber
   & = \dfrac{1}{2} \dfrac{m_1 m_2}{m_1 + m_2} (\vec{v_1} - \vec{v_2})^2 = \dfrac{1}{2} \mu (\vec{v_1} - \vec{v_2})^2  \; \; \; \square 
\end{align}

<br>

То же самое. 
<br>
Таким образом, отдаётся вся кинетическая энергия в системе центра масс.
