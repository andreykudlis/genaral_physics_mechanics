## Аберрация света
**Аберрация света** $\;-\;$ изменение направления фронта плоской волны при смене СО.
<br>
Волновой вектор, характеризующий направление распространение волнового фронта, будем ассоциировать с движением луча (фотонов).

<br>

<br>
Свет идёт к нам в глаза со скоростью $c$. В системе отсчёта Солнца:

<br>

\begin{align}
    &v_x = -c\cos{(\varphi)},
    &v_x' = -c\cos{(\varphi')}.
\end{align}

<br>

Исходя из полученной в прошлой главе формулы для скорости $v_x'$:

<br>

\begin{align}
    &v_x' = \dfrac{v_x-V}{1-\dfrac{c\cos{(\varphi)}V}{c^2}} \implies\\
    &\implies\cos{(\varphi')} = \dfrac{\cos{(\varphi)} + \beta}{1+\beta\cos{(\varphi)}}.
\end{align}

<br>

Несложно высчитать, используя ряд Тейлора и вынеся $\cos\varphi$, что полученное выражение преобразуется до:

<br>

\begin{align}
    &\boxed{\cos\varphi' = \cos\varphi \left[1+\beta\left(\dfrac{1}{\cos\varphi} - \cos\varphi\right)\right] + ...}
\end{align}

<br>

В одном случае $\;-\;$ приближается, а в другом $\;-\;$ отдаляется.

Какова длина волны?
$V$ $\;-\;$ скорость источника относительно среды (>0, если удаляется и <0, если приближается). $T_1$ $\;-\;$ период испускания волн источником.
Пусть удаляется (как на рисунке):

<br>

\begin{align}
    &\lambda = T_1 c + VT_1 = T_1(c+V) \nonumber\\
    &\lambda = \dfrac{1}{\nu_0}(c+v).
\end{align}

<br>

На приёмнике регистрируем:

<br>

\begin{align}
    &\dfrac{1}{T} = \nu = \dfrac{c}{\lambda} = \dfrac{c}{T_1(c+V)},\nonumber\\
    &\boxed{T = T_1 \left(1+\dfrac{V}{c}\right)}.
\end{align}

<br>

В случае сближения: $T = T_1 \left(1 - \dfrac{V}{c}\right)$.
## СТО
$T_1$ $\;-\;$ это время процесса (испускания сигналов) в системе наблюдателя, т. е. замедленная версия того, что в СО источника:

<br>

\begin{align}
    &T_1 = \dfrac{T_0}{\sqrt{1 - \dfrac{v^2}{c^2}}} \implies \nonumber\\
    &\implies T = \dfrac{T_0\left(1 - \dfrac{V}{c}\right)}{\sqrt{1 - \dfrac{V^2}{c^2}}}.
\end{align}

<br>

Продольный эффект: $ T = \dfrac{T_0\left(1 - \dfrac{V}{c}\right)}{\sqrt{1 - \dfrac{V^2}{c^2}}}$.
Поперечный эффект: $T \;=\; \gamma T_0$.

## Пространство Минковского.
Итак, пространство Минковского - пространство событий точек $(ct, x, y, z)$ с длиной, задаваемой интервалом.
<br>
Траектория материальной точки в таком пространстве называется мировой линией. Обычно рассматривается одномерное движение (в осях $OX, O\tau (\tau = ct)$). Получаем диаграмму Минковского:
<br>
В штрихованной области любое событие связано с событием в начале координат времениподобными интервалами. Эти области называются абсолютным прошлым и абсолютным будущим.
<br>
Области точек называются абсолютно удалёнными областями.
<br>
Изобразим события в разных системах отсчёта. Для этого мы хотим изобразить на рисунке оси системы отсчёта $K\,'$, которая двигается со скоростью $V$ вдоль направления $X$ относительно системы отсчёта $K$.
<br>
Из преобразований Лоренца помним, что:

<br>

\begin{align}
    &x\,' = \gamma(x - \beta \tau), 
    <br>
    &\tau\,' = \gamma(\tau - \beta x)
\end{align}

<br>

Ось $OX$ задаётся простым уравнением: $\tau = 0$, ось $O\tau$ задаётся простым уравнением: $x = 0$. Аналогично ось $OX\,'$ задаётся уравнением $\tau\,'= 0$ и ось $O\tau\,'$ уравнением $x\,' = 0$. Таким образом, найдём оси системы отсчёта $K\,'$:

<br>

\begin{align}
    &\tau\,' = 0 \Rightarrow \tau = \beta x, 
    <br>
    &x\,' = 0 \Rightarrow \tau = \dfrac{1}{\beta} x
\end{align}

<br>

Изобразим их, учитывая, что $\beta = v/c < 1$.
<br>
Посмотрим, как ведёт себя единичный отрезок, который в свою очередь определяется интервалом:

<br>

\begin{align}
    \Delta S = \sqrt{\Delta\tau^2 - \Delta x^2}.
\end{align}

<br>

Рассмотрим событие $(0;0)$ и $(1;0)$: между ними $\Delta S = 1$, мы знаем, что он сохраняется.
<br>
Если мы захотим отметить единичную длину (отрезок) в системе отсчёта $K\,'$, то пользуемся интервалом. Например: 

<br>

\begin{align}
    \Delta S = 1 = \sqrt{1 - 0}.
\end{align}

<br>

Нанесём кривую, события которой будут отделены событием $(0;0)$ интервалом $\Delta S= 1$. Посмотрим её пересечение с прямой $\tau = 1/\beta x$.

<br>

\begin{align}
    &1 = \sqrt{\tau^2 - x^2} \Rightarrow \tau = \sqrt{1 + x^2}, \nonumber \\
    & \sqrt{1 + x^2} = \dfrac{1}{\beta}x \Rightarrow x^2 = \dfrac{\beta^2}{\sqrt{1-\beta^2}} \Rightarrow x = \beta \gamma, \ \ \tau = \gamma.
\end{align}

<br>

$(x;\tau)$ - точка пересечения. Точка соответствует также и единичной длине в системе отсчёта $K\,'$. Перепроверим:

<br>

\begin{align}
    &x\,' = \gamma(\beta\gamma - \beta\gamma) = 0, \nonumber \\
    &\tau\,' = \gamma(\gamma - \beta^2 \gamma) = \gamma^2(1-\beta^2) = \dfrac{\gamma^2}{\gamma^2} = 1.
\end{align}

<br>

Таким образом, мы нашли, где будет единичка системы $K\,'$, когда смотрим на оси в системе $K$.
