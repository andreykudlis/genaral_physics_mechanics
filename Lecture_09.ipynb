# Гравитационное взаимодействие. Продолжение

Добавим еще один факт: пусть наши силы будут (опять) центральными: $\vec{F}\sim \vec{r}$

<br>

\begin{align}
    \Rightarrow \vec{M} = [\vec{r}\times \vec{F}] = 0 \Rightarrow \vec{L}=const
\end{align} 

<br>

$\Rightarrow$ движение, как мы уже говорили, будет в плоскости, а помимо этого секториальная скорость будет непостоянной! То есть за равные промежутки времени радиус-вектор материальной точки в центральном поле, отложенный из силового центра, заметает равные площади. 

Теперь к закону всемирного тяготения.

$\sim$ 1666 Исаак Ньютон, $\sim$ 1687 опубликовано. 

<br>

\begin{align}
    \vec{F} = \gamma \cdot \dfrac{mM}{r^2} \cdot \dfrac{\vec{r}}{r}
\end{align}

<br>

$ \gamma \equiv G $  - гравитационная постоянная. 
 G = 6.67 $\cdot 10^{-11}$ [$\dfrac{M^3}{\text{кг} \cdot c}$]

$f$ – модуль кручения стержня (известен);
<br>
$\alpha$ – угол отклонения;
<br>
$M = f\alpha$ – момент сил, возникающий при скручивании.

<br>

\begin{align}
    &\implies f \cdot \alpha_0 \Bigl | \alpha_0 - \text{ измеряется} \quad =\nonumber\\
    &= 2G\dfrac{mM}{r^2} \left(\dfrac{l}{2}\right) \Bigl | m, M, r - \text{ измеряются} \quad = \nonumber\\
    &= G\dfrac{mM}{r^2}l.
\end{align}

<br>

Найдём значение силы через частоту колебаний данной системы: 

<br>

\begin{align}
    &I \frac{d^{2}\alpha}{dt^{2}}=-M=-f\alpha\
\end{align}

<br>

Перед моментом сил стоит знак -, так как момент сил, возникающий при скручивании нити, стремится вернуть стержень в первоначальное положение

<br>

\begin{align}
    &I=2m\left(\dfrac{l}{2}\right)^2
\end{align}

<br>

При выводе будет использоваться уравнение гармонических колебаний :

<br>

\begin{align}
    \ddot{x}+\omega^2x=0
\end{align}

<br>

**Для желающих разобраться, как это решается :** Чтобы его решить, надо искать решение вида $x=c*e^{\gamma t}$ -> подставляем в уравнение, получаем характеристическое уравнение $\gamma^2*c*e^{\gamma t}+\omega^2*c*e^{\gamma t}=0\Rightarrow \gamma_{1,2}=\pm i \omega$, а далее из начальных условий ищем решение вида $x=c_1*e^{\gamma_1t}+c_2*e^{\gamma_2t}$, это и будет всем знакомое уравнение типа $x=x_m*cos(\omega t+\varphi_0)$.

<br>

\begin{align}
    &2m{\left(\dfrac{l}{2}\right)^2}\ddot{\alpha} + f\alpha = 0 \implies\\
    &\implies \omega^2 = \dfrac{2f}{ml^2} \implies \\
    &\implies T =\dfrac{2\pi}{\omega}= 2\pi\sqrt{\dfrac{ml^2}{2f}}.
\end{align}

<br>

\begin{align}
    &\dfrac{T^2}{\left(2\pi\right)^2} = \dfrac{ml^2}{2f} \implies \nonumber\\
    &\implies f = \dfrac{ml^24\pi^2}{2T^2} \implies\\
    &\implies G = \dfrac{2\pi^2lr^2}{MT^2}\alpha_0. \qquad \blacksquare
\end{align}

<br>

Рассчитаем массу Земли $M_{\oplus}$:

<br>

\begin{align}
    & mg = G \dfrac{mM_{\oplus}}{R^2_{\oplus}} \implies\nonumber\\
    & \implies M_{\oplus} = \dfrac{gR^2_{\oplus}}{G} \implies M_{\oplus} = \dfrac{gR^2_{\oplus}MT^2}{2\pi^2lr^2\alpha_0}.
\end{align}

<br>

Рассчитаем среднюю плотность Земли, используя полученные значения её массы и принимая её форму за шар, считая радиус известным.

<br>

\begin{align}
    & \overline{\rho_{\oplus}} = \dfrac{M_{\oplus}}{\frac{4}{3}\pi R^3_{\oplus}} = \dfrac{3gMT^2}{8\pi^3R_{\oplus}lr^2}.\\
    & \overline{\rho_{\oplus}} \approx 5.48 \text{г/см}^3, \text{современные данные: 5.52 г/см}^3.\nonumber
\end{align}

## Законы Кеплера
Переходим к задаче Кеплера (получим все законы Кеплера при её решении).
<br>
**Задача Кеплера:** движение в потенциале: 

<br>

\begin{align}
    U = -\frac \alpha r,\hspace{1em} \vec F = - \nabla U = -\vec{e}_r \cdot \left(\frac{\alpha}{r^2}\right) = -\alpha\frac{\vec{e}_r}{r^2}, \alpha = GmM.
\end{align}

<br>

Движение плоское $\implies$ полярная система координат.
<br>
Силы потенциальны $\implies E = \text{const}$.
<br>
Силы центральные $\implies \vec{L}=\vec{e}_zL=\text{const}$.

<br>

\begin{align}
    & \dfrac{mv^2}{2} - \gamma\dfrac{mM}{r} = E.\\
    & |\vec{L}| = m |[\vec{r}\times\vec{v}]| = m[\vec{r}\times\vec{v}_r+\vec{v}_{\varphi}] = m\rho^2\dot{\varphi} =L_z = \text{const}.\\
    & \boxed{
    m\rho^2\dot{\varphi} =L_z = \text{const}.
    }
\end{align}

<br>

\begin{align}
    & \vec{r} = \vec{e}_\rho\rho \implies \vec{v} = \vec{e}_\rho\dot{\rho} + \rho\vec{e}_\varphi\dot{\varphi} = \vec{v}_\rho + \vec{v}_\varphi \nonumber\\
    & \vec{a} = \dot{\vec{v}} = \underline{\vec{e}_\rho\ddot{\rho}} + \dot{\rho}\dot{\vec{e}}_\rho + \dot{\rho}\dot{\varphi}\vec{e}_\varphi+\rho\ddot{\varphi}\vec{e}_\varphi + \underline{\rho\dot{\varphi}\dot{\vec{e}}_\varphi} = \nonumber\\
    & = \left(\ddot{\rho} - \rho\dot{\varphi}^2\right)\vec{e}_\rho + \left(2\dot{\rho}\dot{\varphi}+\rho\ddot{\varphi}\right)\vec{e}_\varphi = \vec{a}_\rho + \vec{a}_\varphi.
\end{align}

<br>

Посмотрим на $\vec{a}_\varphi$:

<br>

\begin{align}
    &\rho\ddot{\varphi} + 2\dot{\rho}\dot{\varphi} = \dfrac{1}{\rho} \dfrac{\mathrm{d}}{\mathrm{d}t} \left(\rho^2\dot{\varphi}\right) = \dfrac{1}{\rho}\left[2\rho\dot{\rho}\dot{\varphi}+\rho^2\ddot{\varphi}\right] = \nonumber\\
    & 2\dot{\rho}\dot{\varphi} + \rho\ddot{\varphi} = \dfrac{1}{\rho} \dfrac{1}{m}\dfrac{\mathrm{d}}{\mathrm{d}t}L_z = 0 \quad (\text{т. к. } L_z = \mathrm{const}) \implies \nonumber\\
    & \implies \vec{a}_\varphi = 0 \implies \vec{a} = \vec{a}_\rho – \text{ направление к Солнцу}. 
\end{align}

<br>

С другой стороны $\Bigl |$ здесь $\rho \equiv r$, т. к. $z$ = 0 $\Bigl |$:

<br>

\begin{align}
    &m\vec{a}=\vec{F} \implies \nonumber\\
    &\implies \vec{a}_\rho = - \dfrac{\gamma M}{r^2}\dfrac{\vec{r}}{r} \implies\nonumber\\
    &\implies \left[\ddot{\rho} - \rho\dot{\varphi}^2\right] = -\dfrac{\gamma M}{\rho^2}.
\end{align}
