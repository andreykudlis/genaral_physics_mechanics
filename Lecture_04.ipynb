## Вязкое трение

При движении в среде механические объекты испытывают трение.

При высоких скоростях $F_c \sim v^2$.

При низких скоростях $F_c \sim v $.

Рассмотрим баллистическое движение в присутствии сил сопротивления среды. 
   
   Пусть $\vec{F}_{\text{сопр}} =-k\vec{v}$, известен угол броска (запуска) - $\alpha$ и начальная скорость - $v_0$


<br>

1. Найдем связь динамических параметров с $v_\infty$:

Пусть из эксперимента известна скорость снаряда в установившемся режиме при свободном падении с большой высоты: $v_\infty$

Тогда:

<br>

\begin{align}
    &m\vec{a}=\vec{F}\\
    & m\vec{a}=m\vec{g}+\vec{F}_{\text{сопр}}.
\end{align}

**Вариант 1. Способ решения дифференциального уравнения**

<br>

\begin{align}
& m\int^{v_y(t)}_{v_0} \frac{d\tilde{v_y}}{mg-\tilde{kv_y}} = \int^t_0 d\tilde{t}\\

& \ln{\left|\frac{mg - kv_y(t)}{mg - kv_0} \right|} = -\frac{kt}{m} \\
& mg-kv_y(t)=(mg-kv_0)e^{-\frac{kt}{m}}\\
& v_y(t) = \frac{mg - (mg - kv_0)e^{-\frac{kt}{m}}}{k} = \frac{mg(1-e^{-\frac{kt}{m}})+kv_0e^{-\frac{kt}m}}{k}
\end{align}

<br>

Если $v_0$ = 0, то: 

<br>

\begin{align}
& \ln{\left| \frac{mg - kv_y(t)}{mg} \right|} = - \frac{kt}{m}\\
& \left(1-\frac k{mg}v_y(t)\right) = e^{\left(-\frac{kt}m\right)}\\
& v_y = \dfrac{mg}{k} \left(\ 1 - e^{-\frac{kt}{m}} \right) = v_\infty \left(\ 1 - e^{-\frac{kt}{m}} \right)\\
& \frac{mg}k = v_\infty\\
& \frac{g}{v_\infty} = \frac km\\
& \boxed{v_y(t)=v_\infty(1-e^{-\frac{gt}{v_\infty}})}
\end{align}

## Зависимость силы от координаты
Рассмотрим силу как функцию от координаты $F(x)$; тогда второй закон Ньютона выглядит как

<br>

\begin{align}
           &ma = F(x). \\
           &m \dfrac{\mathrm{d}v}{\mathrm{d}t} = F(x), \\
           &\text{Учитывая, что }\dfrac{\mathrm{d}x}{\mathrm{d}t} = v \implies \dfrac{\mathrm{d}x}{v} = \mathrm{d}t \\
           &m \dfrac{\mathrm{d}v}{\mathrm{d}x}v = F(x) \\
           &m v \mathrm{d}v = F(x)\mathrm{d}x.
\end{align}
        
<br>
       
Интегрируя обе части от значений скорости и координаты в начале движения до таковых в конце, получаем

<br>

\begin{align}
            &m \int \limits^v_{v_0} \tilde{v} \mathrm{d}\tilde{v} = \int \limits^x_{x_0} F(\tilde{x})\mathrm{d}\tilde{x} \\
            &m \left. \dfrac{\tilde{v}^2}{2} \right|^v_{v_0} = \int \limits^x_{x_0} F(\tilde{x})\mathrm{d}\tilde{x} \\
            &v^2 - v_0^2 = \dfrac{2}{m} \int \limits^x_{x_0} F(\tilde{x})\mathrm{d}\tilde{x} \\
            &v = \sqrt{v_0^2 + \dfrac{2}{m} \int \limits^x_{x_0} F(\tilde{x})\mathrm{d}\tilde{x}} \\
            &\text{где } v = \dfrac{\mathrm{d}x}{\mathrm{d}t} \\
            &\mathrm{d}t = \dfrac{\mathrm{d}x}{\sqrt{v_0^2 + \dfrac{2}{m} \int \limits^x_{x_0} F(\tilde{x})\mathrm{d}\tilde{x}}}. \\
        \end{align}
        
        <br>
        
Найдём время движения: для упрощения вычислений примем начальную скорость $v_0$ и время начала движения нулевыми, оставив начальную координату $x_0$ произвольной. Будем рассматривать конкретный случай, когда сила прямо пропорциональна координате ($F(x) = \gamma x$, где $\gamma > 0$).

<br>

\begin{align}
             &\int \limits^t_0 \mathrm{d}\tilde{t} = \int \limits^x_{x_0} \dfrac{\mathrm{d}\tilde{x}}{\sqrt{\dfrac{2}{m} \int \limits^x_{x_0} \gamma \tilde{\tilde{x}}\mathrm{d}\tilde{\tilde{x}}}} \\
             &t = \int \limits^x_{x_0} \dfrac{\mathrm{d}\tilde{x}}{\sqrt{\dfrac{2}{m}}\sqrt{\gamma \dfrac{\tilde{x}^2 - x_0^2}{2}}} \\
             &t = \int \limits^x_{x_0} \dfrac{\mathrm{d}\tilde{x}}{ \sqrt{\dfrac{\gamma}{m}} \sqrt{\tilde{x}^2 - x_0^2}} \\
             &t = \sqrt{\dfrac{m}{\gamma}}\int \limits^x_{x_0} \dfrac{\mathrm{d}\tilde{x}} {x_0 \sqrt{\dfrac{\tilde{x}}{x_0}^2 - 1}}
         \end{align}
         Обозначим $y = \dfrac{\tilde{x}}{x_0}$
         \begin{align}
             &t = \sqrt{\dfrac{m}{\gamma}} \int \limits^{\dfrac{x}{x_0}}_{1} \dfrac{\mathrm{d}y}{\sqrt{y^2 - 1}} \\
             &t = \sqrt{\dfrac{m}{\gamma}} \ln \left| \dfrac{x + \sqrt{x^2 - x_0^2}}{x_0} \right| .
        \end{align}

## Реактивное движение
Рассмотрим систему, части которой изменяют массу. На систему может действовать внешняя сила. 

<br>

Пусть $p(t)$ - импульс системы в начальный момент времени.

<br>

\begin{align}
    \vec{p}(t)=\vec{P}=M\vec{V}
\end{align}

<br>

\begin{align}
   & p(t+dt) = \vec{P} + d\vec{P}+ \vec{V}_m\cdot dm = \nonumber \\
    &= M\vec{V} + Md\vec{V} + \vec{V}dM + \vec{V}_mdm,
\end{align}

<br>

Где $(\vec{P} + d\vec{P})$ - изменение импульса ракеты, $\vec{V}_m$ - скорость относительно лабораторной системы отчёта.

<br>

\begin{align}
    \vec{p}(t+dt) - \vec{p}(t)= \vec{F}(t)dt
\end{align}

<br>

\begin{align}
    Md\vec{V}+\vec{V}dM+\vec{V}_mdm=\vec{F}(t)dt
\end{align}

<br>

Понятно, что $dm=-dM$, т.е. $m+M=const$

<br>

\begin{align}
    Md\vec{V}+(\vec{V}-V_m)dM=\vec{F}(t)dt
\end{align}

<br>

$\vec{V}\sim u$, $V_m \sim v$

<br>

\begin{align}
    MdV- \vec{V_{\text{отн}}}dM = \vec{F}(t)dt
    \end{align}
    \begin{align}
    \boxed{
        M(t)\dfrac{dV}{dt}=\vec{F}(t) + \vec{V}_{\text{отн}}\dfrac{dM}{dt}} \text{ – Уравнение Мещерского.}
    \end{align} 

<br>
    
\begin{align}
    \vec{v}=\vec{v}+\vec{u} \nonumber \\ 
    \Rightarrow \vec{v}\ '=\vec{v} - \vec{u} \nonumber
\end{align}

<br>

Рассмотрим решения данного уравнения.

1. Внешние силы отсутствуют и движение одномерно.

<br>

\begin{align}
    &M \dfrac{\mathrm{d}\vec{V}}{\mathrm{d}t} = 0 + \vec{V}_{\text{отн}} \dfrac{\mathrm{d}M}{\mathrm{d}t} \\
    &M \dfrac{\mathrm{d}V}{\mathrm{d}t} = - V_{\text{отн}} \dfrac{\mathrm{d}M}{\mathrm{d}t} \\
    &\mathrm{d}V = - V_{\text{отн}} \dfrac{\mathrm{d}M}{M}.
\end{align}

<br>

Интегрируя обе части от начального до конечного момента, получаем

<br>

\begin{align}
    &V_{\text{отн}} \ln \dfrac{M_{\text{кон}}}{M_{\text{нач}}} = -(V - V_0). \\
    &\boxed{ V = V_0 + V_{\text{отн}} \ln \dfrac{M_{\text{кон}}}{M_{\text{нач}}} } \text{ ― формула Циолковского.}
\end{align}

<br>

2. Движение происходит в поле силы тяжести, масса изменяется линейно: $M = M_0 (1 - \alpha t)$.
Найдём производную массы по времени и запишем уравнение Мещерского для этой ситуации:

<br>

\begin{align}
    &\dfrac{\mathrm{d}M}{\mathrm{d}t} = - M_0 \alpha \\
    &M_0 (1 - \alpha t) \dfrac{\mathrm{d}V}{\mathrm{d}t} = - M_0 (1 - \alpha t) g + V_\text{отн} M_0 \alpha \\
    &\mathrm{d}V = \Bigl(-g + \dfrac{V_\text{отн}\alpha}{1 - \alpha t} \Bigr) \mathrm{d}t
\end{align}

<br>

Интегрируя обе части от начального до конечного момента времени, получаем

<br>

\begin{align}
    &V - V_0 = -gt + V_\text{отн} \alpha \cdot \dfrac{1}{-\alpha}\ln \dfrac{1 - \alpha t}{1} \\
    &V = V_0 - gt + V_\text{отн} \ln \dfrac{1}{1 - \alpha t}
\end{align}
