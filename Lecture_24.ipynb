# Гармонические колебания. Фазовое пространство
В обычной ИСО, где они не вращаются, т. е. не те, что раньше, в любой момент времени $\vec{L} = \mathrm{const}$,

$\vec L$ можно расписать через мгновенную ось вращения! т. е. все точки тела будут $\sim \left[\vec{\omega}, \vec{r}_i\right]$.

У нас симметричный волчок, т. е. есть ось волчка и есть $I_z = I_\parallel, \;\; I_x = I_y = I_\perp$

<br>

$\vec{L}\left(\vec\omega\right)$  $\;-\;$ линейная функция $\vec{\omega} = \vec{\omega}_\parallel + \vec{\omega}_\perp \implies$

<br>

\begin{align}
    &\vec{L}\left(\vec{\omega}\right) = I_\parallel\vec{\omega}_\parallel + I_\perp\vec{\omega}_\perp = \mathrm{const}\\
    &\implies I_\parallel\vec{\omega}_\parallel + I_\perp\vec{\omega}_\perp = \mathrm{const}\nonumber\\
    &\implies \vec{\omega}_\parallel\;\text{и}\;\vec{\omega}_\perp = \mathrm{const}\;\text{длины!!!}
\end{align}

<br>

Кинетическая энергия такого вращения:

<br>

\begin{align}
    &T = \dfrac{1}{2}\vec{L}\vec{\omega} = \dfrac{1}{2}\left(I_\parallel\vec{\omega}^2_\parallel + I_\perp\vec{\omega}^2_\perp\right).
\end{align}

<br>

Ось фигуры, $\vec{L}, \vec{\omega}$ в одной плоскости. Вращение происходит вокруг $\vec{L}$.

## Вынужденная прецессия

**Гироскопом** называют быстро вращающееся тело. Ось вращения может изменять своё положение в пространстве. Точка опоры гироскопа $\;-\;$ точка закреплённая. В более общем смысле -- точка, относительно которой рассматривается движение гироскопа.

Продолжаем рассматривать приближенную теорию симметричного волчка.

При быстром вращении вдоль оси: 

$$\vec{L} \simeq I_\parallel\vec{\omega}_\parallel \simeq I_\parallel\vec{\omega}.$$

Пусть точка приложения силы лежит на оси гироскопа:

<br>

\begin{align}
    &\vec{M} = \left[\vec{a}, \vec{F}\right]\implies\nonumber\\
    &\implies \dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \left[\vec{a}, \vec{F}\right] = \vec{M},\;\text{т.е.}\; \dot{L} \perp L\nonumber\\
    &\implies\;\text{меняется только направление}\nonumber\\
    &\implies \dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \left[\vec{\Omega}\times\vec{L}\right]\;\;\left(\mathrm{d}\vec{L}=\left[\mathrm{d}\varphi_\varphi\times L\right]\right)\nonumber\\
    &\implies \left[\Omega\times\vec{L}\right] = \vec{M}\nonumber\\
    &\implies\Omega I_\parallel\omega = Fa\nonumber\\
    &\implies \boxed{\Omega = \dfrac{M}{I\parallel \omega}}.
\end{align}

## Колебания материальной точки

В общем смысле колебательное движение $\;-\;$ это движение, которое повторяется во времени. Наиболее распространенными колебаниями являются колебания, описывающиеся следующим уравнением:

<br>

\begin{align}
    &\ddot{x} + \omega^2x=0,\;\;\ddot{x} = \dfrac{\mathrm{d}^2}{\mathrm{d}t^2}x
\end{align}

<br>

Решение такого уравнения имеет следующий вид:

<br>

\begin{align}
    &x = A \cos{\omega t} + B\sin{\omega t}\\
    &\text{либо}\nonumber\\
    &x = a\cos{\left(\omega t + \varphi_0\right)},
\end{align}

<br>

где $a$ $\;-\;$ амплитуда колебаний, $\omega$ $\;-\;$ частота, $\varphi_0$ $\;-\;$ начальная фаза.

$\omega = \dfrac{2\pi}{T}$, где Т - период колебания.

Из того, что период $\cos$ равен  $2\pi$, то есть $\omega \cdot t = \dfrac{2\pi}{T} \cdot t = 2\pi k, \; k \in \mathbb{Z}$

$\Rightarrow t = T\cdot k$, то есть да, Т - период, ибо каждые t=T движение повторяется.


Два свободных параметра определяются начальными условиями: значениями $x(t)$ и $\dot{x}(t)$ в моменте t=0.

<br>

\begin{align}
    &\text{Пусть} \; \; x(0) = x_0 \; \text{и} \; \dot{x}(0) = v_0
    \\
    &\Rightarrow x_0 = acos(\varphi_0), \; v_o = -a\omega sin(\varphi_0)
    \\
    &\dfrac{sin(\varphi_0)}{cos(\phi_0)} = tg\varphi_0 = -\dfrac{v_0 a}{a \omega x_0}
    \\
    & \varphi_0 = arctg\bigg[-\dfrac{v_0}{\omega x_0}\bigg]
    \\
    & x_0 ^2 + \bigg( \dfrac{v_0}{\omega}\bigg)^2 = a^2 \Rightarrow a = \sqrt{x_0 ^2 + \bigg( \dfrac{v_0}{\omega}\bigg)^2}.
\end{align}

<br>


### Математический маятник

<br>

\begin{align}
    &ma_{\tau} = -mgsin\alpha
    \\
    & \Rightarrow m\cdot l \cdot \ddot{a} + mg\alpha = 0
    \\
    & \ddot{\alpha} + \dfrac{g}{l}\alpha = 0
    \\
    & \Rightarrow \ddot{\alpha} + \omega^2\alpha = 0, \; \omega = \sqrt{\dfrac{g}{l}} 
    \\
    & \Rightarrow \alpha = \alpha_0 \cdot cos(\omega t + \varphi_0).
\end{align}

### Пружинка

<br>

\begin{align}
    & m\ddot{x} = -k(x-x_0)
    \\
    & \ddot{x} + \dfrac{k}{m}x = \dfrac{k}{m}x_0
    \\
    & \ddot{x} + \omega^2x = \dfrac{k}{m}x_0 = \omega^2 x_0
\end{align}

<br>

**Вариант 1:** Пусть \; $\tilde{x} = x-x_0, \; \ddot{\tilde{x}} = \ddot{x}$

<br>

\begin{align}
    & \Rightarrow \ddot{\tilde{x}} + \omega^2\tilde{x} = 0 \; \; \Rightarrow \tilde{x} = acos(\omega t + \varphi_0) \Rightarrow x = x_0 + acos(\omega t + \varphi_0)
\end{align}

<br>

**Вариант 2:      ОРНДУ = ОРОДУ + ЧРНДУ** $\Rightarrow \boxed{x = acos(\omega t + \varphi_0) + x_0}$

Для одномерных гармонических колебаний можно записать в общем виде:

<br>

\begin{align}
    &\dfrac{1}{2}A\left(\dot{x}\right)^2 + \dfrac{1}{2}Bx^2 = \mathrm{const}\\
    &A\dot{x}\ddot{x} + B\dot{x}x = 0\\
    &\implies \omega^2 = \dfrac{B}{A}.
\end{align}

<br>

То есть, как только записали выражение для энергии, сразу же знаем и частоту.

<br>

\begin{align}
    &x = a\cos{\left(\omega t + \varphi_0\right)}\nonumber\\
    &T = \dfrac{1}{2} \underset{m=A}{m} \dot{x}^2 = \dfrac{1}{2} \underset{m=A}{m} (a\omega)^2\sin^2{\left(\omega t + \varphi_0\right)}\nonumber\\
    &U = \dfrac{1}{2}Bx^2=\dfrac{1}{2}Ba^2\cos^2{\left(\omega t + \varphi_0\right)}\\
    &\text{Ещё раз!}\nonumber\\
    &T = \dfrac{A}{2}(a\omega)^2\sin^2{\left(\omega t + \varphi_0\right)} = \dfrac{B}{2}a^2\sin^2{\left(\omega t + \varphi_0\right)}\\
    &U = \dfrac{B}{2}a^2\cos^2{\left(\omega t + \varphi_0\right)} = \dfrac{B}{2}a^2\cos^2{\left(\omega t + \varphi_0\right)}\\
    &\text{<}T\text{>} = \dfrac{1}{T_\text{п}}\int^{T_\text{п}}_0 \dfrac{B}{2}a^2\sin^2{\left(\omega t + \varphi_0\right)}\mathrm{d}t = \dfrac{Ba^2}{4}\\
    &\text{<}U\text{>} = \dfrac{Ba^2}{4}.
\end{align}

<br>

Кинетическая и потенциальная энергии сов. колебания с удвоенной частотой относительно ф. зн. $\dfrac{Ba^2}{4}$.

## Фазовое пространство, фазовая траектория

Совокупность координаты и скорости материальной точки (обобщенных) даёт такой объект, как фазовая плоскость (пространство).

Кривая, описывающая положение точки в фазовом пространстве, называется фазовой траекторией.

\begin{align}
    &x = a\cos{\left(\omega t + \varphi_0\right)}\nonumber\\
    &v = -a\omega\sin{\left(\omega t + \varphi_0\right)}\nonumber\\
    &\left(\dfrac{x}{a}\right)^2 + \left(\dfrac{v}{a\omega}\right)^2 = 1 \;\text{эллипс}
\end{align}
