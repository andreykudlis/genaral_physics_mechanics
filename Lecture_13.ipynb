# Неинерциальные системы отсчёта

Иногда удобно работать в НИСО.

Запишем уравнение движения на этот случай:

В системе $K$:

<br>

<figure>
    <center>
    <img src="pics/non_inertial_reference_frame.png" width="400" align="center"/>
    <figcaption>Рис. 13.1: Преобразования Галилея.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    m\ddot{\vec{r}}=\vec{F},
\end{align}

<br>

где $\vec{F}$ ― реальные физические силы.

В системе $K'$:

<br>

\begin{align}
    \vec{r} = \vec{r}\;' + \vec{R}.
\end{align}

<br>

Движение $K'$ относительно $K$ раскладываем на поступательное движение со скоростью $\vec{V}$ и вращательное с угловой скоростью $\vec{\omega}$ относительно оси, проходящей через $O'$.

<br>

\begin{align}
    &\vec{r} = \vec{e}_xx+\vec{e}_yy+\vec{e}_zz;\\
    &\dfrac{\mathrm{d}\vec{r}}{\mathrm{d}t}\triangleq \underbrace{\vec{v}}_{\text{скор. м. т. в K}} = \dfrac{\mathrm{d}\vec{r}\;'}{\mathrm{d}t} + \dfrac{\mathrm{d}\vec{R}}{\mathrm{d}t} \triangleq \vec{V} + \dfrac{\mathrm{d}\vec{r}\;'}{\mathrm{d}t};\\
    & \dfrac{\mathrm{d}\vec{r}\;'}{\mathrm{d}t} = \underbrace{\dot{x}'\vec{e}_x\;' + \dot{y'}\vec{e}_y\;' + \dot{z}'\vec{e}_z\;'}_{\vec{v}\;'} + x'\dfrac{\mathrm{d}\vec{e}_x\;'}{\mathrm{d}t} + y'\dfrac{\mathrm{d}\vec{e}_y\;'}{\mathrm{d}t} + z'\dfrac{\mathrm{d}\vec{e}_z\;'}{\mathrm{d}t};\\
    & \mathrm{d}\vec{e}_x\;' = - \vec{e}_z\;'\mathrm{d}\varphi_{y'} + \vec{e}_{y'}\mathrm{d}\varphi_{z'} \implies \dot{\vec{e}}_{x'} = \vec{e}_{y'}\dot{\varphi}_{z'} - \vec{e}_{z'}\dot{\varphi}_{y'} = \vec{e}_{y'} \omega_{z'} - \vec{e}_{z'}\omega_{y'} \implies\\ \nonumber
    &\implies x \, ' \dot{\vec{e}}_x \, ' + y \, ' \dot{\vec{e}}_y \, ' +  z \, ' \dot{\vec{e}}_z \, ' = x \, ' \cdot (\vec{e}_{y \, '} \omega_{z\, '} - \vec{e}_{z\, '}\omega_{y\, '}) + y \, ' \cdot (-\vec{e}_{x \, '} \omega_{z\, '} - \vec{e}_z\, '\omega_x) + \nonumber \\
    &+ z \, ' (-\vec{e}_y \, '\omega_{x\, '} +\vec{e}_x\omega_{y\, '}) = \omega \times \vec{r}\, '.
\end{align}

<br>

Ускорение:

<br>

\begin{align}
    &\vec{a} \overset{def}{=} \dfrac{d\vec{V}}{dt} = \dfrac{d\vec{V}}{dt} + \dfrac{d}{dt}\vec{V}\,' + \dfrac{d}{dt}\vec{\omega} \times \vec{r}\,' = \vec{a}_0 + \vec{a}\,' + \vec{\omega}\times \vec{V}\,' + \dfrac{d\vec{\omega}}{dt}\times\vec{r}\,' + \vec{\omega} \times \vec{V}\,' + \vec{\omega} \times [\vec{\omega}\times \vec{r}\,'] \nonumber \\
    &= \vec{a}_0 + \vec{a}\,' + 2[\vec{\omega}\times \vec{V}\,'] + \dfrac{d\vec{\omega}}{dt} \times \vec{r} + \vec{\omega} \times [\vec{\omega}\times \vec{r}\,'],
\end{align}

<br>

где $\vec{a}_0$ - поступательное ускорение, $\vec{a}\,'$ - ускорение в K'.

<br>

\begin{align}
     &\text{Переносное ускорение:} \ \ \vec{a}_\text{пер} = \vec{a}_0 + \dfrac{d\vec{\omega}}{dt} \times \vec{r}\,', \nonumber \\
     &\text{Кориолисово ускорение:} \ \ \vec{a}_\text{кор} = 2[\vec{\omega} \times \vec{V}\,'].
 \end{align}

<br>

Умножим на массу полученное значение ускорения:

<br>

\begin{align}
     &m\vec{a}\,' = \underset{\overset{||}{m\vec{a}}}{\vec{F}} \underbrace{-m\vec{a}_0 - 2m[\vec{\omega} \times \vec{V}\,'] - m\dfrac{d\vec{\omega}}{dt} \times \vec{r}' - m[\vec{\omega}[\vec{\omega} \times \vec{r}\,']]}_{\vec{F}_\text{in}}. \\
     &\vec{F}_\text{in пост.} = - m\vec{a}_0; \\
     &\vec{F}_\text{in кор.} = -2m [\vec{\omega} \times \vec{V}]; \\
     &\vec{F}_\text{in ц.б.} = -m[\vec{\omega}[\vec{\omega}\times \vec{r}\,']] - m \dfrac{d\vec{\omega}}{dt} \times \vec{r}\,'.
 \end{align}

<br>

Введя силы инерции, рассмотрим примеры их проявления.
## Поступательная сила инерции(сила Д'Аламбера)
### Ускоряющийся автобус
Тело находится в ускоряющемся автобусе/вагоне и т.д. Пусть в нашем случае это будет человек и мы хотим сохранять состояние покоя. Как?

<br>

    <figcaption>Рис. 13.2: Задача на ускоряющееся тело в НИСО.</figcaption>
    </center>
</figure>

<br>

Уравнение:

<br>

\begin{align}
    &m\vec{a} \, '=\vec{F}+\vec{F}_\text{in}, \nonumber \\
    &\vec{v} \, ' =0, \vec{\omega} = 0 \Rightarrow \vec{F}_\text{in}=-m\vec{a}_0.
\end{align}

<br>

Для сохранения состояния покоя, нам необходимо, чтобы $m\vec{a} \, ' =0$. 

<br>

\begin{align}
    &m\vec{a} \, ' =0 = m\vec{g} + \vec{N} + \vec{F}_\text{тр} - m\vec{a}_0, \nonumber \\
    &oy: N = mg, \nonumber \\
    &ox: kmg = ma_0.
\end{align}

<br>

Однако, также необходимо записать уравнение на моменты сил, чтобы человек не упал:

<br>

\begin{align}
    &\dfrac{l}{2} \cos\alpha \, m a_0=\dfrac{l}{2} \sin\alpha \, mg, \nonumber \\
    &\boxed{\dfrac{a_0}{g} = \tan\alpha}.
\end{align}

<br>

\begin{align}
    &m\vec{a} \, ' = \vec{T} + m\vec{g}_0 + \vec{F}_\text{in}, \nonumber \\
    &m\vec{a} \, ' = \vec{T} + m\vec{g}_0 - m\vec{a}_0, \nonumber \\
    &m\vec{a} \, ' = \vec{T} + m(\vec{g}_0 - \vec{a}_0) \Rightarrow \vec{g}_\text{эфф} = \vec{g}_0 - \vec{a}_0 \Rightarrow m\vec{a} \, ' = \vec{T} + m\vec{g}_\text{эфф} \nonumber \\
    &\Rightarrow \underset{\text{период}}{T} = 2\pi \sqrt{\dfrac{l}{|\vec{g}_\text{эфф}|}} = 2\pi \sqrt{\dfrac{l}{\sqrt{\vec{a}_0^2 + \vec{g}_0^2}}}.
\end{align}

Найти новое положение равновесия: ($\vec{v} \, ' = 0, \vec{a} \, ' = 0$)

<br>

<br>

\begin{align}
    &m\vec{a} = \vec{F} + \vec{F}_\text{in}, \nonumber \\
    &\vec{F}_\text{in}= -m [\vec{\omega}, [\vec{\omega} \times \vec{r} \, ']].\\
    &\vec{\omega} \times \vec{r} \, ' = \vec{c}, \nonumber \\
    &|\vec{c}| = |\vec{\omega}| \cdot \underbrace{|\vec{r} \, '| \cdot \sin\alpha}_R, \nonumber \\
    &[\vec{\omega} \times \vec{c}] = \vec{b}, \ \vec{r} \, ', \ \vec{b} \ \text{и} \ \vec{\omega} - \text{в одной плоскости}. \nonumber \\
    &|[\vec{\omega} \times [\vec{\omega} \times \vec{r} \, ']]| = \omega^2 \cdot R, \text{направление всегда к оси вращения}.
\end{align}

<br>

<figure>
    <center>
    <img src="pics/13_5.png" width="350" align="center"/>
    <figcaption>Рис. 13.5: Новое положение равновесия.</figcaption>

<br>

Из условия следует: 

<br>

\begin{align}
    &\vec{0} = m\vec{g} + \vec{T} - m [\vec{\omega} \times [\vec{\omega} \times \vec{r} \, ']], \nonumber \\
    &0 = mg - T \cos\alpha, \nonumber \\
    &0 = m \omega^2 \underbrace{R}_{l\sin\alpha} - T\sin\alpha.
\end{align}

<br>

\begin{align}
    &\begin{cases}
    T\cos\alpha = mg, \\
    m\omega^2 l = T
    \end{cases} \Rightarrow \omega^2 l= \dfrac{g}{\cos\alpha}, \nonumber \\
    &\cos\alpha = \dfrac{g}{\omega^2 l} \leqslant 1.
\end{align}

<br>

То есть либо $\alpha = 0$ если $\dfrac{g}{\omega^2l}\geqslant 1$

<br>

\begin{align}
    &\Rightarrow \sin\alpha \, l \, m\omega^2 = T\sin\alpha, \nonumber \\
    &0 \cdot l \, m\omega^2 = T \cdot 0 \ \text{и} \ T = mg.
\end{align}

<br>

То есть в таком случае отклонение от горизонтальной оси: 

<br>

\begin{align}
    \alpha = \arccos{\left[\dfrac{g}{\omega^2l}\right]}.
\end{align}

### Колебание веса из-за вращения планеты
Учтём вращение Земли, чтобы оценить разницу в весе ($\vec{P}$) тела в разных точках Земли. Вес тела, $\vec{P}$ - сила, с которой тело действует на опору. $|\vec{P}| = P = |\vec{N}| = N \ \ \ \ (\vec{P} = - \vec{N}).$

<br>

\begin{align}
    &m\vec{a} \, ' = m\vec{g} + \vec{N} - m[\vec{\omega} \times [\vec{\omega} \times \vec{r}]], \nonumber \\
    &0 = m\vec{g} - m[\vec{\omega} \times [\vec{\omega} \times \vec{r}]] + \vec{N}, \nonumber \\
    &m\vec{g}_\text{eff} + \vec{N} = 0, \nonumber \\
    &\vec{g}_\text{eff} = \vec{g} \underbrace{-[\vec{\omega} \times [\vec{\omega} \times \vec{r}]]}_{\vec{a}_c} = \vec{g} + \vec{a}_c, \nonumber \\
    &|\vec{a}_c| = \omega^2R_\oplus \cos\alpha, \nonumber \\
    &|\vec{g}_\text{eff}| = \sqrt{(\vec{g}+ \vec{a}_c)^2} = \sqrt{\vec{g}^2 +\vec{a}_c^2 + 2\vec{a}_c\vec{g}} = g \sqrt{1+ \dfrac{\omega^4R_\oplus^2\cos^2\alpha}{g^2} + \dfrac{2\vec{a}_c\vec{g}}{g^2}} \nonumber \\
    &\phantom{|\vec{g}_\text{eff}|} =g\sqrt{1-\dfrac{2\omega^2R_\oplus\cos^2\alpha}{g}+ \dfrac{\omega^4R_\oplus^2\cos^2\alpha}{g^2}}.\\
    &|\vec{g}_\text{eff}| = g_{eff} \simeq g - \omega^2 R_\oplus \cos^2\alpha.
\end{align}

<br>

В случае Земли, 

<br>

\begin{align*}
    &\omega \approx 7{,}3 \cdot 10^{-5} \text{ с}^{-1}, \\
    &\omega^2 \approx 5{,}3 \cdot 10^{-9} \text{ с}^{-2}, \\
    &R_\oplus \approx 6{,}4 \cdot 10^6 \text{ м}.
\end{align*}

<br>

Подставляя это в получившийся вывод, получаем численную поправку к ускорению свободного падения

<br>

\begin{align}
    &|\vec{g}_\text{eff}| \approx g - 0{,}034 \ \dfrac{\text{м}}{\text{с}^2} \cos \alpha.
\end{align}

<br>

Для общего понимания выразим максимальное отклонение $|\vec{g}_\text{eff}|$ от $g$ в процентах от $g$.

<br>

\begin{align}
    & \dfrac{g_{eff max} - g_{eff min}}{g_{eff max}} = \dfrac{g - (g - 0.034)}{g} \simeq 0.0035 \implies 0.35\%.
\end{align}

<br>

Такое отклонение возможно засечь достаточно точными приборами, однако его малость оставляет нам простор случаев, в которых мы можем игнорировать переменчивость ускорения свободного падения.

### Пуля и карусель

<br>

<figure>
    <center>
    <img src="pics/100.png" width="400" align="center"/>
    <figcaption>Рис. 13.6: Вращающаяся карусель, на которой стрелок целится в мишень.</figcaption>
    </center>
</figure>

<br>

Рассмотрим стрелка, который стоит на вращающейся карусели радиуса $R$ и целится в мишень, которую видит под углом $\beta$. (Сама мишень, значит, удалена от диаметра между ней и стрелком на дугу $\varphi = 2 \beta$). Скорость пули, которую выпускает стрелок, примем $u$. Рассмотрим её смещение $x$ из-за угловой скорости $\omega$ и найдём угол $\alpha$, который нужен стрелку, чтобы скомпенсировать как своё вращение, так и вращение мишени.

<br>

$\require{enclose}$
\begin{align}
    &\alpha \approx tg \alpha = \dfrac{\omega R}{u} + \beta, \\
    &\alpha = \dfrac{\omega R}{u} + \dfrac{\varphi}{2} = \dfrac{\omega R}{u} + \dfrac{\omega \cdot \dfrac{\enclose{updiagonalstrike}{2} R}{u}}{\enclose{updiagonalstrike}{2}} = \dfrac{\omega R}{u} + \dfrac{\omega R}{u} = \dfrac{2 \omega R}{u}.
\end{align}

<br>

Рассмотрим динамику пули в системе отсчёта, связанной с вращающимся диском. За $m$ обозначим её массу, за $\vec{v'}$ ― скорость в этой системе отсчёта, за $\vec{a'}$ ― ускорение.

<br>

<figure>
    <center>
    <img src="pics/101-1.png" width="400" align="center"/>
    <figcaption>Рис. 13.7: Динамика пули в неинерциальной системе отсчёта.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &m \vec{a'} = \underbrace{\vec{F}}_{0} - \underbrace{m \vec{a}_0}_0 - \underbrace{m \Bigl[ \vec{\omega} \times [\vec{\omega} \times \vec{r}] \Bigr]}_{\text{мало, почти } 0} - 2 m [\vec{\omega} \times \vec{v'}], \\
    &\text{где } |[\vec{\omega} \times \vec{v'}]| = \omega u.
\end{align}

<br>

Довольно очевидно, что ускорение направлено строго вдоль оси $y'$. ( На самом деле, это не совсем так : кориолисово ускорение всегда перпендикулярно скорости пули в НИСО, а она вообще не совсем параллельна оси $x'$, но при этом из-за высокой скорости пули составляет малый угол с этой осью, поэтому проекция кориолисова ускорения на ось $x'$ будет величиной первого порядка малости и, соответственно, мы можем ей пренебречь. Более точно траекторию пули можно представить в виде дуги окружности, так как кориолисово ускорение всегда перпендикулярно скорости, следовательно изменяет только её направление, но не изменяет абсолютное значение, но так как мы лишь оцениваем величину угла $\alpha$, то замена параболы на дугу окружности даст поправку более высокого порядка малости и ей можно будет пренебречь)

<br>

\begin{align}
    &m a_{y'} = - 2 m \omega u. \\
    &\vec{a'} = \begin{pmatrix} 0 \\ a_{y'} \end{pmatrix} = \begin{pmatrix} a_{x'} \\ a_{y'} \end{pmatrix} = \begin{pmatrix} 0 \\ -2 \omega u \end{pmatrix}. \\
    &\text{Тогда диаметр } 2R = \underbrace{u \cos \alpha}_{v_{0x}} \tau. \\ \nonumber
\end{align}

<br>

Получившаяся ситуация кинематически ничем не отличается от движение в поле тяжести Земли ― как и в том случае, всё время полёта $\tau$ можно разбить на время до достижения максимума по оси $y'$ и на время после, так что $\tau_\uparrow = \tau_\downarrow = \dfrac{1}{2} \tau$.

<br>

<figure>
    <center>
    <img src="pics/101-2.png" width="400" align="center"/>
    <figcaption>Рис. 13.8: Равноускоренное движение пули от стрелка до мишени.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &y'(t) = v_{0y} t + \dfrac{a_{y'} t^2}{2} = u t \sin \alpha - \dfrac{2 \omega u t^2}{2}. \\
\end{align}

<br>

По лемме Ферма при достижении максимума производная обнуляется:

<br>

$\require{enclose}$
\begin{align}
    &\left. \dfrac{\mathrm{d}y'}{\mathrm{d}t} \right|_{t = \tau_\uparrow} = 0 = u \sin \alpha - 2 \omega u \tau_\uparrow. \\
    &\tau_\uparrow = \dfrac{\enclose{updiagonalstrike}{u} \sin \alpha}{2 \omega \enclose{updiagonalstrike}{u}}, \\
    &\tau = 2 \tau_\uparrow = \dfrac{\sin \alpha}{\omega}. \\
    &\dfrac{2 R}{u \cos \alpha} = \dfrac{\sin \alpha}{\omega}. \\
\end{align}

<br>

Возьмём угол $\alpha$ малым ― тогда синус примерно равен ему самому, а косинус можно принять за единицу.

<br>

\begin{align}
    & \alpha \approx \sin \alpha = \dfrac{2 R \omega}{u}.
\end{align}
