# Динамика абсолютно твёрдого тела
## Закон изменение количества движения. Тензор инерции

Можно пойти ещё дальше. Если система центра масс, то и скорость можно брать относительно центра масс (тогда и $\vec{r}_i\;'$ в системе ц.м.):

<br>

\begin{align}
    &L = \sum\vec{r}_i\;'\times m_i\vec{v}_i = \sum \vec{r}_i\;'\times m_i\left(\vec{v}_c + \vec{v}_i\;'\right) = \nonumber\\
    &\sum\vec{r}_i\;' \times m_i\vec{v}_c + \sum \vec{r}_i\;' + m_i\vec{v}_i\;' = \sum\underbrace{\left(m_i\vec{r}_i\;'\right)}_{M\vec{r}_{c.m.} = 0} \times \vec{v}_c + \sum \vec{r}_i\;' \times m_i \vec{v}_i\;' \implies \nonumber\\
    & \implies \boxed{\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \vec{M}},
\end{align}

<br>

где $\vec{M} = \sum \vec{r}_i\;'\times\vec{F}^i, \vec{L} = \sum\vec{r}_i\;'\times m\vec{v}_i\;'$ в системе центра масс.

В общем случае удобно рассмотреть вращение вокруг неподвижной точки или мгновенной оси (хотим посмотреть на связь $\vec{L}$ и $\vec{\omega}$, <u>если нашли такую</u>):

<br>

\begin{align}
    &\vec{L} = \int\vec{r}\times\vec{v}\mathrm{d}m = \int\vec{r}\times\left[\vec{\omega}\times\vec{r}\right]\mathrm{d}m = \nonumber\\
    & = \int\left(\vec{\omega}\;^2\vec{r}\;^2-\vec{r}\left(\vec{\omega},\vec{r}\right)\right)\mathrm{d}m = \int_V\left(r^2\vec{\omega}-\vec{r}\left(\vec{\omega}, \vec{r}\right)\right)\mathrm{d}m = \hat{I}\vec{\omega} \nonumber\\
    &\boxed{\vec{L} = \hat{I}\vec{\omega}},
\end{align}

<br>

где $\hat{I}$ $\;-\;$ тензор инерции.

Если тензор инерции диагонален, то говорят, что мы рассматриваем его в главных осях.

<br>

\begin{align}
    &\hat{I} = \int_V
        \begin{pmatrix}
        y^2+z^2 & -xy & -xz\\
        -xy & x^2+z^2 & -yz\\
        -xz & -yz & y^2+x^2
        \end{pmatrix}dm.
\end{align}

<br>

Таким образом, мы видим, что в общем случае $\vec{L} \not \sim \vec{\omega}$.

В случае вращательного движения относительно фиксированной оси (пусть $\omega_x = \omega_y = 0, \quad \omega_z \not = 0$):

<br>

\begin{align}
    &L_z = \omega I, \quad I - \text{ момент инерции относительно оси вращения}, \\
    & \underbrace{\boxed{\dfrac{\mathrm{d}L_z}{\mathrm{d}t} = I \dfrac{\mathrm{d}\omega}{\mathrm{d}t} = M_z}}_{\text{основное уравнение}} - \text{ относительно неподвижной оси}.
\end{align}

<br>

### Замечание про тензор инерции.
Матрица симметричная и вещественная, значит можно найти ортогональную матрицу $\hat{Q}$:
\begin{align*}
    \hat{I}_d = Q^T \hat{I} Q,
\end{align*}
где $\hat{I}_d$ $\;-\;$ диагональная матрица, $Q^T$ $\;-\;$ матрица поворота.

**Замечание**
Мы с вами помним, что можно записывать:

<br>

\begin{align}
    &\vec{M} = \sum \vec{r}_i\,'\times\vec{F}\,'\nonumber \\
    &\underset{\text{относительно ц.м.}}{\vec{L} = \sum \vec{r}_i\,' \times m_i\vec{V}_i\,'},
\end{align}

<br>

Откуда

<br>

\begin{align}
    \boxed{\dfrac{d\vec{L}}{dt} = \vec{M}.}
\end{align}

<br>

\begin{align}
    &\vec{L} = \sum \vec{r}_i\,' \times m_i [\vec{\omega},\vec{r}_i\,'] = \int \vec{r} \times [\vec{\omega}, \vec{r}] dm = \int dm (\vec{\omega} \vec{r} \ ^2 - \vec{r}(\vec{r}, \vec{\omega})).
\end{align}

<br>

Таким образом, только для неподвижной оси (если вокруг неё вращение), но и относительно центра масс, который, вообще говоря, может двигаться с ускорением.

Резюмируем:
есть уравнение $d\vec{L}/dt = \vec{M}$. Мы его можем писать либо **относительно неподвижной точки**, либо **относительно центра масс** (но скорости берём в инерциальной системе отсчёта):

<br>

\begin{align}
    &\vec{M} = \sum \vec{r}_i\,'\times\vec{F}\,'\nonumber \\
    &\vec{L} = \sum \vec{r}_i\,' \times m_i\vec{V}_i\,'. \\
    &\Rightarrow \dfrac{d\vec{L}}{dt} = \vec{M}
\end{align}

<br>

3) либо можно брать скорости в системе центра масс:

<br>

\begin{align}
    &\vec{M} = \sum \vec{r}_i\,'\times\vec{F}\,'\nonumber \\
    &\vec{L} = \sum \vec{r}_i\,' \times m_i\vec{V}_i\,'.
\end{align}

<br>

Если мы можем записать как вращение относительно оси (это может быть неподвижная точка, либо центр масс, если тело вращается относительно него) то:

<br>

\begin{align}
    &\vec{L} = \int dm(\vec{\omega}\vec{r} \ ^2 - \vec{r}(\vec{r},\vec{\omega})) = \hat{I} \vec{\omega}, \ \ \text{откуда далее}\nonumber \\
    &\dfrac{d\vec{L}}{dt}= \vec{M}.
\end{align}

<br>

Если движение относительно ориентированной оси, то достаточно рассмотреть проекцию момента импульса:

<br>

\begin{align}
    \dfrac{dL_z}{dt}=M_z \Rightarrow I\dfrac{d\omega}{dt} = M_z.
\end{align}

<br>

То есть найдутся оси ― главные оси, ― в которых тензор инерции будет диагонален.\\
Получаем уравнения на шесть степеней свободы:

<br>

\begin{align}
    &\dfrac{d\vec{L}}{dt} = \vec{M}, \nonumber \\
    &m \dfrac{d\vec{v}_c}{dt}=\vec{F}
\end{align}

<br>

Каково будет условие равновесия АТТ ?

$\vec{F} = 0$ и $\vec{M} = 0$.

То есть отсутствие ускорения у центра масс и отсутствие ускорения во вращении + начальные ускорения.

## Кинетическая энергия вращательного движения
Рассмотрим кинетическую энергию абсолютно твёрдого тела. Записываем по определению:

<br>

\begin{align}
    &K \equiv T =  \int \limits _V \dfrac{1}{2}(\vec{V}_O + [\vec{\omega}, \vec{r}_\text{из т. o}])^2 dm = \nonumber \\
    &= \dfrac{M}{2}\vec{V}_O^2 + \vec{V}_O \times \int \limits _V [\vec{\omega},\vec{r}_\text{из т. o}] dm + \underset{\text{связано с чистым вращением.}}{\dfrac{1}{2}\int ([\vec{\omega}, \vec{r}_\text{из ц. м.}])^2 dm} \overset{\text{Если ц.м}}{=} \dfrac{M}{2}V_c^2 + \dfrac{1}{2}\int [\vec{\omega}, \vec{r}_\text{из т. o}]^2 dm.
\end{align}

<br>

Рассмотрим частный случай вращения относительно неподвижной оси $z$, для удобства отсчитывая радиус-векторы точек от центра масс:

<br>

\begin{align}
    &T = \dfrac{MV_c^2}{2} + \dfrac{1}{2} \int \limits_M \omega^2 r_\text{c.m.}^2 \mathrm{d}m \nonumber \\
    &= \dfrac{MV_c^2}{2} + \dfrac{I_z \omega^2}{2} \overset{z\text{ неподвижна}}{=} \dfrac{MV_c^2}{2} + \dfrac{L_z \omega}{2}.
\end{align}

<br>

<br>

\begin{align}
    &T = \dfrac{1}{2} \int \limits_M v^2 \mathrm{d}m = \dfrac{1}{2} \int \limits_M \vec{v} \cdot [\vec{\omega} \times \vec{\vec{r}}] \mathrm{d}m \nonumber \\
    &= \dfrac{1}{2} \vec{\omega} \cdot \int \limits_M [\vec{r} \times \vec{v} \mathrm{d}m] = \dfrac{1}{2} \vec{\omega} \cdot \vec{L} = \dfrac{1}{2} \vec{\omega} \cdot \hat{I} \vec{\omega}.
\end{align}

<br>

Рассмотрим вращение колеса (или любой другой фигуры с окружностью в сечении) относительно какой-то оси, отстоящей на $R$ от оси $z$, проходящей через центр масс.

<br>

\begin{align}
    &\dfrac{1}{2} \omega^2 (I_\text{c.m.} + M R^2) \overset{I_\text{c.m.} = I_z}{=} \dfrac{\omega^2 I_z}{2} + \dfrac{1}{2} \omega^2 M R^2 \overset{\omega R = V_c}{=} \dfrac{I_z \omega^2}{2} + \dfrac{M V_c^2}{2}.
\end{align}

<br>

## Скатывание тел с наклонной плоскости без проскальзывания

Рассмотрим скатывание колеса с наклонной плоскости под действием силы тяжести, без проскальзывания ― то есть, в каждый момент считая точку касания поверхности неподвижной, что достигается за счёт действия силы трения качения.

<br>

### Решение 1, через неподвижную точку
На предмет действуют две силы ― тяжести и трения. Выберем точку $O$, она неподвижна (так как скатывание без соскальзывания), и при этом именно к ней приложена сила трения, а значит момент силы трения относительно $O$ нулевой. Осталось рассмотреть момент силы тяжести $M_g$, величину которого довольно легко посчитать из геометрических соображений.

<br>

\begin{align}
    &\dfrac{\mathrm{d}L_O}{ \mathrm{d}t} = M_g, \\
    &I_O \dfrac{\mathrm{d}\omega}{\mathrm{d}t} = R m g \sin \alpha. \\
    &\left( I_c + M R^2 \right) \dfrac{\mathrm{d} \omega}{\mathrm{d}t} = R mg \sin \alpha, \\
    &\dfrac{\mathrm{d}v}{\mathrm{d}t} = \dfrac{R^2 m g \sin \alpha}{I_c + m R^2}.
\end{align}

<br>

### Решение 2, через центр масс
Как мы доказали, закон изменения момента количества движения можно записывать не только относительно неподвижной точки, но и центра масс: в этом случае сила трения обретает некий момент, и для решения задачи требуется обратиться ко второму закону Ньютона, то есть закону изменения линейного количества движения. Зато, радует то, что момент силы тяжести относительно центра масс нулевой. 

<br>

\begin{align}
    &I_c \dfrac{\mathrm{d}\omega}{\mathrm{d}t} = M_\text{тр}, \\
    &\dfrac{I_c}{R} \dfrac{\mathrm{d}v}{\mathrm{d}t} = R F_\text{тр}. \\
    &m \dfrac{\mathrm{d}v}{\mathrm{d}t} = m g \sin \alpha - F_\text{тр}. \\
    &\dfrac{I_c}{R} \dfrac{\mathrm{d}v}{\mathrm{d}t} = R (m g \sin \alpha - m \dfrac{\mathrm{d}v}{\mathrm{d}t}), \\
    &\dfrac{\mathrm{d}v}{\mathrm{d}t} = \dfrac{g \sin \alpha}{1 + \dfrac{I_c}{m R^2}}, \\
    &\dfrac{\mathrm{d}v}{\mathrm{d}t} = \dfrac{R^2 m g \sin \alpha}{I_c + m R^2}.
\end{align}

<br>

### Решение 3, через закон сохранения энергии
Вместо рассмотрения сил и их моментов можно посмотреть на энергию системы, которая будет складываться из кинетической и потенциальной: работа силы трения качения будет нулевой, поэтому механическая энергия сохранится.

<br>

\begin{align}
    &K_2 - K_1 = U_1 - U_2, \\
    & \dfrac{1}{2}(I_c + mR^2)\omega^2 = mgl\sin\alpha \\
    &\Rightarrow 2 \cdot \dfrac{1}{2} (I_c + mR^2)\omega \dfrac{d\omega}{dt} = mgv\sin\alpha, \\
    &(I_c + mR^2)\dfrac{d\omega}{dt} = mgR\sin\alpha, \\
    &\dfrac{dv}{dt} = \dfrac{R^2 mg\sin\alpha}{I_c + mR^2}. \
\end{align}

<br>

Как видно, все три рассмотренных метода эквивалентны в том смысле, что они дают одинаковые ответы ― ведь мы рассматриваем один и тот же физический процесс, просто разными методами. Полезно овладеть всеми тремя, так как в более сложных ситуациях уравнения одного из предложенных подходов могут оказаться сильно проще остальных, что сильно экономит мыслительные силы при обсчёте какой-то ситуации.

## Уравнения Эйлера
До сего момента мы рассматривали лишь плоское вращение. Что же будет в случае более сложного движения, которое не происходит вдоль определённой оси?

Как мы уже вывели ранее, для момента количества движения и суммарного момента сил относительно центра масс всегда справедливо

<br>

\begin{align}
    &\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \vec{M}. \\
    &\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \dfrac{\mathrm{d}L_x}{\mathrm{d}t} \vec{e'}_x + \dfrac{\mathrm{d}L_y}{\mathrm{d}t} \vec{e'}_y + \dfrac{\mathrm{d}L_z}{\mathrm{d}t} \vec{e'}_z + L_x \dfrac{\mathrm{d}\vec{e'}_x}{\mathrm{d}t} + L_y \dfrac{\mathrm{d}\vec{e'}_y}{\mathrm{d}t} + L_z \dfrac{\mathrm{d}\vec{e'}_z}{\mathrm{d}t},
\end{align}

<br>

где последнюю тройку слагаемых можно вместе представить как $[\vec{\omega} \times \vec{L}].$

<br>

\begin{align}
    &\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = \dfrac{\mathrm{d}L_x}{\mathrm{d}t} \vec{e'}_x + \dfrac{\mathrm{d}L_y}{\mathrm{d}t} \vec{e'}_y + \dfrac{\mathrm{d}L_z}{\mathrm{d}t} \vec{e'}_z + [\vec{\omega} \times \vec{L}] = \vec{M}.
\end{align}

<br>

Расписывая покоординатно,

<br>

\begin{align}
    &\left\{ 
    \begin{array}{l}
        \dfrac{\mathrm{d}L_x}{\mathrm{d}t} + (\omega_y L_z - \omega_z L_y) = M_x, \\ \\
        \dfrac{\mathrm{d}L_y}{\mathrm{d}t} + (-\omega_x L_z + \omega_z L_x) = M_y, \\ \\
        \dfrac{\mathrm{d}L_z}{\mathrm{d}t} + (\omega_x L_y - \omega_y L_x) = M_z. 
    \end{array} \right.
\end{align}

<br>

Если мы возьмём за оси $x$, $y$, $z$ главные оси, в которых тензор момента инерции имеет диагональный вид, то получим 

<br>

\begin{align}
    &\left\{ 
    \begin{array}{l}
        \dfrac{\mathrm{d}L_x}{\mathrm{d}t} + (I_z - I_y) \omega_z \omega_y = M_x, \\ \\
        \dfrac{\mathrm{d}L_y}{\mathrm{d}t} + (-I_z + I_x) \omega_x \omega_z = M_y, \\ \\
        \dfrac{\mathrm{d}L_z}{\mathrm{d}t} + (I_y - I_x) \omega_x \omega_y = M_z.
    \end{array} \right.
\end{align}

<br>

Эти три уравнения называют уравнениями Эйлера.

## Свободное вращение

Рассмотрим уравнения Эйлера для некоторых предметов в случае нулевого внешнего момента сил, $\vec{M} = 0$.

### Шар

Из-за симметричности шара все его проекции на оси равны, $I_x = I_y = I_z$, из-за чего все вторые слагаемые в уравнениях Эйлера обращаются в ноль.

<br>

\begin{align}
    &\left\{ 
    \begin{array}{l}
        \dfrac{\mathrm{d}L_x}{\mathrm{d}t} + 0 = 0, \\ \\
        \dfrac{\mathrm{d}L_y}{\mathrm{d}t} + 0 = 0, \\ \\
        \dfrac{\mathrm{d}L_z}{\mathrm{d}t} + 0 = 0.
    \end{array} \right. \\
    &\dfrac{\mathrm{d}\vec{L}}{\mathrm{d}t} = 0, \\
    &\vec{L} = \text{const}, \\
    &\vec{\omega} = \text{const}.
\end{align}

<br>


### Симметричный волчок
Посмотрим на фигуру с осевой симметрией относительно оси $z$, причём такую, что $I_x = I_y \neq I_z$ и $I_z - I_x = \Delta I$.

<br>

\begin{align}
    &\left\{ 
    \begin{array}{l}
        I_x \dfrac{\mathrm{d}\omega_x}{\mathrm{d}t} + \Delta I\ \omega_y \omega_z = 0, \\ \\
        I_y \dfrac{\mathrm{d}\omega_y}{\mathrm{d}t} - \Delta I\ \omega_x \omega_z = 0, \\ \\
        I_z \dfrac{\mathrm{d}\omega_z}{\mathrm{d}t} + 0 = 0.
    \end{array} \right. \\
    &\left\{ 
    \begin{array}{l}
        \dfrac{\mathrm{d}\omega_x}{\mathrm{d}t} + \dfrac{\Delta I}{I_x}\ \omega_y \omega_z = 0, \\ \\
        \dfrac{\mathrm{d}\omega_y}{\mathrm{d}t} - \dfrac{\Delta I}{I_y}\ \omega_x \omega_z = 0, \\ \\
        \dfrac{\mathrm{d}\omega_z}{\mathrm{d}t} = 0.
    \end{array} \right. \\
    &\omega_z = \text{const}.
\end{align}

<br>

Положим приведённую угловую частоту $\Omega = \dfrac{\Delta I}{I_x} \omega_z$, которая будет постоянной из-за постоянства $\omega_z$.

<br>

\begin{align}
    &\left\{
    \begin{array}{l}
        \dfrac{\mathrm{d}\omega_x}{\mathrm{d}t} + \Omega \omega_y = 0, \\ \\
        \dfrac{\mathrm{d}\omega_y}{\mathrm{d}t} - \Omega \omega_x = 0.
    \end{array} \right. \\
    &\omega_x = \dfrac{1}{\Omega}\dfrac{\mathrm{d}\omega_y}{\mathrm{d}t}, \\
    &\dfrac{\mathrm{d}}{\mathrm{d}t} \left( \dfrac{1}{\Omega}\dfrac{\mathrm{d}\omega_y}{\mathrm{d}t} \right) + \Omega \omega_y = 0, \\
    &\dfrac{1}{\Omega} \ddot{\omega}_y + \Omega \omega_y = 0, \\
    &\ddot{\omega}_y + \Omega^2 \omega_y = 0.
\end{align}

<br>

Это дифференциальное уравнение задаёт колебание величины $\omega_y$ с угловой частотой $\Omega$! А именно, (это можно проверить прямой подстановкой) решение имеет вид

<br>

\begin{align}
    &\omega_y = A \cos (\Omega t + \varphi), \\
    &\omega_x = - A \sin (\Omega t + \varphi),
\end{align}
где $A$ ― амплитуда, $\varphi$ ― начальная фаза колебания.
Положим $\varphi = 0$ и $A = \omega_0$.

<br>

\begin{align}
    &\omega_y = \omega_0 \cos (\Omega t), \\
    &\omega_x = - \omega_0 \sin (\Omega t). \\
    &\text{Заметим, что } \omega_x^2 + \omega_y^2 + \omega_z^2 = \omega_z^2 + \omega_0^2 = \text{const}.
\end{align}

<br>

Таким образом, направление угловой скорости будет меняться периодическим образом, а её величина сохраняться ― в целом, вектор угловой скорости будет двигаться по окружности. Подобные явления называют прецессией (в данном случае, она происходит вокруг оси $z$).
