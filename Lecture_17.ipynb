<br>

Будем считать, что в системе $K$ в момент времени, когда $t=0$, совпадают начала отсчётов систем $K$ и $K\, '$. Когда это происходит, соответственно $t\,' = 0$. Проведём аналогию: кладём один конец стержня в начало координат системы $K\, '$, обозначим его изначальную длину за $l_0$. В релятивистском случае из материала прошлой лекции мы знаем, что в системе отсчёта $K$ его длина при движении составляет:

<br>

\begin{align}
    l = l_0\sqrt{1-\dfrac{V^2}{c^2}} = \dfrac{l_0}{\gamma}, \ \text{где} \ \gamma = \dfrac{1}{\sqrt{1 - \dfrac{V^2}{c^2}}}.
\end{align}

<br>

У другого конца стержня же в это время должна быть какая-то координата. Положим $x = l$ (численно). Понятно, что:

<br>

\begin{align}
    x = \dfrac{x \,'}{\gamma}.
\end{align}

<br>

Это же справедливо для любой координаты (например, в нулевой момент времени):

<br>

\begin{align}
    x(0) = \dfrac{x\, '}{\gamma}.
\end{align}

<br>

За время $t$ смещение составит:

<br>

\begin{align}
    &x = \dfrac{x\,'}{\gamma} + Vt \nonumber \\ 
    &\Rightarrow \boxed{x\,' = \gamma(x-Vt)}. 
    \\ \nonumber
\end{align}

<br>

**Небольшое замечание :** Если мы решим перейти в систему отсчёта $K'$, то несложно будет получить аналогичные преобразования для $x$ (скорость движения $K$ относительно $K'$ в таком случае будет равна $-V$, а $K'$ и $K$ абсолютно равноправны, поэтому) :

<br>

\begin{align}
    & \boxed{x = \gamma(x\,'+Vt\,')}.
\end{align}

<br>

Теперь займемся преобразованием времени. Рассмотрим такое событие: запускаем луч из начала координат обеих систем (поскольку в начальный момент времени они совпадают) в начальные моменты времени $t = 0$ и $t \,' = 0$. Пусть в каждой системе отсчёта находится очень много синхронизированных во всякой точке часов. За время $t$ в системе $K$ луч доходит в точку с координатой $x_2 = ct$(в этой точке стоят часы, синхронизированные с началом координат). В системе отсчёта $K\,'$ это же событие имеет координату $x_2\,' = ct_2\,'$. Из преобразования координат нам известно: 

<br>

\begin{align}
    ct = x_2 = \dfrac{x_2\,'}{\gamma} + Vt = \dfrac{ct_2\,'}{\gamma} + Vt.
\end{align}

<br>

Отсюда время события $t\,'$ в системе отсчёта $K\,'$ выражается через время $t$ в системе отсчёта $K$ следующим образом:

<br>

\begin{align}
    t_2\,' = \gamma t(1-\beta), \ \text{где} \ \beta = \dfrac{V}{c}.
\end{align}

<br>

С другой стороны, часы системы $K\,'$, с точки зрения наблюдателя в системе отсчёта $K$ находятся в тот же момент времени $t$ в координате $x_1 = Vt$, однако мы знаем, что сами часы в системе отсчёта $K\,'$ показывают время: 

<br>

\begin{align}
    t_1\,' = t \sqrt{1-\dfrac{V^2}{c^2}} = \dfrac{t}{\gamma}.
\end{align}

<br>

Это событие занимает меньше времени в системе отсчёта $K\,'$. Таким образом, есть два события, которые в системе отсчёта $K$ происходят в один момент времени $t$ в разных местах. Из-за этого в системе отсчёта $K\,'$ они происходят в разные моменты времени:

<br>

\begin{align}
    &x_1 = Vt \Rightarrow t_1\,' = \dfrac{t}{\gamma}, \\
    &x_2 = ct \Rightarrow t_2\,' = \gamma t(1-\beta).
\end{align}

<br>

Рассмотрим разность во времени в системе отсчёта $K\,'$ между этими двумя одновременными событиями в системе отсчёта $K$:

<br>

\begin{align}
    \Delta t\,' = t_2\,' - t_1\,' = \gamma t(1-\beta) - \dfrac{1}{\gamma} = t(c-V) \dfrac{\gamma(1 - \beta) - \dfrac{t}{\gamma}}{c - V}. \\ \nonumber
\end{align}

<br>

С другой стороны, 

<br>

\begin{align}
    \Delta x = x_2 - x_1 = (c-V)t.
    \\ \nonumber
\end{align}

<br>

Подставляя это в выведенную ранее формулу для разницы времени и проводя некоторые преобразования, получаем:

<br>

\begin{align}
    &\Delta t\,' = \dfrac{\gamma^2(1-\beta) - 1}{\gamma(c-V)}\Delta x = \dfrac{\dfrac{1}{1-\beta^2}(1-\beta) - 1}{\gamma c (1-\beta)} \Delta x = \dfrac{\dfrac{1}{1 + \beta} - 1}{\gamma c (1-\beta)} \Delta x = \dfrac{1 - 1 - \beta}{\gamma c (1 - \beta)(1 + \beta)} \Delta x = \nonumber \\
    &= - \dfrac{\beta}{\gamma c \dfrac{1}{\gamma^2}} \Delta x = \boxed{- \dfrac{\beta \gamma}{c} \Delta x}.
\end{align}

<br>

Вернёмся немного назад. У нас есть координаты и их связь в виде какого-то события: $x\,' = \gamma(x-Vt)$. Посмотрим на время в каждой из точек в каждой из систем отсчёта: если в этой произвольной точке едут и показывают время в системе отсчёта $K\,'$ часы, то координата этих часов в системе отчёта $K$ в начальный момент времени равна $x - Vt$, то есть отставание от опорных часов в системе отсчёта $K\,'$ (часов в начале координат) составляет:

<br>

\begin{align}
    &\Delta t\,' = (x - Vt - 0)\left(-\dfrac{\gamma \beta}{c}\right) = (x - Vt)\left(-\dfrac{\gamma \beta}{c}\right). \\
    &t\,' = \dfrac{t}{\gamma} + \Delta t\,'. \\
    &t'\bigr|_{x' \neq 0} - t'_{\text{опорные}}\bigr|_{x' = 0} = \Delta t', \\
    &t' = \dfrac{t}{\gamma} + \Delta t = \dfrac{t}{\gamma} + (x - vt)\left( \dfrac{-\gamma \beta}{c} \right), \\
    &\text{При этом, чисто алгебраически } \beta^2 = 1 - \dfrac{1}{\gamma^2}. \\
    &t' = \dfrac{t}{\gamma} - x \dfrac{\gamma \beta}{c} + t \gamma \beta^2 \nonumber \\
    & \phantom{t' } = t \left( \dfrac{1}{\gamma} + \gamma \beta^2 \right) - \dfrac{\gamma \beta}{c} x \nonumber \\
    & \phantom{t' } = t \left( \dfrac{1}{\gamma} + \gamma - \dfrac{1}{\gamma} \right) - \dfrac{\gamma \beta}{c} x = \gamma \left( t - \dfrac{\beta}{c} x \right) = \gamma \left( t - \dfrac{\beta}{c} x \right).
\end{align}

<br>

Таким образом, преобразования Лоренца для одномерного случая выглядят как:

<br>

\begin{align*}
    \boxed{
            x' = \gamma (x - vt). \\
            t' = \gamma \left( t - \dfrac{\beta}{c} x \right). \\
            y' = y. \\
            z' = z.
            }
\end{align*}

<br>

Если же мы будем переходить в обратную сторону, то координата и время примут вид

<br>

\begin{align}
    &x' = \gamma (x + vt). \\
    &t' = \gamma \left( t + \dfrac{\beta}{c} x \right).
\end{align}

<br>

При решении различных задач удобно зафиксировать такие отношения между координатами во времени и пространстве: для двух событий с координатами в $K$ $(x_1, t_1)$ и $(x_2, t_2)$, а в $K'$ ― $(x_1', t_1')$ и $(x_2', t_2')$ изменения координат,

<br>

\begin{align}
    &\Delta x = x_2 - x_1, \\
    &\Delta x' = x_2' - x_1', \\
    &\Delta t = t_2 - t_1, \\
    &\Delta t' = t_2' - t_1'. \\
    \end{align}

<br>

Будут связаны как:

<br>

\begin{align}
    &\Delta x' = \gamma (\Delta x - v \Delta t), \\
    &\Delta t' = \gamma (\Delta t - \dfrac{\beta}{c} \Delta x).
\end{align}

<br>

В этот момент наших рассуждений мы замечаем, что было бы удобнее уравнять в правах время и пространство ― для этого перейдём от времени $t$ к величине $\tau = ct$, которая измеряется в метрах. Тогда наши соотношения запишутся как

<br>

\begin{align}
    &\Delta x' = \gamma (\Delta x - \beta \Delta \tau), \\
    &\Delta \tau' = \gamma (\Delta \tau - \beta \Delta x).
\end{align}

<br>

Чудеса симметрии!

## Собственное время

Собственное время какого-то объекта ― время часов, которые перемещаются в месте с ними. Покажем, что собственное время некого процесса ― минимальное среди времени этого процесса, измеряемого в различных ИСО. Вспомним об эффекте замедления времени: если есть некая система $K'$, вместе с которой движутся часы, то они и будут показывать собственное время системы.

<br>

\begin{align}
    &t_{\text{собст}} = \Delta t'.
\end{align}

<br>

При этом в $K$, относительно которого движется наша система, это же движение произойдёт за

<br>

\begin{align}
    &\Delta t = \gamma(\Delta t' - \beta \Delta x'), \\
    &\text{где, по построению, } \Delta x' = 0, \\
    &\Delta t = \gamma \Delta t' = \dfrac{\Delta t'}{\sqrt{1 - \dfrac{v^2}{c^2}}}. \\ \nonumber
\end{align}

<br>

Так как мы постулируем максимальность скорости света, то подкоренное выражение не больше единицы ― так как корень находится в знаменателе, то $\Delta t \geq \Delta t'$, причём равенство достигается лишь при $v = 0$, что мы и хотели показать.

### Регистрация мюонов на поверхности Земли

При попадании высокоэнергетческих атомов в атмосферу они вступают в реакцию, которая приводит к расщеплению старых частиц и рождению новых. Один из продуктов распада ― мюон $\mu^\pm$. Время жизни мюона в его системе отсчёта $t_\text{собст} = 2 \cdot 10^{-6}$ с. Скорость мюона составляет где-то $v = 0{,}998 c$. В классической механике это бы значило, что за своё существование он бы пролетал не больше $v t_\text{собст} = 3 \cdot 10^8 \cdot 2 \cdot 10^{-6} \approx 600$ м.

Но в реальности мюоны наблюдаются даже у поверхности Земли! Это связано с релятивистскими эффектами:

1. Для мюона продольные расстояния до поверхности значительно сокращаются,
    \begin{align}
        &l = l_0 \sqrt{1 - \dfrac{v^2}{c^2}}. \\
        &v t_\text{собст} = l = l_0 \sqrt{1 - \dfrac{v^2}{c^2}}.
    \end{align}
2. Для наблюдателя на поверхности Земли время распада мюона также становится больше
    \begin{align}
        &t = \gamma t_\text{собст} = \dfrac{t_\text{собст}}{\sqrt{1 - \dfrac{v^2}{c^2}}}. \\
        &l_0 = vt = \gamma v t_\text{собст} = \dfrac{v t_\text{собст}}{\sqrt{1 - \dfrac{v^2}{c^2}}}.
    \end{align}

С обоих точек зрения мюон достигает поверхности Земли.

## Сокращение длины

Заметим, что из наших общих преобразований промежутков пространства можно вывести полученный нами раннее результат сокращения длины.

<br>

<figure>
    <center>
    <img src="pics/130.png" width="400" align="center"/>
        <figcaption>Рис. 17.2: Сокращение длины.</figcaption>
    </center>
</figure>

<br>

Рассмотрим систему $K'$, которая движется относительно $K$ со скоростью $v$, и стержень, длина которого в $K'$ $\Delta x' = l$. Рассмотрим его длину в системе $K$ $\Delta x = l_0$ в какой-то момент времени ($\Delta t = 0$)

<br>

\begin{align}
    &\Delta x = l_0 = \gamma (\Delta x' - v\Delta t) = \gamma l, \\
    &l = \dfrac{l_0}{\gamma} = l_0 \sqrt{1 - \dfrac{v^2}{c^2}}.
\end{align}

<br>

Другой способ получить тот же результат ― рассмотреть какую-то одну точку ($\Delta x = 0$), через которую в системе $K$ пролетает сначала конец, а потом и начало стержня. Тогда, из простых кинематических соображений,

<br>

$$\require{enclose}$$
\begin{align}
    &\Delta t = \dfrac{l}{v}. \\
    &\Delta x' = x_\text{начало}' - x_\text{конец}' = - l_0. \\
    &-l_0 = \Delta x' = \gamma (\Delta x - v \Delta t) = - \gamma\ \enclose{updiagonalstrike}{v}\ \dfrac{l}{\enclose{updiagonalstrike}{v}}, \\
    &\boxed{l = \dfrac{l_0}{\gamma} = l_0 \sqrt{1 - \dfrac{v^2}{c^2}}}.
\end{align}

<br>

Таким образом, преобразования Лоренца позволяют нам получать частные результаты не хитростью ума и ловкостью рук, а выводить их из более общего принципа.
