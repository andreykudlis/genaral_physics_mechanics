## Релятивистское движение под действием постоянной силы

Рассмотрим частный случай одномерного прямолинейного движения:

<br>

$$\require{enclose}$$
\begin{align}
    &\vec{F}\parallel\vec{v} \implies \vec{F} = \dfrac{\vec{v}}{v^2}\left(\vec{F}, \vec{v}\right) \implies \nonumber\\
    &m\dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = \dfrac{1}{\gamma} \left[\dfrac{\vec{v}}{v^2}\left(\vec{F}, \vec{v}\right) - \dfrac{\vec{v}}{c^2}\left(\vec{F}, \vec{v}\right)\right] \nonumber\\\
    &m\dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = \dfrac{1}{\gamma}\vec{v}\left(\vec{F},\vec{v}\right)\left(\dfrac{c^2-v^2}{v^2c^2}\right)\nonumber\\
    &m\dfrac{\mathrm{d}v_x}{\mathrm{d}t} = \dfrac{\enclose{updiagonalstrike}{v_x}F\enclose{updiagonalstrike}{v_x}}{\enclose{updiagonalstrike}{v_x^2}\enclose{updiagonalstrike}{c^2}}\enclose{updiagonalstrike}{c^2}\left(1-\dfrac{v_x^2}{c^2}\right)\sqrt{1-\dfrac{v_x^2}{c^2}}\nonumber\\
    &m\dfrac{\mathrm{d}v_x}{\mathrm{d}t} = F \left(1-\left(\dfrac{v^2_x}{c^2}\right)\right)^{3/2} \implies\nonumber\\
    &m\dfrac{\mathrm{d}v_x}{\left(1-\left(\dfrac{v_x}{c}\right)^2\right)^{3/2}} = F\mathrm{d}t\nonumber\\
    & m\ \mathrm{d}\left(\dfrac{v_x}{\sqrt{1-\dfrac{v_x^2}{c^2}}}\right) = F\mathrm{d}t.
\end{align}

<br>

Проинтегрируем обе части:

<br>

\begin{align}
    m\left(\dfrac{v_x}{\sqrt{1-\dfrac{v^2_x}{c^2}}}\right)=Ft.
\end{align}

<br>

\begin{align}
    &v_x = \dfrac{Ft}{m}\sqrt{1-\dfrac{v_x^2}{c^2}}\nonumber\\
    &v_x^2 = \left(\dfrac{Ft}{m}\right)^2\left(1-\dfrac{v_x^2}{c^2}\right)\nonumber\\
    &v_x^2 \left(1+\left(\dfrac{Ft}{mc}\right)^2\right) = \left(\dfrac{Ft}{m}\right)^2\nonumber\\
    &v_x = \dfrac{Ft}{m}\dfrac{1}{\sqrt{1+\left(\dfrac{Ft}{mc}\right)^2}}.
\end{align}

<br>

Для рисунка 21.1: $\dfrac{v_x}{c} = \dfrac{t}{c} = \dfrac{1}{\sqrt{1+\left(\dfrac{t}{c}\right)^2}}$.

## Реактивное движение
### Формула Циолковского в СТО
Запишем ЗСЭ и И, помня, что $\gamma m U = \dfrac{EU}{c^2}$, где $E$ -- энергия отлетевшей массы, $U$ -- скорость газа относительно наблюдателя:

<br>

$$\require{enclose}$$
\begin{align}
    &\begin{cases}
    \gamma m c^2 = \gamma\;'m\;'c^2+E\\
    \gamma m v = \gamma\;'m\;'v\;'-\dfrac{EU}{c^2}
    \end{cases}\nonumber\\
    &\begin{cases}
    \gamma\;'m\;'c^2 - \gamma m c^2 = -E\\
    \gamma\;'m\;'v\;'-\gamma m v = \dfrac{EU}{c^2} = \dfrac{U}{c^2}(-1)\left(\gamma\;'m\;'c^2-\gamma m c^2\right)
    \end{cases}\nonumber\\
    &\Delta t \to 0\nonumber\\
    &\mathrm{d}(\gamma m v) = -\dfrac{U}{c^2}\mathrm{d}(\gamma mc^2)\nonumber\\
    &\gamma m\mathrm{d}v + v\mathrm{d}\gamma m = -\dfrac{U}{\enclose{updiagonalstrike}{c^2}}\enclose{updiagonalstrike}{c^2}\mathrm{d}(\gamma m)\nonumber\\
    &\gamma m\mathrm{d}v + v\mathrm{d}(\gamma m) = -U\mathrm{d}\gamma m\nonumber\\
    \dfrac{\mathrm{d}v}{v+u} = -\dfrac{\mathrm{d}\gamma m}{\gamma m} = -\mathrm{d}\ln{(\gamma m)}.
\end{align}

<br>

Параметр же выбирается относительно ракеты.

<br>

<figure>
    <center>
    <img src="pics/21.158.png" width="400" align="center"/>
        <figcaption>Рис. 21.2: Направление скорости испускания газов.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &v_x\,' = \dfrac{v_x - V}{1-\dfrac{v_xV}{c^2}},\\
    &u_0 - v = u, \\
    &\Rightarrow u = \dfrac{u_0- v}{1 - \dfrac{u_0v}{c^2}}.\\
    &\Rightarrow v + u = v + \dfrac{u_0-v}{1-\dfrac{u_0v}{c^2}} = \dfrac{v - u_0\dfrac{v^2}{c^2} + u_0 - v}{1 - \dfrac{u_0v}{c^2}}= \dfrac{u_0\left(1-\dfrac{v^2}{c^2}\right)}{1 - \dfrac{u_0v}{c^2}} \blacksquare \\
    &\dfrac{dv\left(1-\dfrac{u_0v}{c^2}\right)}{u_0\left(1-\dfrac{v^2}{c^2}\right)}= - \dfrac{d\gamma m}{\gamma m} \\
    &\Rightarrow - d\ln(\gamma m) = \dfrac{1}{u_0} \cdot \dfrac{dv}{1-\dfrac{v^2}{c^2}} - \dfrac{\dfrac{u_0v}{c^2}dv}{u_0\left(1-\dfrac{v^2}{c^2}\right)},\\
    &- d \ln[\gamma m] = \dfrac{1}{u_0} \dfrac{dv}{1-\dfrac{v^2}{c^2}} - \dfrac{1}{2} \dfrac{d\dfrac{v^2}{c^2}}{1-\dfrac{v^2}{c^2}}, \\
    &-\ln[\gamma m] \bigg|_{\gamma_1 m_1} ^{\gamma_2 m_2} = \dfrac{1}{u_0}\left[\dfrac{1}{2} \ln\left(1+ \dfrac{v}{c}\right) - \dfrac{1}{2}\ln\left(1-\dfrac{v}{c}\right) \right] + \dfrac{1}{2} \left[\ln\left(1-\dfrac{v^2}{c^2}\right)\right].
\end{align}

<br>

Разложим дробь в сумму двух других:

<br>

\begin{align}
    &\dfrac{1}{1 - \dfrac{v^2}{c^2}} = \dfrac{1}{\left(1-\dfrac{v}{c}\right)\left(1 + \dfrac{v}{c}\right)} = \dfrac{A}{1-\dfrac{v}{c}} + \dfrac{B}{1+\dfrac{v}{c}}, \\
    &A\left(1 + \dfrac{v}{c}\right) + B\left(1 - \dfrac{v}{c}\right) = 1, \nonumber \\
    &A\dfrac{v}{c} - B\dfrac{v}{c} +A +B = 1, \nonumber \\
    &A + B = 1, \nonumber \\
    & A - B = 0, \nonumber \\
    &\Rightarrow  A = B = \dfrac{1}{2}
\end{align}

<br>

Вернёмся:

<br>

\begin{align}
    &-\ln[\gamma m] \bigg|_{\gamma_1 m_1} ^{\gamma_2 m_2} = \dfrac{c}{2u_0}\ln\left[\dfrac{1+\dfrac{v}{c}}{1-\dfrac{v}{c}}\right]\bigg|_1^2 + \dfrac{1}{2}\ln\left(1-\dfrac{v^2}{c^2}\right)\bigg|_1^2.
\end{align}

<br>

Пусть мы имеем изменение скорости от 0 до $v$ и массы от $m_0$ до $m$:

<br>

\begin{align}
    &- \ln\left[\dfrac{\gamma m}{m_0}\right] = \dfrac{c}{2u_0} \ln\left[\dfrac{1+\dfrac{v}{c}}{1-\dfrac{v}{c}}\right] + \dfrac{1}{2}\ln\left(1-\dfrac{v^2}{c^2}\right) - \ln\dfrac{\gamma m}{m_0} + \ln\gamma = \dfrac{c}{2u_0}\ln\dfrac{1+\dfrac{v}{c}}{1-\dfrac{v}{c}}, \\
    &\ln\dfrac{\gamma m}{\gamma m_0} = - \dfrac{c}{2u_0} \ln\dfrac{1 + \dfrac{v}{c}}{1 - \dfrac{v}{c}}, \\
    &\ln\left[\dfrac{m}{m_0}\right] = -\dfrac{c}{2u_0} \ln\dfrac{1 + \dfrac{v}{c}}{1 - \dfrac{v}{c}}, \\
    & -2\dfrac{u_0}{c}\ln\dfrac{m}{m_0} = \ln\dfrac{1 + \dfrac{v}{c}}{1 - \dfrac{v}{c}}, \\
    &\left(\dfrac{m}{m_0}\right)^{-2\dfrac{u_0}{c}} = \dfrac{1 + \dfrac{v}{c}}{1 - \dfrac{v}{c}}, \\
    &\left(1-\dfrac{v}{c}\right)e^{-2\dfrac{u_0}{c}\ln\dfrac{m}{m_0}} = 1 +\dfrac{v}{c},\\
    &e^{-2\dfrac{u_0}{c}\ln\dfrac{m}{m_0}} = \dfrac{v}{c}\left(1 + e^{-2\dfrac{u_0}{c}\ln\dfrac{m}{m_0}}\right)+1,\\
    &\boxed{\dfrac{e^{-2\dfrac{u_0}{c}\ln\dfrac{m}{m_0}}-1}{e^{-2\dfrac{u_0}{c}\ln\dfrac{m}{m_0}}+1} \cdot c = v.}
\end{align}

## Столкновение релятивистских частиц.

Мы знаем, что "длина" 4-импульса это инвариант. Пусть у нас есть 4-импульс системы:

<br>

\begin{align}
    (\sum E_i)^2 - c^2(\sum P_i)^2 = const
\end{align}

<br>

Излучение фотона.\\
Канонически рассматриваемый вопрос: а может ли свободный электрон поглотить или излучить фотон?\\
Пусть на покоящийся электрон (перешли в такую систему отсчёта) налетает фотон.\\
ЗСЭ:

<br>

\begin{align}
    E_{ph} + mc^2 = \gamma m c^2 = E_{el}.
\end{align}

<br>

ЗСИ:

<br>

\begin{align}
    &\underset{\text{до}}{P_{ph}} = \underset{\text{после}}{P_{el}} \Rightarrow E_{ph} = \sqrt{E_{el}^2- (mc^2)^2}, \\
    &cP_{ph} = cP_{el} = \sqrt{E_{el}^2-(mc^2)^2}\\
    &\Rightarrow E_{ph} = \sqrt{\gamma^2(mc^2)^2 - (mc^2)^2} = mc^2 (\gamma^2 - 1)^{1/2}, \\
    &E_{ph} = mc^2 (\gamma - 1).
\end{align}

<br>

Подставим $E_{ph}$ из ЗСЭ:

<br>

\begin{align}
    (\gamma - 1)^2 = \gamma^2 - 1 \Rightarrow \gamma = 1 \Rightarrow E_{ph} = 0.
\end{align}

## Слияние и распад частиц

Рассмотрим распад частицы массой $M$ на две массами $m_1$ и $m_2$. Считая систему замкнутой, рассмотрим энергию и импульс в системе отсчёта, в которой первоначальная частица покоилась. Обозначим как $E_1$, $E_2$ ― энергии распавшихся частиц, как $\vec{p}_1$, $\vec{p}_2$ ― импульсы распавшихся частиц.

<br>

$$\require{enclose}$$
\begin{align}
    &Mc^2 = E_1 + E_2. \\
    &0 = \vec{p}_1 + \vec{p}_2. \\
    &-\vec{p}_1 = \vec{p}_2, \\
    &p_1^2 = p_2^2, \\
    &E_1^2 - (m_1 c^2)^2 = E_2^2 - (m_2 c^2)^2. \\
    &E_1^2 = E_2^2 + (m_1 c^2)^2 - (m_2 c^2)^2, \\
    &E_1 + E_2 = Mc^2, \\
    &E_1^2 = (Mc^2 - E_1)^2 + (m_1 c^2)^2 - (m_2 c^2)^2, \\
    &\enclose{updiagonalstrike}{E_1^2} = (Mc^2)^2 - 2 E_1 M c^2 + \enclose{updiagonalstrike}{E_1^2} + (m_1 c^2)^2 - (m_2 c^2)^2, \\
    &2 E_1 M c^2 = (M c^2)^2 + (m_1 c^2)^2 - (m_2 c^2)^2, \\
    &E_1 = \dfrac{(M c^2)^2 + (m_1 c^2)^2 - (m_2 c^2)^2}{2 M c^2}
\end{align}

<br>

В случае слипания можно перейти в систему, где одна из частиц покоится, после чего найти общую массу.

<br>

$$\require{enclose}$$
\begin{align}
    &E_1 + E_2 = E, \\
    &E_1 + m_2 c^2 = E. \\
    &\vec{p} = \vec{p}_1 + \underbrace{\vec{p}_{2}}_{0} = \vec{p}_1, \\
    &\vec{p} = \gamma m \vec{v}_c = \dfrac{\gamma m c^2 \vec{v}_c}{c^2} = \dfrac{E \vec{v}_c}{c^2}. \\
    &E^2 - p^2c^2 = \left(MC^2\right)^2 \\
    &E^2 - p^2c^2 = \left(E_1+m_2c^2\right)^2 - c^2p^2 = (E_1+m_2c^2)^2 - c^2p_1^2 \nonumber\\
    &= m_1c^2+(m_2c^2)^2+2E_1m_2c^2 \\
    &\implies M = \dfrac{1}{c^2}\sqrt{m^2_1c^4+m_2^2c^4+2E_1m_2c^2} \\
    &M = \sqrt{m_1^2+m^2_2+\dfrac{2E m_2}{c^2}}.
\end{align}

<br>

## Кинематика абсолютно твёрдого тела
Пока что вернёмся к классической механике.

Абсолютно твёрдое тело ― модель тела как системы материальных точек, разделённых между собой постоянным расстоянием.

Полная масса абсолютно твёрдого тела из конечного числа частиц это всего лишь сумма их масс,

<br>

\begin{align}
    &M = \sum_i m_i.
\end{align}

<br>

Если же тело известно нам как некий объём с плотностью, которая как-то меняется в пространстве, то полную массу можно найти, проинтегрировав по поверхности:

<br>

\begin{align}
    &M = \int \limits_V \rho (\tilde{V}) \mathrm{d}\tilde{V}.
\end{align}

<br>

### Инвариантность угловой скорости
Рассмотрим произвольную точку абсолютно твёрдого тела относительно некой другой точки этого тела $O$, имеющей радиус-вектор $R_O$.

<br>

<figure>
    <center>
    <img src="pics/lecture21_164.png" width="400" align="center"/>
        <figcaption>Рис. 21.3: Радиус-вектор точки относительно O и O'.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &\vec{r} = \vec{R}_O + \vec{r}_O.
\end{align}

<br>

Рассмотрим приращение этого вектора за малое время $\mathrm{d}t$,

<br>

\begin{align}
    &\mathrm{d}\vec{r} = \mathrm{d}\vec{R}_O + \mathrm{d}\vec{r}_O = \mathrm{d}\vec{R}_O + [\mathrm{d}\vec{\varphi} \times \vec{r}_O], 
\end{align}

<br>

то есть, движение точки можно разложить на поступательное движение $O$ и вращение вокруг $O$.

<br>

\begin{align}
    &\vec{v} = \dfrac{\mathrm{d}\vec{r}}{\mathrm{d}t} = \vec{V}_O + [\vec{\omega} \times \vec{r}_O].
\end{align}

<br>

Теперь повторим наши рассуждения для движения точки не относительно $O$, а относительно какой-то третьей точки тела ― $O'$,

<br>

\begin{align}
    &\vec{v} = \vec{V}_{O'} + [\vec{\omega}' \times \vec{r}_{O'}]. \\
    &\vec{V}_O + [\vec{\omega} \times \vec{r}_O] = \vec{V}_{O'} + [\vec{\omega}' \times \vec{r}_{O'}]. \\
\end{align}

<br>

Завершим «любовный треугольник» наших трёх точек, рассмотрев движение $O$ относительно $O'$.

<br>

$$\require{enclose}$$
\begin{align}
    &\vec{R_O} = \vec{R}_{O'} + \vec{R}_{O'O}, \\
    &\vec{V}_O = \vec{V}_{O'} + [\vec{\omega}' \times \vec{R}_{O'O}]. \\
    &\vec{r}_{O'} = \vec{r_O} + \vec{R}_{O'O}. \\
    &\enclose{updiagonalstrike}{\vec{V}_{O'}} + \enclose{updiagonalstrike}{[\vec{\omega}' \times \vec{R}_{O'O}]} + {\vec{\omega} \times \vec{r}_O} = \enclose{updiagonalstrike}{\vec{V}_{O'}} + [\vec{\omega}' \times \vec{r}_O] + \enclose{updiagonalstrike}{[\vec{\omega}' \times \vec{R}_{O'O}]}, \\
    &[\vec{\omega} \times \vec{r}_O] = [\vec{\omega}' \times \vec{r}_O], \\
    &\boxed{\vec{\omega} = \vec{\omega}'}.
\end{align}

<br>

Последнее следует из того, что мы не накладывали дополнительных условий на $\vec{r}_O$, а значит равенство выполняется для всех таких векторов без исключения.

Таким образом, угловая скорость не зависит от выбора точки отсчёта абсолютно твёрдого тела.

### Плоское движение абсолютно твёрдого тела
Плоское движение ― движение, при котором все точки двигаются параллельно некоторой плоскости, которую называют плоскостью вращения. При таком движении всегда можно выделить в каждый момент времени выделить ось, вокруг которой происходит вращение всех точек абсолютно твёрдого тела.

Выразим это формульной записью:

<br>

\begin{align}
    &\vec{v} = \underbrace{\vec{v}_c}_0 + [\omega \times \vec{r}_c] = [\omega \times \vec{r}_c].
\end{align}

<br>

    <center>
    <img src="pics/lecture21_166-1.png" width="400" align="center"/>
        <figcaption>Рис. 21.4: Положение точки относительно оси вращения и точки O.</figcaption>
    </center>
</figure>

<br>

Мы показали, что скорость данной точки относительно любой другой произвольной можно представить как

<br>

\begin{align}
    &\vec{v} = \vec{V}_O + [\vec{\omega} \times \vec{r}_O].
\end{align}

<br>

Обозначим положение мгновенной оси вращения как $\vec{R}_c$, тогда $\vec{r}_O = \vec{R}_{OC} + \vec{r}_c$.

<br>

\begin{align}
    &\vec{V} = \vec{V_O} + [\vec{\omega} \times (\vec{R}_{OC} + \vec{r}_c)] = [\vec{\omega} \times \vec{r}_c]. \\
    &[\vec{\omega} \times \vec{V}_O] = [\vec{\omega} \times [- \vec{\omega} \times \vec{R}_{OC}]] = - (\vec{\omega} (\vec{\omega} \cdot \vec{R}_{OC}) - \vec{R}_{OC} (\vec{\omega} \cdot \vec{\omega})). \\
    &\text{Заметим, что } \omega \perp \vec{R}_{OC}, \\
    &\vec{\omega} \times \vec{V}_O = \omega^2 \vec{R}_{OC}. \\
    &\boxed{\vec{R}_{OC} = \dfrac{\vec{\omega} \times \vec{V_O}}{\omega^2}}. \\
    *\text{При этом }\vec{V}_O \perp \vec{\omega}, \\
    &|\vec{R}_{OC}| = \dfrac{\omega V_O}{\omega^2} = \dfrac{\omega \omega R}{\omega^2} = R.
\end{align}
