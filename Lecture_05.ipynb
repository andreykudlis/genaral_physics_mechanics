# Механическая работа. Энергия.
Работа (интуитивная версия): что-то, что определяется прикладываемой силой объемом "результатов", которых мы достигли, прикладывая данную силу.

**Определение.** Механической работой силы $\vec{F}$ на перемещение d$\vec{r}$ называется величина:

<br>

\begin{align}
    \delta A = \vec{F} \cdot d\vec{r}\; =\; (\vec{F}, d\vec{r})\;=\; |\vec{F}|\cdot |d\vec{r}|\cdot \cos{(\angle (\vec{F}, d\vec{r}))}.
\end{align}

**Пример.**
Пусть $B(x,y)\;=\;3xy\;+\;8y$, тогда dB:

<br>

\begin{equation} 
\begin{split}
   dB \ = \ 3ydx \ + \ 3xdy\;+\;8dy\;=\;(3y, 3x\;+\;8)\cdot (dx, dy)\;=\; \vec{F} \cdot d\vec{r}. \nonumber
\end{split}
\end{equation}

<br>

\begin{align}
    \int_{A}^{B} \vec{F} \cdot d\vec{r} \;=\; \int_{A}^{B} dB\;=\; 12\;+\;16\;-\; (3\;+\;8)\;=\; 17.\nonumber
\end{align}

<br>

Или $\Rightarrow$

<br>

\begin{align}
     \int_{A}^{B}\;=\;\int_{1,1}^{2,1}\;+\;\int_{2,1}^{2,2}\;=\;3\cdot \int_{1}^{2} dx \;+\; 14\cdot \int_{1}^{2} dy \;=\;17.\nonumber
\end{align}

<br>

\begin{align}
     \int_{A}^{B}\;=\;\int_{1,1}^{1,2}\;+\;\int_{1,2}^{2,2}\;=\;11\cdot \int_{1}^{2} dy \;+\; 6\cdot \int_{1}^{2} dx \;=\;17.\nonumber
\end{align}

<br>

А если:

<br>

$\vec{F}\;=\;(3y^2, 4x\;+\;8)$, т.е. нельзя найти такое u: $dB\;=\;(\vec{P},d\vec{r})$

<br>

\begin{align}
     \int_{A}^{B}\;=\;\int_{1,1}^{2,2}\;+\;\int_{2,1}^{2,2}\;=\;3\cdot \int_{1}^{2} dx \;+\; 16\cdot \int_{1}^{2} dy \;=\;19.\nonumber
\end{align}

<br>

\begin{align}
     \int_{A}^{B}\;=\;\int_{1,1}^{1,2}\;+\;\int_{1,2}^{2,2}\;=\;12\cdot \int_{1}^{2} dy \;+\; 12\cdot \int_{1}^{2} dx \;=\;24.\nonumber
\end{align}

<br>

$19\;\neq\;24$, т.е. $3y^{2}dx\;+\;(4x\;+\;8)dy \;-\;$ не полный дифференциал, а $3ydx\;+\;(3x\;+\;8)dy \;-\;$ полный дифференциал.

Выведем условие (в двумерии) того, что сила является потенциальной, тогда 

<br>

\begin{align}
    F_{x}=-\dfrac{\partial U}{\partial x}\Rightarrow \dfrac{\partial F_{x}}{\partial y}=-\dfrac{\partial^{2} U}{\partial x \partial y}\\ \nonumber
    F_{y}=-\dfrac{\partial U}{\partial y} \Rightarrow \dfrac{\partial F_{y}}{\partial x}=-\dfrac{\partial^{2} U}{\partial x \partial y}\\ \nonumber
    \dfrac{\partial F_{x}}{\partial y}=\dfrac{\partial F_{y}}{\partial x}\\ \nonumber
    \dfrac{\partial F_{y}}{\partial x}-\dfrac{\partial F_{x}}{\partial y}=[\nabla \times \vec F]_{z}=0\\ \nonumber
\end{align}

<br>

В трёхмерном случае условие потенциальности силы будет выглядеть аналогично, только за счёт того, что у нас не 2 координаты, а 3, надо будет приравнивать к 0 разности частных производных проекций силы F для всех пар координат (их всего 3) : $[\nabla \times \vec F]=0$

## Кинетическая энергия
Введём величину: $E_\text{к} = K = T = \frac{mv^2}2$ – кинетическая энергия материальной точки.
<br>
Зачем вводить такую величину?
<br>
Рассмотрим работу суперпозиции произвольных сил по  перемещению материальной точки:

<br>

\begin{align}
    & A = \int^2_1 \vec{F}d\vec{r} = \int^2_1\vec{F}\vec{v}dt =\nonumber\\
    & = \int^2_1m\frac{d\vec{v}}{dt}\vec{v}dt = m\int^2_1\vec{v}d\vec{v} =\nonumber\\
    & = m\int^2_1 vdv = m\int^2_1\frac{dv^2}{2} = \int^2_1d\frac{mv^2}{2} = \nonumber\\
    & = \frac{mv^2}{2}\Bigg | ^2_1 = T_2 - T_1.
\end{align}

<br>

Таким образом, суммарная работа сил по перемещению материальной точки равна разности её кинетических энергий.

<br>

\begin{align}
    & \vec{V}\cdot\vec{V} = V^2 \Rightarrow d(\vec{V},\vec{V}) = \vec{V}d\vec{V}+d\vec{V}\vec{V}=\nonumber\\
    & = 2 (\vec{V}, d\vec{V}) = 2VdV \nonumber \Rightarrow\\
    & \Rightarrow \vec{V}\cdot d\vec{V} = VdV.
\end{align}

<br>

**Кинетическая энергия системы материальных точек** есть сумма кинетических энергий всех материальных точек.
<br>
Кинетическая энергия **зависит** от выбора системы отсчёта!

<br>

\begin{align}
    T_{\text{системы}} = \sum^N_{i=1} T_i = \sum^N_{i=1} \frac{m_iv_i^2}{2}.
\end{align}

<br>

Напишем закон преобразования кинетической энергии при переходе к другой системе отсчёта:

<br>

$K$ и $K'$, пусть $K'$ движется со скоростью $\vec{V}$ относительно $K, \vec{v}_i$ – скорость i-ой мат. точки в системе $K$, $\vec{v}_i\text{ }'$ – скорость i-ой мат. точки в системе $K'$, тогда  $\Rightarrow$

<br>

\begin{align}
& \Rightarrow \vec{v}_i = \vec{v}_i\text{ }' + \vec{V} \nonumber \\
    & \Rightarrow \vec{v}_i^2 = v_i^2 = \left( \vec{v}_i\text{ }' + \vec{V} \right)^2 = \nonumber \\
    & = \vec{v}_i\text{ }'^2 + 2\vec{v}_i\text{ }'\vec{V}+\vec{V}^2 = \nonumber \\
    & = v_i'^2 + V^2 + 2\vec{v}_i\text{ }'\vec{V}. \\
    & \Rightarrow T_{\text{к}} = \sum^N_{i=1}\frac{m_iv_i^2}{2} = \sum^N_{i=1} \frac{m_i}{2} \left(v_i'^2 + V^2 + 2\vec{v}_i\text{ }'\vec{V}\right) = \nonumber \\
    & = T_{\text{к'}} + \frac{MV^2}{2} + \sum^N_{i=1} m_i\vec{v}_i\text{ }'\vec{V} = \nonumber \\
    & = T_{\text{к'}} + \frac{MV^2}{2} + \left( \vec{P}_{k}, \vec{V} \right).
\end{align}

## Теорема Кёнига.
Вспоминаем: 

<br>

\begin{align}
\vec{ṙ}_c = \dfrac{\sum_{i=1}^Nm_i\: \vec{v}_i}{\sum_{i=1}^Nm_i}, \ \  \vec{r}_c = \dfrac{\sum_{i=1}^Nm_i\: \vec{r}_i}{\sum_{i=1}^Nm_i},
\end{align}

<br>

Импульс в системе центра масс равен: 

<br>

\begin{align}
    \sum_{i=1}^Nm_i \: \vec{v}_i \, ' = \sum_{i=1}^Nm_i[\vec{v}_i-\vec{v}]= \sum_{i=1}^Nm_i \: \vec{v}_i - M \cdot \dfrac{\sum_{i=1}^Nm_i \: \vec{v}_i }{\sum_{i=1}^Nm_i}=0.
\end{align}

<br>

Получаем теорему Кёнига:

<br>

\begin{align}
    T_k=T_{k \, '} + \dfrac{MV^2}{2}
\end{align}

<br>

где $T_k\geqslant0$ - полная кинетическая энергия системы k, $T_k \, '$ - полная кинетическая энергия системы k', $MV^2/2\geqslant0$ -  кинетическая энергия центра масс. Таким образом, кинетическая энергия будет минимальна в системе центра масс.
## Потенциальная энергия.
В том случае, когда работа выполняется консервативными (потенциальными) силами, она будет равна изменению некоторой функции. Рассмотрим консервативную силу $\vec{F}$:

<br>

\begin{align}
    & \delta A = \vec{F}\cdot d\vec{r}, \; \; \oint \delta A = 0 \Rightarrow \delta A =dB, \nonumber \\
    & \oint dB=0=\int_1^2 dB + \int_2^1 dB= B(2) - B(1) + B(1) - B(2) = 0, \nonumber \\
    & \text{т. е. } dB= \vec{F}\cdot d\vec{r}= F_xdx + F_ydy + F_zdz= \left(\vec{\nabla}B, d\vec{r}\right) = -dU = - \left(\nabla U, d\vec{r}\right).
\end{align}

<br>

\begin{align}
\boxed{ \vec{F}=-\nabla U }    
\end{align}

<br>

<a name="nabla"></a>

<p> Небольшое математическое отступление : $\nabla$ - векторный дифференциальный оператор набла, над ним обычно вектор не ставят</p>

<br>

<h4><center>
$\nabla=\vec e_{x} \dfrac{\partial}{\partial x}+\vec e_{y} \dfrac{\partial}{\partial y} + \vec e_{z}\dfrac{\partial}{\partial z}$
</h4></center>

$\vec{F}$ - Потенциальная сила, то есть выражается через $\nabla$ скалярной функции $U$ - потенциальной энергии.\\

## Работа центральной силы

Центральная сила ― такая сила, линия действия которой всегда проходит через какую-то заданную точку (силовой центр). Если обозначить радиус-вектор тела относительно силового центра как $\vec{r}$, то центральную силу можно выразить как

<br>

\begin{align}
    &\vec{F}(\vec{r}) = f(r) \cdot\dfrac{\vec{r}}{r}.
\end{align}

<br>

Рассмотрим работу такой силы на примере Ньютоновой силы всемирного тяготения между объектами массами $m$ и $M$:

<br>

\begin{align}
    &\vec{F}(\vec{r}) = - \gamma \dfrac{mM}{r^2} \dfrac{\vec{r}}{r^3} \\
    \int \limits^2_1 \mathrm{\delta} &A = \int \limits^2_1 \vec{F}(\vec{r}) \cdot \mathrm{d}\vec{r} = - \gamma mM \int \limits^2_1 \dfrac{\vec{r} \cdot \mathrm{d} \vec{r}}{r},
\end{align}

<br>

где, при независимости работы от траектории, траекторию можно разложить на часть, сонаправленную силе ($\vec{r}\cdot\mathrm{d}\vec{r} = r\mathrm{d}r$), и перпендикулярную ей ($\vec{r}\cdot\mathrm{d}\vec{r} = 0$), то есть

<br>

\begin{align}
    &= - \gamma mM \int \limits^2_1 \dfrac{r \mathrm{d}r}{r^3} = - \gamma mM \int \limits^2_1 \dfrac{\mathrm{d}r}{r^2} \\
    &= -\gamma mM \left.\Bigl(-\dfrac{1}{r}\Bigr)\right|^2_1 = - \gamma mM \Bigl( \dfrac{1}{r_1} - \dfrac{1}{r_2}\Bigr) \\
    &U = - \gamma \dfrac{mM}{r} + \text{const}.
\end{align}

<br>

Для удобства потенциальную энергию бесконечно удалённого от силового центра объекта примем нулевой, что приводит нас к

<br>

\begin{align}
    &U = - \gamma \dfrac{mM}{r}.
\end{align}

## Полная механическая энергия

Рассмотрим векторную сумму всех $N$ сил, действующих на некое тело. Разложим получившуюся равнодействующую силу на консервативную и неконсервативную компоненты $\vec{F}_\textit{к}$ и $\vec{F}_\textit{нк}$.

<br>

\begin{align}
    &\sum^N_{k = 1} \vec{F}_k = \vec{F}_\textit{к} + \vec{F}_\textit{нк}.
\end{align}

<br>

Тогда работа этих сил при перемещении из некого положения 1 в положение 2

<br>

\begin{align}
    &A_{1 \to 2} = \int \limits^2_1 \vec{F} \cdot \mathrm{d}\vec{r} \\
    &= \int \limits^2_1 \vec{F}_\textit{к} \cdot \mathrm{d}\vec{r} + \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r} \\
    &= -(U_2 - U_1) + \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r}.
\end{align}

<br>

При этом совершённая работа будет затрачена на изменение кинетической энергии тела,

<br>

\begin{align}
    &A_{1 \to 2} = T_2 - T_1. \\
    &T_2 - T_1 = -(U_2 - U_1) + \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r} \\
    &(T_2 - T_1) + (U_2 - U_1) = \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r} \\
    &(T_2 + U_2) - (T_1 + U_1) = \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r}. \\
\end{align}

<br>

Величину $T + U$ называют полной механической энергией и обозначают $E$, что приводит нас к

<br>

\begin{align}
    E_2 - E_1 = \int \limits^2_1 \vec{F}_\textit{нк} \cdot \mathrm{d}\vec{r}.
\end{align}

<br>

**Определение :** Полная механическая энергия - скалярная физическая величина, равная сумме потенциальной и кинетической энергий : $E=T+U$.

Стоит заметить, что если в системе действуют лишь консервативные силы, то полная механическая энергия $E$ сохраняется! Этот так называемый «интеграл движения» очень удобен при рассмотрении конкретных ситуаций.
