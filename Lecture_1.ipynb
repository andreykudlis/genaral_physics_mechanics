### Анализ размерностей
Рассмотрим случай равноускоренного движения.
Пусть начальная скорость равна нулю. Какова зависимость скорости от координаты?

Из точного решения мы узнаем:
\begin{align}
    \upsilon=\sqrt{2ax},\; \; \; \;  \text{где a - ускорение, x - координата.}
\end{align}


Получим данное соотношение из размерного анализа:
Самая простая зависимость:
\begin{align}
    \upsilon^l=a^k \cdot x^m \Rightarrow \upsilon=a^{\tfrac{k}{l}}\cdot x^{\tfrac{k}{l}} =a^{k'} \cdot x^{m'} \; \underset{\text{новые k,m}}\Rightarrow \upsilon=a^k \cdot x^m
\end{align}

\begin{align}
    {[V]}=\dfrac{L}{T}; \; {[a]}=\dfrac{L}{T^2}; \; {[x]}= \text{L};
\end{align}

\begin{align}
    \Rightarrow L \cdot T^{-1}= L^k \cdot T^{-2k} \cdot L^m \Rightarrow
    \begin{cases}
1 = k+ m,\\
-1=-2k.
\end{cases} \Rightarrow
\begin{cases}
k=\tfrac{1}{2},\\
m=\tfrac{1}{2}.
\end{cases} \Rightarrow \upsilon \approx \sqrt{ax},\;
\end{align}
что достаточно близко к точному ответу $\sqrt{2ax}$.

Повторим эти же рассуждения для вывода скорости звука. От чего может зависеть скорость?

\begin{align} c = f(\rho, P, T) = g(\rho, P) \end{align}

Самая простая форма:

\begin{align} c = P^l \cdot \rho^k, \; \;
    {[\rho]}=\dfrac{M}{L^3}; \; {[P]}=\dfrac{M}{L \cdot T^2};
\end{align}

\begin{align}
    \Rightarrow L \cdot T^{-1} = M^l \cdot L^{-l} \cdot T^{-2l} \cdot M^k \cdot L^{-3k}
\end{align}

\begin{align}
    \text{(т.к.} {[c]}=\dfrac{M}{C}\text{(СИ), либо} \ {[c]}=\dfrac{L}{T})
\end{align}

 \begin{align}
    \Rightarrow M^l \cdot M^k = 1 \Rightarrow l=-k
\end{align}
\begin{align}
    \Rightarrow L \cdot T^{-1} =  L^k \cdot T^{2k} \cdot L^{-3k} = L^{-2k} \cdot T^{2k}
\end{align}
\begin{align}
    \Rightarrow 1=-2k \Rightarrow k =- \dfrac{1}{2} \Rightarrow l = \dfrac{1}{2}
\end{align}
\begin{align}
    \Rightarrow c=\sqrt{\dfrac{P}{\rho}},\; \text{точное же решение} \; c=\sqrt{\gamma \cdot \dfrac{P}{\rho}}. \; \text{Для воздуха:}\; \sqrt{\gamma} \approx 1.18
\end{align}


### Производная
**Виды обозначений:**
\begin{align}
    ẋ(t), \;x'(t), \;x'_t,\; ẋ,\; \dfrac{dx}{dt}\; \text{ и т.д.}
\end{align}

**Определение.** Функция $f(x)$ имеет производную в точке $x_0$, если существует следующий предел:
\begin{align}
    &f'(x_0)=\dfrac{df}{dx}(x_0)=\underset{\Delta x\rightarrow 0}{lim}\dfrac{f(x_0+\Delta x)-f(x_0)}{\Delta x}
\end{align}

**Пример:**

\begin{align}
    f(x)=x^2, \; \underset{\Delta x\rightarrow 0}{lim}\dfrac{{(x_0+\Delta x)}^2-{x_0}^2}{\Delta x} = 2x_0
\end{align}

**Определение.** Дифференциал функции f в точке $x_0$ может быть определен как линейная по аргументу часть приращения функции:

\begin{align}
   df=f'(x_0) \cdot dx
\end{align}

**Пример:**

\begin{align}
    \Delta f(x)=f(x + \Delta x) - f(x) = 2x \Delta x + {\Delta x}^2 \\
    df(x)=2xdx, \; \text{в физике часто заменяют} \; \Delta x \rightarrow dx
\end{align}

Необходимо помнить, что само приращение (d) является математической абстракцией (моделью бесконечно малой величины).

Пользуйтесь дифференциалами, т.к. с ними можно работать как с дробями.


**Рассмотрим несколько свойств:**
1) Производная обратной функции:

\begin{align} \text{Если} \; f(x)=x^2,\; x=\sqrt{f}, \; f > 0 \\
\dfrac{dx}{df}= \dfrac{1}{\dfrac{df}{dx}}= \dfrac{1}{2x} = \dfrac{1}{2 \sqrt{f}} =x'(f)
\end{align}

2) Производная сложной функции:

\begin{align} f'_x(y(x))=f'_y(y) \bigg|_{y=y(x)} \cdot y'_x(x)
\end{align}

Как это доказать? При помощи дифференциалов это очень просто:

\begin{align} \dfrac{df}{dx}(y(x)) = \dfrac{df}{dy}\bigg|_{y=y(x)} \cdot \dfrac{dy}{dx}(x)
\end{align}

**Пример:**

\begin{align}
    f(x)=sin^3x^2
\end{align}
\begin{align}
    f'(x)=6 \cdot sin^2x^2\cdot cosx^2 \cdot 2x
\end{align}

**Как получить?**

\begin{align}
W(z(y(x)))=sin^3x^2
\end{align}
\begin{align}
W(z)=z^3  \; ; \;  z(y)=siny  \; ; \;  y(x)=x \; ?
\end{align}
\begin{align}
\Rightarrow \dfrac{df}{dx}=\dfrac{dW}{dx}=\dfrac{dW}{dz} \bigg|\dfrac{dz}{\underset{z=z(y(x))}{dy}} \bigg|\dfrac{dy}{\underset{y=y(x)}{dx}}=3z^2\bigg|_{z=z(y(x))} \cdot cosy \bigg|_{y=y(x)}\cdot 2x= \\ = 3sin^2y \bigg | _ {y=y(x)} \cdot cosx^2 \cdot 2x = 6sin^2x^2 \cdot cosx^2 \cdot 2x
\end{align}

Что ещё? **Полезное свойство:**

\begin{align}
    d(f\cdot g)=fdg+gdf
\end{align}

### Интеграл
**Определение.** Неопределённым интегралом функции $f(x)$ (первообразная) называется другая функция $F(x):F'(x)=f(x)$

\begin{align}
    F(x) = \int f(x)dx + const
\end{align}

**Определение.** Определённым интегралом от функции $f(x)$  на промежутке $[a,b]$ $\exists$ $\Leftrightarrow$ $\exists$  $lim$ $следующей суммы:$  $\underset{\mu(\Delta x_i)\rightarrow0}{lim} \sum\limits_{i=1}^n f(g_i) \Delta x $.

**Обозначение:**
\begin{align}
    \int\limits_a^b f(x)dx = \underset{\mu(\Delta x_i)\rightarrow 0}{lim} \sum\limits_{i=1}^n f(g_i) \Delta x_i
\end{align}

**Геометрический смысл:** Площадь $S$ фигуры под графиком функции.

### Система координат и система отсчёта
**Определение.** Система отсчёта - совокупность системы координат, начала отсчёта, привязанного к некоторому физическому объекту, и отсчитывающих время часов. \
**Определение.** Система координат - способ определения положения и перемещения точки (тела) при помощи чисел (символов).

На курсе мы будем использовать три системы координат: декартову, цилиндрическую (полярную), сферическую.

Положение точки определяется координатами $(r, \theta, \varphi)$.

\begin{align}
\overrightarrow{r} = \overrightarrow{e}_r \cdot r, \quad \overrightarrow{e}_r, \overrightarrow{e}_\theta, \overrightarrow{e}_\varphi.
\end{align}
\begin{align}
\text{Связь координат: }
\begin{cases}
x = r\sin{\theta}\cos{\varphi}\\
y = r\sin{\theta}\sin{\varphi}\\
z = r\cos{\theta}
\end{cases}
\end{align}
\begin{align}
\text{В другую сторону: }
\begin{cases}
r = \sqrt{x^2+y^2+z^2}\\
\theta = \arccos{\left[\frac z{\sqrt{x^2+y^2+z^2}}\right]}\\
\varphi = \text{arctg} \left[ \frac yx \right]
\end{cases}
\end{align}
\begin{align}
\text{Связь между ортами: }
\begin{cases}
\overrightarrow{e}_r = \left(\overrightarrow{e}_x\cos{\varphi}+\overrightarrow{e}_y\sin{\varphi} \right) \sin{\theta} + \overrightarrow{e}_z\cos{\theta}\\
\overrightarrow{e}_{\varphi} = - \overrightarrow{e}_x\sin{\varphi}+\overrightarrow{e}_y\cos{\varphi}\\
\overrightarrow{e}_{\theta} = \left(\overrightarrow{e}_x\cos{\varphi}+\overrightarrow{e}_y\sin{\varphi} \right) \cos{\theta} - \overrightarrow{e}_z\sin{\theta}
\end{cases}
\end{align}
**Упражнения:**
$\text{выразить} \overrightarrow{e}_x, \overrightarrow{e}_y, \overrightarrow{e}_z \text{через } \overrightarrow{e}_r, \overrightarrow{e}_{\theta}, \overrightarrow{e}_{\varphi}$.
