\begin{align}
    |\vec{M}|=M=|\vec{F}||\vec{r}|\cdot \sin\alpha ; \\
    |\vec{r}|=r; |\vec{r}|\cdot \sin\alpha = l -\text{плечо.}
\end{align}

<br>

Момент силы является аддитивной величиной: 
Момент сил системы материальных точек равен сумме моментов сил для каждой материальной точки:

<br>

\begin{align}
    \vec{M}=\sum _i \vec{M_i} = \sum_i \vec{r_i} \times \vec{F_i}
\end{align}

<br>

Запишем правило преобразования момента силы при переходе из одной системы отсчёта в другую:

<br>

\begin{align}
    & \vec{r}_i = \vec{r}_i\;' + \vec{R}.\\
    & \vec{M} = \sum_i\left(\vec{r}_i\;' + \vec{R}\right) \times \vec{F}_i =\nonumber\\
    & = \sum_i\vec{r}_i\;'\times\vec{F}_i+\vec{R}\times\sum_i\vec{F}_i=\nonumber\\
    & =\vec{M}\;'+\vec{R}\times\vec{F}, \qquad \blacksquare
\end{align}

<br>

$\vec{F}$ -- равнодействующая всех сил.

### Небольшое замечание : 
 В векторном произведении можно раскрывать скобки, однако нельзя менять порядок множителей, так как это приведёт к изменению знака векторного произведения

# Момент импульса
### Момент импульса материальной точки -- вектор, который определяется так:

<br>

\begin{align}
    \vec{L} = \vec{r} \times \vec{p}.
\end{align}

<br>

Для системы материальных точек:

<br>

\begin{align}
    \vec{L} = \sum_i \vec{L}_i = \sum_i \vec{r}_i \times \vec{p}_i.
\end{align}

<br>

Как преобразуется момент импульса (если $\vec{r}_i = \vec{r}_i\;'+\vec{R}$):

<br>

\begin{align}
    & \vec{L} = \sum_i\vec{r}_i\times\vec{p}_i=\sum_i\left(\vec{r}_i\;'+\vec{R}\right)\times m_i\left(\vec{v}_i\;'+\vec{V}\right)=\nonumber\\
    & =\sum_im_i\vec{r}_i\;'\times\vec{v}_i\;' + \sum_im_i\vec{r}_i\;'\times\vec{V}+\sum_im_i\vec{R}\times\vec{v}_i\;'+\sum_im_i\vec{R}\times\vec{V} =\nonumber\\
    & = \vec{L}\;'+(-\vec{V})\times\sum_im_i\vec{r}_i\;'+\vec{R}\times\sum_im_i\vec{v}_i\;'+M\vec{R}\times\vec{V}=\nonumber\\
    & =\vec{L}\;'+M\vec{R}\;'_{c.m.}\vec{V}+\vec{R}\times\vec{p}\;'+M\vec{R}\times\vec{V}.\\ \nonumber
\end{align}

<br>

$\vec{R}\;'_{c.m.}=\dfrac{\sum_i m_i\vec r_i\;'}{M}$ - по определению радиус-вектора центра масс

Выражение упрощается для системы центра масс :

<br>

\begin{align}
    \boxed{
    \vec{L} = \vec{L}_{c.m.} + M\vec{R}_{c.m.}\times\vec{V}_{c.m.},
    }\\
    \vec{p}\;'=0, \quad \vec{R}\;'_{c.m.} = 0.
\end{align}

<br>

Получим связь между $\vec{L}$ и $\vec{M}$:
1. Для одной материальной точки:

<br>

\begin{align}
    & \vec{L} = \vec{r} \times \vec{p}\\
    & \frac{d\vec{L}}{d{t}} = \frac{d}{dt}\vec{r}\times \vec{p} + \vec{r}\times\frac{d\vec{p}}{dt} =\nonumber\\
    & = \vec{v}\times m\vec{v} + \vec{r}\times \vec{F} = \vec{M} = \frac{d\vec{L}}{dt}.
\end{align}

<br>

$\vec v || \vec{mv}$, поэтому $\vec{v}\times m\vec{v}=0$ (синус угла между сонаправленными векторами  = 0)
$\dfrac{d\vec p}{dt}=\vec F$  - второй закон Ньютона в импульсной форме

2. Для системы материальных точек:

<br>

\begin{align}
    & \frac{d\vec{L}}{dt} = \sum_i\frac{d\vec{L}_i}{dt} = \sum_i\left[\frac{d}{dt}\vec{r}_i\times m_i\vec{v}_i+\vec{r}_i\times m_i\frac{d\vec{v}_i}{dt}\right]=\nonumber\\
    & = \sum_i m_i\vec{r}_i\times\frac{d\vec{v}_i}{dt} = \sum_i \vec{r}_i \times \vec{F}_i = \sum_i \vec{M}_i. \qquad \blacksquare
\end{align}

<br>

$\dfrac{d}{dt}\vec{r}_i=\vec v_i || \vec{mv_i}\Rightarrow\dfrac{d}{dt}\vec{r}_i\times m_i\vec{v}_i=0$

Закон сохранения момента импульса: момент импульса системы материальных точек сохраняется, если на систему не действуют внешние силы, либо суммарный момент внешних сил равен нулю.

<br>

\begin{align}
    & \frac{d\vec{L}}{dt} = \sum_i \vec{r}_i\times\vec{F}_i = \sum_i\vec{r}_i\times\left(\vec{F}_i ^{ ext} +\sum_{k \neq i} \vec{F}_{ki}\right)=\nonumber\\
    & = \sum_i\vec{r}_i\times F_i^{ ext} + \sum_{i, k \neq i}\vec{r}_i\times\vec{F}_{ki} =\nonumber\\
    & = \sum_i\vec{r}_i\times\vec{F}_i^{ ext} + \sum_{i, k>i} \left(\vec{r}_i - \vec{r}_k \right) \times \vec{F}_{ki} =\nonumber\\
    & = \sum_i \vec{M}_{i}^{ ext}. \qquad \blacksquare
\end{align}

<br>

Как мы можем заметить, суммарный момент импульса системы меняется только при наличии момента внешних сил $\vec{M}_{i}^{ ext}$.

Пример: посчитаем предыдущее выражение для случая трёх сил, тогда:

<br>

\begin{align}
    & \underset{i, k\neq i}{\sum} \vec{r}_i\times\vec{F}_{ki} = \vec{r}_1 \times \vec{F}_{21} + \vec{r}_{1}\times\vec{F}_{31}+\vec{r}_2\times\vec{F}_{12}+\vec{r}_2\times\vec{F}_{32}+\vec{r}_3\times\vec{F}_{13}+\vec{r}_3\times\vec{F}_{23}=\nonumber
\end{align}

<br>

т.к. $\vec{F}_{21} = - \vec{F}_{12}$ и т.д., то

<br>

\begin{align}
    & = \left(\vec{r}_2 - \vec{r}_1\right)\times\vec{F}_{12}+\left(\vec{r}_3-\vec{r}_1\right)\times\vec{F}_{13}+\left(\vec{r}_3 - \vec{r}_2\right)\times\vec{F}_{23} = 0. \nonumber
\end{align}

<br>

Данный закон сохранения связан с особенностью пространства, называемой  _изотропностью_.
Интересен случай: $\vec{F} \sim \vec{r} \Rightarrow \vec{M} = 0 \Rightarrow \vec{L} = \text{const}.$
    Рассмотрим случай движения вокруг фиксированной оси: найдём проекцию момента импульса вдоль оси $OZ$:

<br>

\begin{align}
        L_z=(\vec{r} \times \vec{p})_z= ((\vec{r}_{\parallel} + \vec{r}_{\perp}) \times (\vec{p}_{\parallel} + \vec{p}_{\perp})_z= (\vec{r}_{\perp} \times \vec{p}_{\perp})_z=L_z.
    \end{align}

<br>

<br>

    \end{align}

момент инерции: $I=mR^2$.

      1& 0& 0\\
      0& 1& 0
      \end{Vmatrix} = \rho^2 m \dot{\varphi}\vec{e}_z,  
    \end{align}

<br>

   Где $\rho^2 m = L_z$.

<br>

\begin{align}

    \end{align}

Внимательно рассмотрев геометрическое построение, запишем условие на угол рассеяния, обозначив его $\theta$ (для простоты вычислений возьмём его ориентированным, чтобы не рассматривать несколько случаев):

Вернёмся к следствию из II закона Ньютона,

<br>

\begin{align}
        &\mu \dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = k \dfrac{q_1 q_2}{r^2} \dfrac{\vec{r}}{r} \\
        &\mu \dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} \cdot \mathrm{d}\vec{r} = k \dfrac{q_1 q_2}{r^2} \dfrac{\vec{r}}{r} \cdot \mathrm{d}\vec{r} \\
        &\text{где } \dfrac{\mathrm{d} \vec{v}}{\mathrm{d} t} \cdot \mathrm{d} \vec{r} = \dfrac{\mathrm{d} \vec{r}}{\mathrm{d} t}\cdot \mathrm{d} \vec{v} = \vec{v} \cdot \mathrm{d} \vec{v} \\
        &\mu \vec{v} \cdot \mathrm{d}\vec{v} = k \dfrac{q_1 q_2}{r^2} \dfrac{\vec{r}}{r} \cdot \mathrm{d}\vec{r} \\
        &\mu \int v \mathrm{d} v = k \int \dfrac{q_1 q_2}{r^2} \mathrm{d}r \\
        &\dfrac{\mu v^2}{2} = - k \dfrac{q_1 q_2}{r} + \text{const} \\
        &\dfrac{\mu v^2}{2} + k\dfrac{q_1 q_2}{r} = \text{const}. \label{eq:scattering_energy}
    \end{align}

В ходе вывода мы пользовались тем, что $\vec r\vec dr=rdr$, покажем это : ($\vec v\vec dv=vdv$ - аналогично )

<br>
\begin{align}
    & (\vec r,\vec r)=r^2 \\
    & (\vec r,\vec dr)+(\vec dr,\vec r)=2rdr \\
    & (\vec r,\vec dr)=rdr
\end{align}

\end{align}

        &= \int \limits^{+\infty}_{-\infty} F \cos \alpha \mathrm{d}t = k q_1 q_2 \int \limits^{+\infty}_{-\infty} \dfrac{dt}{\rho^2} \cos \left(\dfrac{1}{2} (\pi - \theta) - \varphi \right) = (*)
    \end{align}

<br>

\begin{align}
        &\dfrac{\mathrm{d}\phi}{\mathrm{d}t} = \dfrac{v_\infty b}{\rho^2} \\
        &\dfrac{\mathrm{d}t}{\rho^2} = \dfrac{\mathrm{d}\phi}{v_\infty b}  %НЕ МИНУС, А ТИРЕ! (:
    \end{align}

<br>

\begin{align}
        &(*) = k q_1 q_2 \int \limits^{\pi - \theta}_0 \dfrac{\mathrm{d}\varphi}{v_\infty b} \cos \left(\dfrac{1}{2} (\pi - \theta) - \varphi \right) \nonumber \\
        &= \dfrac{k q_1 q_2}{v_\infty b} \int \limits^{\pi - \theta}_0 \mathrm{d}\varphi\cos \left(\dfrac{1}{2} (\pi - \theta) - \varphi \right) \nonumber \\
        &= \dfrac{k q_1 q_2}{v_\infty b} \left. \left( -\sin \left( \dfrac{1}{2} (\pi - \theta) - \varphi \right) \right) \right|^{\pi - \theta}_0
    \end{align}

<br>

\begin{align}
        &= \dfrac{k q_1 q_2}{v_\infty b} \left[ \left( -\sin \left( \dfrac{1}{2} (\pi - \theta) - (\pi - \theta) \right) \right) - \left( -\sin \left( \dfrac{1}{2} (\pi - \theta) - 0 \right) \right) \right] \nonumber \\
        &= \dfrac{k q_1 q_2}{v_\infty b} \left[ \left( -\sin \left( - \dfrac{1}{2} (\pi - \theta) \right)\right) - \left( -\sin \left( \dfrac{1}{2} (\pi - \theta) \right) \right) \right] \nonumber \\
        &= \dfrac{k q_1 q_2}{v_\infty b} \left[ \sin \left(\dfrac{1}{2} (\pi - \theta) \right) +\sin \left( \dfrac{1}{2} (\pi - \theta)\right) \right] \nonumber  \\
        &= \dfrac{k q_1 q_2}{v_\infty b} \cdot 2 \sin \left( \dfrac{\pi}{2} - \dfrac{\theta}{2} \right) = 2 \dfrac{k q_1 q_2}{v_\infty b} \cos \dfrac{\theta}{2} \nonumber
\end{align}

<br>

\begin{align}
        |\Delta \vec{p}| = 2 \dfrac{k q_1 q_2}{v_\infty b} \cos \dfrac{\theta}{2}.
\end{align}

Получив два выражения для величины вектора $\Delta \vec{p}$, мы можем просто приравнять их:

Важной характеристикой рассеяния является величина d$\sigma$ - эффективное сечение рассеяния, определяемого как:
    
<br>

\begin{align}
    d\sigma\;=\;\frac{dN}{n},\nonumber
\end{align}
    
<br>

- где dN - число частиц, рассеиваемых в единицу времени на углы $\theta$ до $\theta \;+\; d\theta$, n - число частиц, пролетающих через единичную площадку поперечного сечения пучка в единицу времени.

По размерности это площадь.
Между углом рассеяния $\theta$ и прицельным расстоянием b есть связь!

<br>

<figure>
    <center>
    <img src="pics/8_diffusion_of_elements.png" width="700" align="center"/>
        <figcaption>Рис. 7.12:Рассеивание частиц.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    \Rightarrow dN\;=\;n \cdot 2\pi bdb, \nonumber\\
    \Rightarrow d\sigma \;=\; 2\pi bdb.
\end{align}
    
<br>

\begin{align}
     d\sigma \;=\; 2\pi b \cdot \bigg |\frac{db}{d\theta} \bigg | \cdot d\theta \;=\; b \cdot \bigg |\frac{db}{d\theta} \bigg |\cdot \frac{2\pi d\theta \sin{\theta}}{\sin{\theta}}\;=\; b \cdot \bigg |\frac{db}{d\theta} \bigg | \cdot \frac{d\Omega}{\sin{\theta}}\;=\; \frac{b}{\sin{\theta}} \cdot \bigg |\frac{db}{d\theta} \bigg | \cdot d\Omega,
\end{align}
    
<br>

- где $d\Omega$ - телесный угол.

Дифференциальное сечение рассеяния:
    
<br>

\begin{align}
    \frac{d\sigma}{d\Omega}\;=\;\frac{b}{\sin{\theta}} \cdot \bigg |\frac{db}{d\theta} \bigg |.
    \\ \nonumber
\end{align}
