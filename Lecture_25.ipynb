# Затухающие колебания. Вынужденные колебания
## Метод комплексных амплитуд

Мы понимаем, что для гармонических колебаний верно следующее соотношение:

<br>

\begin{align}
    x\sim\cos{\left(\omega t + \varphi_0\right)}
\end{align}

<br>

Рассмотрим следующее:

<br>

\begin{align}
    &e^{i\varphi} = \cos{\varphi} + i\sin{\varphi}, \;\; \varphi = \omega t + \varphi_0\nonumber\\
    &x = \mathrm{Re}\left[\underbrace{ae^{i\varphi}}_{z}\right],\\
    &y = \mathrm{Im}\left[z\right],
\end{align}
где $a$,$\;\varphi \in \mathbb{R},\;\;\varphi$ - фаза, $\varphi_0$ - начальная фаза.

<br>

Т. е. $x$ $\;-\;$ проекция комплексного числа $z$ на ось $\mathrm{Re}$.

<br>

\begin{align}
    &\dfrac{\mathrm{d}^2z}{\mathrm{d}t^2} + \omega^2z = 0\nonumber\\
    &\text{или}\nonumber\\
    &\dfrac{\mathrm{d}^2}{\mathrm{d}t^2} (x+iy) + \omega^2(x+iy) = 0 \implies \nonumber\\
    &\implies \underbrace{\dfrac{\mathrm{d}^2}{\mathrm{d}t^2}x+\omega^2x}_{=0} + \underbrace{i\left[\dfrac{\mathrm{d}^2}{\mathrm{d}t^2}y+\omega^2y\right]}_{=0} = 0.
\end{align}

<br>

Решение начального уравнения нужно искать в виде $e^{\lambda t}$

<br>

\begin{align}
    &\implies \dfrac{\mathrm{d}^2}{\mathrm{d}t^2} e^{\lambda t} + \omega^2e^{\lambda t} = 0\\
    &\implies \lambda^2+\omega^2=0\implies\lambda=\pm i\omega\\
    &\implies z = c_1e^{i\omega t} + c_2 e^{-i\omega t}.
\end{align}

<br>

\begin{align}
    &x = Re[Z] = Re[c_1 e^{i\omega t} + c_2e^{-i\omega t}] = \nonumber \\
    &=a_1\cos{\omega t} + a_2 \sin{\omega t} = \sqrt{a_1^2 + a_2^2} \left[\dfrac{a_1}{\sqrt{a_1^2+a_2^2}}\cos{\omega t} - \dfrac{(-a_2)}{\sqrt{a_1^2+a_2^2}} \sin{\omega t}\right]  \nonumber \\
    &= \sqrt{a_1^2 +a_2^2}\cos{\omega t + \varphi_0},\nonumber \\
    &tg\varphi_0 = \dfrac{a_1}{-a_2} = - \dfrac{a_1}{a_2}.
\end{align}

<br>

## Затухающие колебания.
Запишем второй закон Ньютона с силами сопротивления:

<br>

\begin{align}
    &m\dfrac{d^2}{dt^2}x = -kx -2\gamma\dot{x} m\nonumber \\
    &\implies \ddot{x} + \dfrac{k}{m}x + 2\gamma\dot{x} =0, \nonumber \\
    &\boxed{\ddot{x} + 2\gamma \dot{x} + \dfrac{k}{m}x =0}
\end{align}

<br>

Для дальнейшего рассмотрения необходимо разделять частоту собственных колебаний без затухания $\omega_0 = \sqrt{k/m}$ и частоту колебаний в присутствии затухания $\omega$. Можно решить методом комплексных амплитуд:

<br>

\begin{align}
    &\ddot{x} + 2\gamma\dot{x} + \omega_0^2x = 0,\nonumber \\
    &x = Re[z].
\end{align}

<br>

А можно решить так:

<br>

\begin{align}
    &x = e^{\lambda t}\nonumber \\
    &\implies \lambda^2 +2\gamma\lambda + \omega_0^2 = 0,\nonumber \\
    &\implies \lambda_{1,2} = \dfrac{-2\gamma \pm \sqrt{4\gamma^2 - 4\omega_0^2}}{2},\nonumber \\
    &\lambda_{1,2} = -\gamma \pm i \sqrt{\omega_0^2 -\gamma^2}.
\end{align}

<br>

Найдём решение $x$:

<br>

\begin{align}
    &x = a_1 e^{[-\gamma + i\omega]t} +  a_2e^{[-\gamma - i\omega]t},\nonumber \\
    &x = e^{-\gamma t}[a_1e^{i\omega t} + a_2e^{-i \omega t}],\nonumber \\
    &x = e^{-\gamma t} a\cos{(\omega t + \varphi_0)}\nonumber \\
    &x = ae^{-\gamma t}\cos{(\sqrt{\omega_0^2 - \gamma^2}t + \varphi_0)}.
\end{align}

<br>

Мы рассмотрели вариант, когда $\omega_0 > \gamma$
<br>
Если $\omega_0 = \gamma $, то $\lambda_1 = \lambda_2$, то есть корень кратности два , следовательно:

<br>

\begin{align}
    x = (a_1 + a_2t)e^{-\gamma t}.
\end{align}

<br>

Если же $\omega_0 < \gamma$, то 

<br>

\begin{align}
    x = e^{-\gamma t} \left[a_1 e^{-\sqrt{\gamma^2 - \omega_0^2}t} + a_2 e^{\sqrt{\gamma^2 - \omega_0^2}t}\right].
\end{align}

<br>

## Характеристики затухающих колебаний ($\omega_0 > \gamma $).
Амплитуда колебаний затухает:

<br>

\begin{align}
    a(t) = ae^{-\gamma t}.
\end{align}

<br>

Логарифмический декремент затухания:

<br>

\begin{align}
    \delta \equiv \lambda = \ln{\dfrac{a(t)}{a(t+ T)}} = \ln\dfrac{a e^{-\gamma t}}{a e^{-\gamma(t+T)}} = \ln{e^{\gamma T}} = \gamma T = \dfrac{2\pi \gamma}{\sqrt{\omega_0^2 - \gamma^2}}.
\end{align}

<br>

Добротность: 

<br>

\begin{align}
    Q = 2\pi \dfrac{E(t)}{\Delta E} = 2\pi \dfrac{E(t)}{E(t) - E(t+T)} = 2\pi \dfrac{e^{-2\gamma t}}{e^{-2\gamma t} - e^{-2\gamma(t+T)}} = 2\pi \dfrac{1}{1 - e^{-2\gamma T}},\nonumber \\
    E = \dfrac{1}{2} Ba^2(t) =\dfrac{1}{2}Ba^2e^{-2\gamma t}.
\end{align}

<br>

Если $\gamma << \omega_0$, то $T/\omega_0<<1 \implies \gamma T <<1$

<br>

\begin{align}
    Q = 2\pi \dfrac{1}{1-1 +2\gamma T} = \dfrac{\pi}{\gamma T} = \dfrac{\pi}{\lambda}.
\end{align}

<br>

## Вынужденные колебания
Рассмотрим вынужденные колебания: такие, при которых присутствует внешняя вынуждающая сила.

<br>

\begin{align}
    &m \dfrac{\mathrm{d}^2 x}{\mathrm{d}t^2} = - kx - 2 \gamma m \dfrac{\mathrm{d}x}{\mathrm{d}t} + F(t). \\
    &\text{Введём обозначение } f(t) = \dfrac{F(t)}{m}, \\
    &\boxed{\ddot{x} + 2 \gamma \dot{x} + \omega_0^2 x = f(t)}.
\end{align}

<br>

Получив дифференциальное уравнение, самым естественным чувством будет желание его решить. Рассмотрим частные случаи того, как может выглядеть решение в зависимости от вида $F(t)$.

### Постоянная вынуждающая сила
Примем $F(t) = F_0$, то есть $f(t) = f_0$.

<br>

\begin{align}
    &\ddot{x} + 2 \gamma \dot{x} + \omega_0^2 x = f_0.
\end{align}

<br>

Решение неоднородного уравнения - это сумма решения однородного и частного решения неоднородного: решение однородного мы знаем, попробуем угадать частное решение неоднородного.

<br>

\begin{align}
    &\text{Положим } x(t) = \tilde{x} = \text{const}, \\
    &0 + 0 + \omega_0^2 \tilde{x} = f_0, \\
    &\tilde{x} = \dfrac{f_0}{\omega_0^2}.
\end{align}

<br>

Получаем общий вид решения неоднородного:

<br>

\begin{align}
    &x(t) = A e^{- \gamma t} \cos (\omega t + \varphi_0) + \dfrac{f_0}{\omega_0^2}, \\
    &\text{где } \omega = \sqrt{\omega_0^2 - \gamma^2}.
\end{align}

<br>

## Гармоническая вынуждающая сила
Примем $F(t) = F_0 \cos (\Omega t)$, $f(t) = f_0 \cos (\Omega t)$.
<br>
Довольно естественно сказать, что $x(t)$ будет действительной частью некой комплексно-значной функцией $z(t)$ ― такой, что

<br>

\begin{align}
    &\ddot{z} + 2 \gamma \dot{z} + \omega_0^2 z = f_0 e^{i \Omega t}. \label{eq:complex_forced_osc} \\
    &\ddot{x} + 2 \gamma \dot{x} + \omega_0^2 x = f_0 \cos (\Omega t).
\end{align}

<br>

Раз правая часть пропорциональна $e^{i \Omega t}$, то и частное решение нужно искать формы
<br>
$\tilde{z}(t) = C e^{i \Omega t}$. Подставим это в уравнение 25.28.

<br>

$$\require{enclose}$$
\begin{align}
    &C (i \Omega)^2 \enclose{updiagonalstrike}{e^{i \Omega t}} + 2 \gamma C (i \Omega) \enclose{updiagonalstrike}{e^{i \Omega t}} + \omega_0^2 C \enclose{updiagonalstrike}{e^{i \Omega t}} = f_0 \enclose{updiagonalstrike}{e^{i \Omega t}}, \\
    &C( - \Omega^2 + 2 \gamma i \Omega + \omega_0^2) = f_0, \\
    &C = \dfrac{f_0}{\omega_0^2 - \Omega^2 + 2 \gamma i \Omega} = \dfrac{f_0 (\omega_0^2 - \Omega^2 - 2 \gamma i \Omega)}{(\omega_0^2 - \Omega^2)^2 + 4 \gamma^2 \Omega^2}.
\end{align}

<br>

В данном случае $C$ это комплексная амплитуда нашего колебания ― запишем её в тригонометрической форме, разделив, собственно, амплитуду и смещение по фазе.

<br>

\begin{align}
    &C = |C| e^{i\Psi}, \\
    &tg \Psi = \dfrac{- 2 \gamma \Omega}{\omega_0^2 - \Omega^2} = \dfrac{2 \gamma \Omega}{\Omega^2 - \omega_0^2}, \\
    &|C| = \dfrac{f_0}{\sqrt{(\omega_0^2 - \Omega^2)^2 + 4 \gamma^2 \Omega^2}}. \\
    &\tilde{x}(t) = \mathrm{Re} \left( |C| e^{i ( \Omega t + \Psi)}\right). \\
    &\tilde{x}(t) = |C| \cos (\Omega t + \Psi). \\
    &x(t) = A e^{- \gamma t} \cos (\omega t + \varphi_0) + |C| \cos (\Omega t + \Psi), \\
    &\text{где } \omega = \sqrt{\omega_0^2 - \gamma^2}.
\end{align}

<br>

Заметим, что через какое-то время первое слагаемое ― решение однородного уравнения ― затухнет и полностью исчезнет, так что в конечном счёте останется лишь колебание, вызванное вынуждающей силой. \\
Амплитуда этого второго слагаемого зависит от частоты прикладываемой силы. Есть такие частоты, на которых амплитуда резко возрастает ― это явление называется резонансом. Проанализируем $C(\Omega)$ и найдём частоты, при которых достигается максимум.

<br>

1. $\Omega \approx 0$.
    \begin{align}
        &|C| = \dfrac{f_0}{\omega_0^2}, \\
        &\Psi < 0.
    \end{align}
2. $\Omega \gg \omega_0$.
    \begin{align}
        &|C| = \dfrac{f_0}{\Omega \sqrt{\Omega^2 + 4 \gamma^2}} \approx 0, \\
        &\Psi \approx 0.
    \end{align}
3. Достигается максимум: $\dfrac{\mathrm{d}|C|}{\mathrm{d}\Omega} = 0$.
    \begin{align}
        &\dfrac{f_0 \cdot \Bigl(2 (\omega_0^2 - \Omega^2) \cdot 2 \Omega \cdot (-1) + 8 \gamma^2 \Omega \Bigr)}{\Bigl((\omega_0^2 - \Omega^2)^2 + 4 \gamma^2 \Omega^2\Bigr)^{\dfrac{3}{2}}} = 0, \\
        &-\cancel{4 \Omega} (\omega_0^2 - \Omega^2) + \cancelto{2}{8} \gamma^2 \cancel{\Omega} = 0, \\
        &2\gamma^2 = \omega_0^2 - \Omega^2, \\
        &\Omega^2 = \omega_0^2 - 2 \gamma^2, \\
        &\boxed{\Omega = \sqrt{\omega_0^2 - 2 \gamma^2}}.
    \end{align}
