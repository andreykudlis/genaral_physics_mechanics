# Кинематика материальной точки

## Основные определения
**Определение.** Материальная точка - модель физического объекта, размерами которого в условиях рассматриваемой задачи можно пренебречь.
В кинематике изучаются законы движения, без анализа причин возникновения самого движения. Для задания положения материальной точки нам нужна система отсчета и минимальный набор координат (они называются обобщенными координатами).
Положение можно определять по-разному. Это может быть как радиус-вектор, так и значение пути, пройденного по фиксированной траектории.

Пусть положение задаётся радиус-вектором: $\vec{r}(t)=(x(t),\; y(t),\; z(t))$.

Вектор $\Delta \vec{r}$ называется вектором перемещения.

Понятно, что $|\Delta \vec{r}|\leqslant S$.

## Скорость
**Определение.** Скорость - производная по времени от радиус вектора.
\begin{align}
 \vec{v} = \dfrac{d\vec{r}}{dt} = \vec{e_x} \cdot \dfrac{dx}{dt} + \vec{e_y} \cdot \dfrac{dy}{dt} + \vec{e_z} \cdot \dfrac{dz}{dt} = \vec{e_x} \cdot v_x + \vec{e_y} \cdot v_y + \vec{e_z} \cdot v_z
\end{align}
Средняя скорость:
\begin{align}
 \langle\vec{v}\rangle = \dfrac{\vec{r}(t_2)-\vec{r}(t_1)}{t_2-t_1}.
\end{align}
Если же абсолютная величина:
\begin{align}
 \langle v \rangle = \dfrac{S_{all}}{t_2-t_1}.
\end{align}
## Ускорение
**Определение.** Ускорение - производная по времени от скорости, записывается следующим образом:
\begin{align}
 \vec{a} = \dfrac{d\vec{v}}{dt} = \dfrac{d^2\vec{r}}{dt^2} = \vec{e_x} \cdot \dfrac{dv_x}{dt} + \vec{e_y} \cdot \dfrac{dv_y}{dt} + \vec{e_z} \cdot \dfrac{dv_z}{dt} = \vec{e_x} \cdot a_x + \vec{e_y} \cdot a_y + \vec{e_z} \cdot a_z.
\end{align}

<ins> Замечание. </ins>

Скорость всегда направлена по касательной к траектории, так как $\vec{v} \sim d\vec{r}$. 

## Естественная параметризация
Пусть $\vec{v} = v \cdot \vec{\tau}$, где $\tau = d\vec{r}/dS$, где $|d\vec{r}|=dS$, а $S$ - пройденный путь.

Получим ускорение в общем случае криволинейного движения: 
\begin{align}
    \vec{W} = \dfrac{d}{dt}\vec{v} = \vec{\tau} \cdot \dfrac{dv}{dt} + v \cdot \dfrac{d\vec{\tau}}{dt} = a_\tau\vec{\tau} + v \cdot \vec{n} \cdot \dfrac{dS}{Rdt} =  a_\tau\vec{\tau} + \vec{n}\cdot \dfrac{v^2}{R}
\end{align}
Где $\vec{n}$ - вектор нормали. $a_\tau$ отвечает за изменение абсолютного значения скорости.
\begin{align}
   d\vec{\tau} =  \vec{n} \cdot |\tau| \cdot d \varphi = \vec{n} \cdot \dfrac{dS}{R}
\end{align}

## Циклоида

При помощи циклоиды можно понять, какую траекторию описывает точка на ободе колеса относительно поверхности.

Рассматривая равномерное качение, скорость центра колеса 

\begin{align*}
    v_c = \omega R
\end{align*}

Введём радиус-вектор произвольной точки на ободе колеса: $\vec{r}$

\begin{align*}
    \vec{r} = \vec{r}_c + \vec{r}_A
\end{align*}


Для радиус-векторов различных точек и соответствующих скоростей мы можем записать ряд следующих соотношений:

\begin{align}
    &\vec{r}_c = \begin{pmatrix} x_0 + v_c t \\ y_0 \end{pmatrix} = \begin{pmatrix} \omega R t \\ R \end{pmatrix} \implies \vec{v}_c = \begin{pmatrix} \omega R \\ 0 \end{pmatrix}, \\
    %
    &\vec{r}_A = \begin{pmatrix} R \sin (\omega t) \\ R \cos (\omega t) \end{pmatrix} = R \begin{pmatrix} \sin (\omega t) \\ \cos (\omega t) \end{pmatrix} \implies \vec{v}_A = \begin{pmatrix} \phantom{-} R \omega \cos (\omega t) \\ - R \omega \sin (\omega t) \end{pmatrix},\\
    %
    &\vec{r} = \begin{pmatrix} R\Bigl( \omega t + \sin (\omega t) \Bigr) \\ R\Bigl( \phantom{\omega} 1 + \cos (\omega t) \Bigr) \end{pmatrix} \implies \vec{v} = \begin{pmatrix} \omega R \Bigl( 1 + \cos (\omega t) \Bigr) \\ - \omega R \sin (\omega t) \end{pmatrix}.
\end{align}

Помимо этого, мы можем выразить касательный вектор $\vec{\tau}$, описывающий направление скорости точки на ободе относительно центра колеса, соответствующую ему нормаль, а также ускорение материальной точки $\vec{a}$:

\begin{align}
    \vec{\tau} &= \begin{pmatrix} \phantom{-} \cos (\omega t) \\ - \sin (\omega t) \end{pmatrix}, \quad 
    \vec{n} = \begin{pmatrix} - \sin (\omega t) \\ - \cos (\omega t) \end{pmatrix},\quad
    \vec{a} = R \omega^2 \vec{n}.
\end{align}
Свойства:
* Время соскальзывания без трения не зависит от начального положения.
* Кривая скорейшего спуска (Брахистохрона).
* Маятник Гюйгенса.

## Радиус кривизны

Найдём кривизну известной траектории $y(x)$ в каждый момент времени, предполагая, что она достаточно гладкая:

<figure>
    <center>
    <img src="pics/17.png" width="320" align="center"/>
    </center>
</figure>
\begin{align*}
    \mathrm{d}S &= \sqrt{\mathrm{d}x^2 + \mathrm{d}y^2} = \\
    &= \mathrm{d}x \sqrt{1 + \left( \dfrac{\mathrm{d}y}{\mathrm{d}x} \right)^2} \\
    \mathrm{d}S &= R \mathrm{d}\alpha \\
    y'_x &= tg \varphi = \dfrac{\mathrm{d}y}{\mathrm{d}x} \\
    \mathrm{d} \varphi &= \mathrm{d} \alpha \\
    \dfrac{\mathrm{d}}{\mathrm{d} x} (tg \varphi) &= \dfrac{1}{\cos^2 \varphi} \cdot \dfrac{\mathrm{d} \varphi}{\mathrm{d} x} = (1 + tg^2 \varphi) \cdot \dfrac{\mathrm{d} \varphi}{\mathrm{d} x} \\
    \dfrac{\dfrac{\mathrm{d}}{\mathrm{d} x} (tg \varphi)}{(1 + tg^2 \varphi)}\mathrm{d}x &= \mathrm{d}\varphi \\ 
    \dfrac{\dfrac{\mathrm{d}}{\mathrm{d} x} y'_x}{\Bigl( 1 + (y'_x)^2 \Bigr)}\mathrm{d}x &= \mathrm{d}\varphi \\
    \mathrm{d}x \sqrt{1 + (y'_x)^2} &= \mathrm{d}S
\end{align*}
\begin{align*}
    \implies R = \dfrac{\mathrm{d} S}{\mathrm{d} \varphi} &=  \dfrac{\mathrm{d}x \sqrt{1 + (y'_x)^2}\Bigl( 1 + (y'_x)^2 \Bigr)}{\mathrm{d} x \left( \dfrac{\mathrm{d}}{\mathrm{d} x} y'_x \right)} = \dfrac{\Bigl( 1 + (y'_x)^2 \Bigr)^{\dfrac{3}{2}}}{y''_x} \\
    R &= \dfrac{\Bigl( 1 + (y'_x)^2 \Bigr)^{\dfrac{3}{2}}}{y''_x} \text{ ― радиус кривизны траектории}
\end{align*}


**Пример.** Баллистическое движение. 
Воспользуемся полученными знаниями дифференциального исчисления и решений для полёта материальной точки.
\begin{align*}
    \text{Решение было: } x &= v_{0_x} t = v_0 t \cos \varphi \\
    y &= v_{0_y} t - \dfrac{gt^2}{2} = v_0 t \sin \varphi - \dfrac{gt^2}{2} \\
    R &= \dfrac{\Bigl( \sqrt{1 + (y'_x)^2} \Bigr)^3}{y''_x} \text{ ― то есть, необходимо найти } y'_x, y''_x 
\end{align*}
\begin{align*}
    y'_x &= \dfrac{\mathrm{d}y}{\mathrm{d}x} = \dfrac{\mathrm{d}y}{\mathrm{d}t} \cdot \dfrac{\mathrm{d}t}{\mathrm{d}x} = \dfrac{\mathrm{d}y}{\mathrm{d}t} \cdot \dfrac{1}{\dfrac{\mathrm{d}x}{\mathrm{d}t}} = \dfrac{v_0 \sin \varphi - gt}{v_0 \cos \varphi} \\
    y''_x &= \dfrac{\mathrm{d}}{\mathrm{d}x} y'_x = \dfrac{\mathrm{d}t}{\mathrm{d}x} \cdot \dfrac{\mathrm{d}}{\mathrm{d}t} y'_x = \dfrac{1}{\dfrac{\mathrm{d}x}{\mathrm{d}t}} \cdot \dfrac{\mathrm{d}}{\mathrm{d}t} \left( \dfrac{v_0 \sin \varphi - gt}{v_0 \cos \varphi} \right) = \\
    &= \dfrac{1}{v_0 \cos \varphi} \cdot \dfrac{\mathrm{d}}{\mathrm{d}t} \left( \dfrac{v_0 \sin \varphi - gt}{v_0 \cos \varphi} \right) = \dfrac{1}{v_0 \cos \varphi} \cdot \left( - \dfrac{g}{v_0 \cos \varphi} \right) = - \dfrac{g}{v_0^2 \cos^2 \varphi} \\
    \implies \Bigl( (1 + (y'_x)^2 \Bigr)^{\dfrac{3}{2}} &= \left( 1 + \dfrac{(v_0 \sin \varphi - gt)^2}{v_0^2 \cos^2 \varphi} \right)^{\dfrac{3}{2}} = \dfrac{\left( v_0^2 - 2 v_0 g t \sin \varphi + g^2t^2 \right)^{\dfrac{3}{2}}}{v_0^3 \cos^3 \varphi} \\
    y''_x &= - \dfrac{g}{v_0^2 \cos^2 \varphi} \\
    \implies R &= \dfrac{\left( v_0^2 - 2 v_0 g t \sin \varphi + g^2t^2 \right)^{\dfrac{3}{2}}}{g v_0 \cos \varphi} = \dfrac{\left( v_0^2 - 2 v_0 g t \sin \varphi + g^2t^2 \right)^{\dfrac{3}{2}}}{g v_x} =\dfrac{v^3}{g v_x} \\
    \text{где } v& \text{ ― полная скорость, } v = \sqrt{v_x^2 + v_y^2} = \\
    &= \sqrt{v_{0_x}^2 + v_{0_y}^2 - 2 v_{0_y} gt + g^2 t^2} = \sqrt{v_0^2 - 2 v_{0_y} gt + g^2 t^2}
\end{align*}

$$R = \dfrac{v^3}{g v_x}$$
