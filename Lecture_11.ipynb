# Космические скорости

## Первая космическая скорость

Рассмотрим движение некого тела вокруг некого массивного объекта (например, планеты). Будем пренебрегать сопротивлением среды и гравитационным взаимодействием с другими телами: то есть, из сил рассматривать лишь взаимодействие тела с массивным объектом.

Полная механическая энергия тела тогда составит

\begin{align}
    &E = \dfrac{mv^2}{2} - \gamma \dfrac{mM}{r}.
\end{align}

<br>

Давайте рассмотрим вращение тела вокруг нашего массивного объекта: для этого удобно перейти в полярную систему координат и положить $\varphi \neq 0$ и $L > 0$. Понятно, что случаев вращения существует огромное множество ― интересно посмотреть на экстремальный случай, когда полная механическая энергия (а, следовательно, и скорость тела) минимальна.

<br>

\begin{align}
    &E_\text{min} = \dfrac{m \dot{\rho}^2}{2} + \dfrac{m}{2} \rho^2 \dot{\varphi}^2 - \gamma \dfrac{mM}{\rho} \nonumber \\
    &\phantom{E_\text{min}} = \dfrac{m \dot{\rho}^2}{2} + \underbrace{\dfrac{L^2}{2 m \rho^2} - \gamma \dfrac{mM}{\rho}}_{U_\text{eff}}.
\end{align}

<br>

Если кто-то не понимает, почему энергия записана именно таким образом, вернитесь к лекции 9, там показано разложение скорости по базисным векторам в полярной системе координат.

Мы хотим вращение!

<br>

\begin{align}
    \Rightarrow \dot{\varphi} \neq 0 \Rightarrow L\;>\;0 \Rightarrow E_{min},\; \text{когда}\; \dot{\varphi} \; \text{и} \; U_{eff} \; \text{минимальна} \Rightarrow \nonumber
\end{align}

<br>

\begin{align}
    \dfrac{-2L^2}{2m\rho_0^3}\;+\;\dfrac{\gamma m M}{\rho_0^2}\;=\;0 \Rightarrow \rho_0 \;=\; \dfrac{L^2}{\gamma m^2 M},
\end{align}

<br>

$\Rightarrow$ т.к. по окружности $L\;=\;m\rho_0 v_{\varphi}$, т.е. $L\;=\;m\rho v$

<br>

\begin{align}
\Rightarrow \rho_0\;=\;\dfrac{m^2 \rho_0^2 v^2}{\gamma m^2 M} \Rightarrow v\;=\;\sqrt{\gamma \dfrac{M}{\rho_0}}.
\end{align}

<br>

Вспомним, что мы на поверхности Земли:

<br>

\begin{align}
    \Rightarrow v_1\;=\;\sqrt{\gamma \dfrac{M_{\oplus}}{R_{\oplus}}} \sim 8 \text{км/с}.
    \\ \nonumber
\end{align}

## Вторая космическая скорость

Минимальная скорость, которую необходимо задать телу, чтобы оно покинуло небесное тело:

<br>

\begin{align}
    &E = \dfrac{mv^2}{2} - \gamma \dfrac{mM}{r}\;=\;0.
\end{align}

<br>

\begin{align}
    \Rightarrow v_2\;=\;\sqrt{2\gamma \dfrac{M_{\oplus}}{R_{\oplus}}}\;=\;\sqrt{2}v_1 \approx 11.2 \text{км/с}.
\end{align}

<br>

Если $E\;>\;0$ $\Rightarrow$ движение по гиперболе $\Rightarrow$ превосходит 2-ую космическую скорость.

## Третья космическая скорость

Скорость, которую необходимо задать телу относительно Земли, чтобы оно (тело) покинуло пределы Солнечной системы. Для простоты расчётов, разобьём наше движение на две части: удаление из гравитационного поля Земли, а затем уже удаление из гравитационного поля Солнца. Расстояние от Земли до Солнца (примерно 150 млн км) намного больше радиуса Земли (примерно 6400 км), поэтому когда тело уже удалится от Земли настолько, что практически перестанет с ней взаимодействовать, расстояние от него до Солнца принципиально не изменится, поэтому наши приближённые рассуждения будут верны с высокой степенью точности.


Рассмотрим процесс удаления из гравитационного поля Земли :

<br>

\begin{align}
   (*) \dfrac{m\vec{v}^2}{2} - \gamma \dfrac{mM}{r}\;=\;\dfrac{m\vec{v}_{\infty}^2}{2}.
\end{align}

<br>

$\vec{v} \;-\;$ скорость (стартовая) относительно Земли.

$\vec{v}_{\infty} \;-\;$ скорость тела относительно Земли, когда тело уже на большом расстоянии от Земли.

Скорость, когда отлетели от Земли: 

<br>

\begin{align}
    \vec{v}_{{\text{отн. Солнца}\;\infty}} \;=\; \vec{V}_{з} \;+\; \vec{v}_{\infty}
\end{align}

<br>

$\vec{V}_{з} \;-\; $ скорость кругового движения Земли вокруг Солнца.

<br>

\begin{align}
    v_{\text{отн. Солнца}\; \infty}^2 \;=\; V_{з}^2 \;+\; v_{\infty}^2 \;+\;2V_{з}\cdot v_{\infty} \cos{\theta}.
\end{align}

<br>

$\theta \;-\;$ угол между скоростью орбитального движения Земли и скоростью, с которой тело выходит из Земного тяготения.

<br>

Предположим, что Земля вращается вокруг Солнца по круговой орбите, тогда будет верно, что она движется с первой космической скоростью для Солнца. Чтобы тело могло покинуть гравитационное поле Солнца, оно должно двигаться со второй космической скоростью, а для 1 и 2 космических скоростей верно следующее : $V_2=\sqrt{2}V_1$

Чему должна быть равна $\vec{v}_{\text{отн. солнца.} \infty}$:

<br>

\begin{align}
    \vec{v}_{\text{отн. солнца.} \infty} \;=\; \sqrt{2}V_{\text{з}}\; \text{(2-ая космическая с орбиты Земли отн. Солнца)}. \nonumber
\end{align}

<br>

\begin{align}
    &\Rightarrow 2V_{\text{з}}^2 \;=\; V_{\text{з}}^2 \;+\; V_{\infty}^2 \;+\; 2V_{\text{з}}{v}_{\infty} \cos{\theta}, \nonumber \\
    &\Rightarrow v_{\infty}\;=\;-\; \dfrac{2V_{з} \cos{\theta + \sqrt{4 V_{з}^2 \cos{\theta}^2\;+\; 4V_{з}^4}}}{2}, \nonumber \\
    &\Rightarrow v_{\infty}\;=\; V_{з} \cos{\theta} \cdot \left( \dfrac{\sqrt{1\;+\;\cos(\theta)^2}}{\cos{\theta}}\;-\;1 \right)
\end{align}

<br>

$\Rightarrow$ из (*) $\vec{v}^2 \;=\; 2\vec{V}_1^2\;+\;\vec{v}_{\infty}^2 $.

<br>

\begin{align}
    &\Rightarrow \vec{v}^2 \;=\; 2\vec{V}_1^2\;+\;V_{з} \cdot {\left( \sqrt{1\;+\;\cos(\theta)^2}\;-\; \cos{\theta} \right)}^2. \qquad \blacksquare
    \\ \nonumber
\end{align}
