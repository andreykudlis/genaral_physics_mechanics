# СТО. Часть 2

В нерелятивистской теории нам были хорошо известны различные сохраняющиеся величины (которые при переходе из одной ИСО в другую не меняются), такие величины называются инвариантами, например, время, длина. Однако такие величины уже не будут инвариантами относительно преобразований Лоренца.

Знакомимся с первым инвариантом: $\Delta S$ - интервал

Пусть $\Delta \tau \;=\; c \Delta t$.

<br>

\begin{align}
    &\Delta S \;=\; \sqrt{(c \Delta t)^2\;-\; (\Delta x)^2}\;=\\ 
    &=\;\gamma \sqrt{(\Delta \tau^{'} \;+\; \beta \Delta x^{'})^2\;-\;(\beta \Delta x^{'}\;+\; \beta \Delta \tau^{'})^2} \nonumber\\
    &=\sqrt{(\Delta \tau^{'})^2\;-\;(\Delta x^{'})^2}\;=\; \Delta S'.
\end{align}

<br>

где $\Delta x'\;=\;\gamma(\Delta x \;-\; \beta \Delta \tau)$,

$\Delta x\;=\;\gamma(\Delta x' \;+\; \beta \Delta \tau)$, 

$\Delta \tau'\;=\;\gamma(\Delta \tau \;-\; \beta \Delta x)$, 

$\Delta \tau\;=\;\gamma(\Delta \tau' \;+\; \beta \Delta x')$.

**Обсудим причинность.**

Рассмотрим 2 события с координатами (1)$(x_1; \tau_1)$ и (2)$(x_2; \tau_2)$ в одной из ИСО.

Пусть (1) есть причина (2), т.е. $x_2\;-\;x_1\;<\;c(t_2\;-\;t_1)$.

Пусть скорость взаимодействия будет $v_0$, тогда $x_2\;-\;x_1\;=\;v_0(t_2\;-\;t_1)\;<\;c(t_2\;-\;t_1)$.

Сохраняется ли его причинность? (или время второго события (его наступления) случается после времени первого)

<br>

\begin{align}
    &\Delta \tau \;=\; c(t_2'\;-\;t_1')\;=\; \gamma (\Delta \tau \;-\; \beta \Delta x) \;=\; \gamma \left[ c \Delta t \;-\; \dfrac{V}{c}v_0 \Delta t \right]\;= \\
    &=\; c \gamma \Delta t \left[ 1 \;-\; \dfrac{V v_0}{c^2} \right]\;>0.
\end{align}

<br>

## Парадокс шеста и сарая
Вопрос об одновременном нахождении обоих концов стержня в сарае. 

Парадокс: пусть шест будет более длинным, чем сарай и он летит со скоростью $v$. 

Собственная длина шеста: $l_0$

Длина сарая: $L$

Вопрос нашего видения (из нашей ИСО)
В этот момент одновременно можем закрыть двери, то есть он поместился в сарай. 

Тогда говорим: СТОП! В системе отсчёта стержня сарай кажется ещё короче. И как же мы закроем двери? Дело в том, что одновременное закрытие дверей в СО сарая не будет одновременным в СО стержня!

В системе сарая:

Событие 1: влетает передним концом. 

<br>

\begin{align}
    (ct_1, x_1)=(0, 0). 
\end{align}

<br>


Событие 2: влетает задний конец. 

<br>

\begin{align}
    &\sqsupset L = l = l_0 \sqrt{1 - \dfrac{v^2}{c^2}}.
    \\
    &(ct_2, x_2)=(\dfrac{l}{v} \cdot c, 0)=(\dfrac{l_0 \sqrt{1 - \dfrac{v^2}{c^2}}}{v} \cdot c, 0) = (L \cdot \dfrac{c}{v}, 0). 
\end{align}

<br>


Событие 3: вылетает передний конец (одновременно с событием 2). 

<br>

\begin{align}
    (ct_3, x_3) = (\dfrac{l}{v}\cdot c, 0+l) = (L \dfrac{c}{v}, L). 
\end{align}

<br>


Событие 4: вылетает задний конец. 

<br>

\begin{align}
    (ct_4, x_4) = (2 \cdot \dfrac{l}{v} \cdot c, l). 
\end{align}

<br>


А с точки зрения шеста:

<br>

\begin{align}
    &(ct'_1, x'_1) = (0,0)
    \\
    &\boxed{x' = \gamma (x-vt)}
    \\
    &\boxed{t' = \gamma (t - \dfrac{v}{c^2} \cdot x)}
    \\
    &(ct'_2 , x'_2) = (\gamma \dfrac{Lc}{v}, -\gamma \cdot L) = (l_0 \dfrac{c}{v}, -l_0) \text{ -- логично }
    \\
    &(ct'_3, x'_3) = (c \gamma (\dfrac{l}{v} - \dfrac{v}{c^2}l), \gamma (l - v \cdot \dfrac{l}{v})) = (\dfrac{c}{v}l_0(1 - \dfrac{v^2}{c^2}), 0), 
    \\
    &\text{то есть не одновременно с событием 2, а раньше.}
\end{align}

<br>

Событие 4:

<br>

$$\require{enclose}$$
\begin{align}
    &\left(ct_4', x_4' \right) = \left(c\gamma\left(2\dfrac{l}{v}-\dfrac{v}{c^2}l\right), \gamma\left(l-\enclose{updiagonalstrike}{v}2\dfrac{l}{\enclose{updiagonalstrike}{v}}\right)\right) = \nonumber\\
    &= \left(c\dfrac{l_0}{v}\left(2-\dfrac{v^2}{c^2}\right), l_0(-1)\right) = \left(2c\dfrac{l_0}{v}\left(1-\dfrac{v^2}{2c^2}\right), -l_0\right). \; \; \blacksquare
\end{align}

## Преобразование скорости
Пусть у нас есть материальная точка, которая движется со скоростью $v_x$ в системе $K$ и $v_x'$ в системе $K'$ (обе системы $\;-\;$ инерциальные). В обычном случае (не СТО):

<br>

\begin{align}
    \boxed{v_x = V + v_x'}.
\end{align}

<br>

А как же в СТО?

<br>

\begin{align}
    &v_x' = \dfrac{\mathrm{d}x'}{\mathrm{d}t} = \dfrac{\mathrm{d}x - V\mathrm{d}t}{\mathrm{d}t-\dfrac{V}{c^2}\mathrm{d}x} = \dfrac{\dfrac{\mathrm{d}x}{\mathrm{d}t}-V}{1-\dfrac{V}{c^2}\dfrac{\mathrm{d}x}{\mathrm{d}t}} \implies \nonumber\\
    &\implies \boxed{v_x' = \dfrac{v_x - V}{1 - \dfrac{V}{c^2}v_x}}.
\end{align}

<br>

Значения $\mathrm{d}x'$ и $\mathrm{d}t'$ получили, взяв дифференциал следующих знакомых выражений:

<br>

\begin{align*}
    &x' = \gamma(x-Vt),
    &t' = \gamma\left(t - \dfrac{V}{c^2}x\right).
\end{align*}

<br>

А что будет с перпендикулярным направлением (проекцией на ось $y$ или $z$)? Вспомним, что $\mathrm{d}y|z'$ = $\mathrm{d}y|z$

<br>

\begin{align}
    &v_{y|z}' = \dfrac{\mathrm{d}{y|z}'}{\mathrm{d}t'} = \dfrac{\mathrm{d}{y|z}}{\gamma\left(\mathrm{d}t - \dfrac{V}{c^2}\mathrm{d}x\right)} = \dfrac{v_{y|z}}{\gamma\left(1 - \dfrac{V}{c^2}v_x\right)} \implies \nonumber\\
    \implies \boxed{v_{y|z} = \dfrac{\sqrt{1 - \dfrac{V^2}{c^2}}v_{y|z}}{1-\dfrac{V}{c^2}v_x}}.
\end{align}
