### Научный подход

Многие из пришедших на курс хотели бы связать свою жизнь с научной деятельностью. По этой причине сперва мы рассмотрим особенности научного подхода, вдруг после прослушивания данной части кто-то решит, что это точно не для него. Приведем некоторый примерный список, он может отличаться от своих аналогов, которые вы можете найти в интернете. Это нормально, ибо в рамках того или иного научного исследования некоторые шаги могут просто не понадобиться. 

Все начинается с вашего внимания(**наблюдения**), с которым вы отмечаете те или иные особенности поведения элементов окружающего вас мира. Банально, это могут быть прочитанные вами научные публикации, из прочтения которых вы могли заметить некую особенность/закономерность. Или же, например, у вас есть загородный дом, на крыше которого в определенном месте в зимний период постоянно образуется большой слой наледи. После этого у вас может возникнуть **вопрос**, который связан с тем или иным аспектом проделанного вами наблюдения. В нашем случае это: “А почему именно там образуется наледь?”. Затем вы начинаете **сбор данных**, проходите курсы по термодинамике и инженерному делу, т.е. комбинируете теоретический и эмпирический пути познания. У вас уже хватает опыта и знаний, чтобы что-то предположить. И вы это делаете, вы **предлагаете гипотезу**, почему так происходит. В нашем случае это может быть как выход горячего воздуха(если проблемный участок находится недалеко от входа), так и плохая вентиляция кровли. Пусть это будет первый вариант.  Помните, что гипотеза не обязательно должна быть правильной, в этом вам лишь предстоит убедиться в дальнейшем. Теперь же нам необходимо сделать **предсказание**, которое для нас могло бы звучать так: “А если я больше не буду выходить через эту дверь, будет ли образовываться наледь?”. Затем идет **проверка вашего предсказания**. Всю следующую зиму вы перестаете пользоваться данным выходом, наблюдая за появлением наледи. Вы проводите **анализ полученных данных**. Например, если наледь все-таки образовалась, является ли толщина слоя такой же как и раньше. Если в итоге наледь не образовалась, то значит ваша гипотеза оказалась правильной, и вы делаете **соответствующий вывод**. После чего, вам необходимо **проверить вашу гипотезу в аналогичных условиях**, но в другой лаборатории, т.е. мы говорим про **воспроизводимость**. Таким образом вы начинаете ездить по всему коттеджному поселку в поисках наледи у входа в дом. Убедившись в своей правоте, вы **публикуете** ваши результаты. Более того, ваши данные должны быть **общедоступны**(речь о деталях вашего эксперимента, ну или теории, которую вы построили), чтобы альтернативная группа исследователей, т.е. в нашем случае это какой-то другой ненормальный садовод, мог бы провести такую же серию экспериментов, т.е. осуществить **рецензию** вашего исследования(воспроизводимость другой группой).
  
Стоит сказать, что вы должны трезво относиться к существующим законами физики и быть готовыми к тому, чтобы переосмыслить теорию, если при ее очередной проверке каким-то хитрым экспериментом она потерпит неудачу.  


### Что будем изучать на курсе?
Физику. Для чего нам нужна физика? По всей видимости, наиболее правильным был бы следующий ответ: физика нам нужна для поиска и объяснения законов природы. С течением времени наука приходит к все более глубокому пониманию наблюдаемых в природе явлений, посредством создания максимально общих теорий, способных описать на первый взгляд абсолютно несвязанные процессы. Одним из наиболее ярких примеров здесь может служить квантовая теория, в рамках которой было объяснено корпускулярно-волновое поведение объектов микромира: в одних условиях мы наблюдаем поведение электрона или фотона, свойственное волне, тогда как в других – рассматриваемый объект ведет себя как частица. Формульно данное утверждение было подкреплено соотношением $\lambda=h/p$, предложенное Луи де Бройлем в 1923 году. Оно демонстрирует фундаментальную связь между волновыми и корпускулярными характеристиками объекта. Детальное знакомство с квантовой теорией у вас произойдет на третьем курсе.

Наряду с появившейся в начале 20-ого века квантовой теории, которая полностью изменила представления о физике атомных масштабов, релятивистская теория, сформулированная Альбертом Эйнштейном в 1905 году, продемонстрировала несостоятельность классической физики при описании явлений, где наблюдаемые скорости достигают световых значений.

Однако, когда масштабы анализируемых явлений не слишком малы(много больше размеров атома или же размеры системы достаточно велики), а характерные скорости имеют значения много меньше скорости света($<<3\cdot 10^8$ м/c), классическая физика полностью справляется с описанием соответствующих процессов. В данном курсе мы будем заниматься вопросами именно **классической механики**, без которой немыслимо полноценное образование любого физика или инженера. Без ее качественного освоения понять более сложные концепты будет значительно сложнее. По этой причине не стоит пренебрегать столь важным разделом.

### Роль математики
Современная физика появляется в тот момент, когда ей наконец-то удается отделиться(или отделаться) от философии. Некоторые из историков считают, что современная физика появилась в 1543 году, когда Коперник опубликовал свой знаменитый труд “О вращении небесных тел”. Затем появились работы Гюйгенса, Галилея, Кеплера и многих других. Однако наиболее фундаментальный вклад был сделан Ньютоном, опубликовавшим в 1687 году свои “Начала”, где им были сформулированы основные законы движения(законы Ньютона) и законы тяготения. Отличительной чертой данной книги стало то, что многие вещи доказывались аналитическим способом. Для этого Ньютон изобрел дифференциальное исчисление. Да, именно физик изобрел один из основных инструментов математики! С этого момента математика стала основным языком физической науки. Именно математика позволяет строить современные модели физики. Она дает возможность достаточно строго ввести критерии применимости той или иной модели, найти следствия при помощи хорошо структурированных методов. ### Системы мер и единиц

Окунемся немного в детали. Так как физикам постоянно приходиться иметь дело с измерениями различных физических величин, среди которых масса, время, длина и многое другое, для их корректного сравнения друг с другом, например для сравнения результатов, которые могли бы быть получены различными лабораториями, возникла необходимость договориться о некой единой системе единиц. Например, в чем следует измерять длину – в сантиметрах или дюймах? Более того, большинство величин могут быть выражены через некий базовый набор из независимых единиц. Например, скорость выражается через единицу длины $L$ и единицу времени $T$ как $L/T$, объем же это произведение трех длин $L^3$ и т.д. Такие основные величины называются размерностями.

Сразу скажем, что мы не будем углубляться в некоторые особенные системы единиц, которые характерны для ювелирного дела (карат = 0.2 г), нефтедобычи (баррель~160 литров) или автомобильной промышленности (лошадиная сила ~ 750 Вт), хотя о них все-таки стоит знать, что полезно для вашего собственного кругозора.

Итак, чтобы упростить жизнь ученых, в 1875 году была принята Метрическая конвенция, а также сформировано Международное бюро мер и весов, задачей которого стало поддержание работы единой системы измерений и контроль стандартов в области метрологии. 

В 1960 году на 11 Генеральной конференции по мерам и весам была предложена современная система единиц – Международная система единиц, СИ. Она определяет семь основных единиц физических величин, а также их производных. 
Основные единицы: **килограмм**, **метр**, **секунда**, ампер, кельвин, моль, кандела. В рамках системы СИ считается, что эти единицы имеют независимую размерность. 

Каждая из этих единиц несколько раз меняла способ своего определения. Как правило, изначально они все были привязаны к каким-то физическим объектам.

Проследим за эволюцией метра. 

В 1791 за метр было принято взять $1/10^7$ расстояния от экватора до Северного полюса. После этого в 1872 международным эталоном метра считается расстояние между двумя штрихами на стержне из платинового сплава, хранящемся в Международном бюро мер и весом во Франции. Почти полвека назад было принято следующее улучшение – 1 метр равен $1650763.73$ длин волн в вакууме, которая соответствует переходу между двумя уровнями атома криптона-86. И уже в 1980 году метр был определен как расстояние, которое проходит свет в вакууме за $1/299792458$ секунды. 

В механике нас будут интересовать только три единицы:

Частоты перехода между двумя сверхтонкими уровнями основного состояния атома цезия-133 равна $\Delta\nu_{\text{Cs}}=9192631770$ Гц. Измерив такой период, мы определим что такое секунда. 

После этого, мы берем свет в вакууме и измеряем длину, которую он проходит за $1/299792458$ с, т.е. скорость света имеет точное значение: $c=299792458$ м/c.
 
Затем при помощи весов Киббла мы можем измерить постоянную планка, которую было принято считать равной $6.62607015 \cdot 10^{-34}$  Дж$\cdot$с.
До недавнего времени именно определение килограмма было самой большой проблемой. Сравнив соответствующие эталоны по всему миру, ученые обнаружили, что значения их масс сильно разняться (вплоть до микро килограммов). 


Остальные единицы: [<a href="https://www.bipm.org/en/measurement-units">The International System of Units</a>].


Есть еще система СГС, в которой за единицу длины берется сантиметр, а в качестве единицы массы используется грамм. Вы должны быть готовы к переходу от одной системы к другой. Однако в механике мы будем в основном пользоваться системой СИ, которая так хорошо вам знаком со школьной скамьи.
