# Кинематика СТО. Часть 4

В пространстве Минковского $\gamma\;=\;\dfrac{1}{\sqrt{1\;-\;\beta^2}}$, $\beta\;=\;\dfrac{u}{c}$ можно строить и другие векторы, в которых "длина", а точнее норма задается разностью квадратов временной и пространственной компонент.

<br>

\begin{align}
    \vec{A}\;=\;(A^0,\;A^1,\;A^2,\;A^3),\\
    \vec{A^0\;'}\;=\;\gamma (A^0\;-\;\beta A^1),\\
    \vec{A^1\;'}\;=\;\gamma (A^1\;-\;\beta A^0).
\end{align}

<br>

Расположение значков говорит о том, что данный вектор является контравариантным (противоположно базису) (нижние значки - ковариантными).

# Вектор 4-скорости

Определим вектор 4-скорости (вводим его так, чтобы уже корректно определенная длина сохранялась).

<br>

$$d\vec{R}\;=\;(d\tau,\;dx,\;dy,\;dz)$$
\begin{align}
    &\Rightarrow \; \vec{V}\;=\;\dfrac{d\vec{R}}{dt_c}\;=\;(\gamma l ,\; \gamma \vec{v})\;=\;\left( \dfrac{c}{\sqrt{1\;-\; \dfrac{v^2}{c^2}}},\; \dfrac{\vec{V}}{\sqrt{1\;-\;\dfrac{v^2}{c^2}}} \right),\\
    &dt_c\;=\;dt\sqrt{1\;-\;\dfrac{v^2}{c^2}},\nonumber\\
    &|\vec{V}|\;=\;\sqrt{\gamma_v^2 c^2 \;-\; v^2 \gamma_v^2}\;=\;\sqrt{c^2}\;=\;c.
\end{align}

<br>

**Посмотрим на преобразования:**

<br>

\begin{align}
    \text{1.}\left( \dfrac{c}{\sqrt{1\;-\;v^2}},\; \dfrac{v_x}{\sqrt{1\;-\;v^2}}, \; ... \right)\;=\;(\gamma_v c, \; \gamma_v \vec{V});
    \end{align}
    
   <br>
   
\begin{align}
    \text{2.}\gamma_{v'} c,\; \gamma_{v'}\vec{V'} \text{K' движется относительно K со скоростью u.}
\end{align}

<br>

\begin{align}
    \gamma_{v'}c\;=\;\gamma_u(\gamma_v c \;-\; \beta_u \gamma_v v_x), \nonumber \\
    \gamma_{v'}v'_x\;=\;\gamma_u(\gamma_v v_x\;-\; \beta_u \gamma_v c). \nonumber
\end{align}

<br>

Поделим одно на другое:

<br>

\begin{align*}
    \dfrac{v'_x}{c}\;=\;\dfrac{v_x\;-\;\beta_u c}{c\;-\;\beta_u v_x},\\
    v'_x\;=\;\dfrac{v_x\;-\;u}{1\;-\;\dfrac{u v_x}{c}}.
\end{align*}

<br>

Аналогично вводим 4-импульс: $\vec{P}\;=\;m\vec{V}\;=\;(\gamma m c,\; \gamma m \vec{V})$.

Инвариантное уравнение движения:

<br>

\begin{align}
    \dfrac{d\vec{P}}{dt_c}\;=\;\left( \gamma \dfrac{d}{dt} \gamma mc, \; \gamma \dfrac{d\gamma m \vec{v}}{dt} \right)\;=\;\Phi.
\end{align}

$$\Phi\;=\;(\phi_0, \; \vec{\phi}) \; \text{- 4-сила.}$$

**Для 2-х ИСО**: $dt_c\;=\;\dfrac{dt}{\gamma}$.

Если скорость материальной точки меняется, то такая система будет **сопутствующей**, она будет все время обновляться.

Наблюдали, что пространственная часть $\vec{P}$ это $\gamma \vec{p}\;=\;\gamma m \vec{V}$, тогда логично предположить, что пространственная часть $\Phi$ это $\gamma \vec{F}$, где $\vec{F}$ все обычные силы.

т.е. $\dfrac{d}{dt} \dfrac{m \vec{V}}{\sqrt{1\;-\; \dfrac{v^2}{c^2}}}\;=\;\vec{F}$.

А что с нулевой компонентой?

<br>

$$|\vec{V}|^2\;=\;(V^0)^2\;-\;(V^1)^2\;-\;(V^2)^2\;-\;(V^3)^2$$.

\begin{align}
    &\dfrac{d}{dt_c}|\vec{V}|\;=\;\dfrac{d}{dt_c}c\;-\;\gamma \dfrac{d}{dt} c \;=\;0,\\
    &V^0 \dfrac{dV^0}{dt_c}\;-\; V^1 \dfrac{dV^1}{dt_c}\;-\; V^2 \dfrac{dV^2}{dt_c}\;-\; V^3 \dfrac{dV^3}{dt_c}\;=\;0, 
\end{align}

<br>

$V^0\;=\;\gamma c$, $dt_c\;=\;\dfrac{dt}{\gamma}$ и еще на массу

<br>

\begin{align}
    &\gamma^2 c \dfrac{d}{dt} \gamma cm \;-\; \gamma^2 \vec{V} \dfrac{d}{dt} \gamma m \vec{V} \;=\;0,\\
    &c \cdot \phi_0 \;=\; \gamma \dfrac{\vec{V} \vec{F}}{1},\\
    &\phi_0 \;=\; \gamma \dfrac{\vec{V} \vec{F}}{c},\\
    &\gamma \dfrac{d}{dt} \gamma mc \;=\; \gamma \dfrac{\vec{v} \vec{F}}{c}\cdot c,\\
    &\dfrac{d}{dt} \dfrac{mc^2}{\sqrt{1\;-\; \dfrac{v^2}{c^2}}}\;=\;\dfrac{\vec{V} \vec{F}}{c} \cdot c.
\end{align}

$$\vec{V} \vec{F} \; \text{- мощность}$$ 
$$E\;=\; \gamma mc^2$$

$$\Rightarrow \; \vec{P} \;=\; \left( \dfrac{E}{c}, \; \vec{p} \right)\; \text{p - релятивистский импульс.}$$

\begin{align}
    & |\vec{P}|^2 = \dfrac{\gamma ^2 m^2 c^4}{c^2} - \gamma ^2 m^2 V^2 = \gamma ^2 (m^2 c^2 - m^2 V^2) = \gamma^2 m^2 (c^2 - V^2) = \dfrac{m^2 c^2 (1 - \dfrac{V^2}{c^2})}{1 - \dfrac{V^2}{c^2}} = m^2 c^2 ,
    \\
    & \text{Посчитать квадрат модуля 4-импульса можно и значительно проще :}\\ &|\vec{P}|^2=|m\vec{V}|^2=m^2c^2 (\text{Так как квадрат модуля 4-скорости равен } c^2) \\
    & \text{то есть} \boxed{E^2 - \vec{p}^2 \cdot c^2 = (mc^2)^2}. 
    \\
    &\nonumber mc^2 = E_0 - \text{энергия покоя}
    \\
    & \nonumber T = \gamma mc^2 - mc^2 - \text{кинетическая энергия}. 
\end{align}

<br>

Из $E = \dfrac{mc^2}{\sqrt{1 - \dfrac{V^2}{c^2}}}$ следует, что для $m \neq 0$ необходима бесконечная энергия для достижения скорости света. Таким образом, согласно СТО, только безмассовые частицы. Такая запись $E^2 - \vec{p}^2 \cdot c^2 = (mc^2)^2$ является более общей, чем запись для массивной частицы. 

Для световых частиц:

<br>

\begin{align}
    E^2 = p^2c^2, \; \text{то есть} \; \boxed{p = \dfrac{E}{c}}. 
\end{align}

<br>

# Дефект массы
Если у нас есть система из невзаимодействующих частиц, то:

<br>

\begin{align}
    &F\cdot \vec{v} = -dU/dt
    \\
    &\Rightarrow \dfrac{d\gamma mc^2}{dt}dt + dU = 0
    \\
    &\Rightarrow \gamma mc^2 + U = const
\end{align}

<br>

В релятивизме: 

<br>

\begin{align}
    & T= mc^2 (\gamma - 1)
    \\
    &\Rightarrow T+U = mc^2(\gamma - 1) + U = const, \text{сохраняется в любой ИСО, но меняет значения.}
\end{align}

<br>

**4-импульс:**
 
 Раньше: $ (\dfrac{E}{c}, \vec{p})$
 
 Теперь: $(\dfrac{E-U}{c}, \vec{p})$
 
 
 Мы нигде не говорим об элементарности частицы!
 Пусть у нас есть составная частица, тогда:
 
 <br>
 
 \begin{align}
     M^2 \cdot c^2 = \dfrac{E^2}{c^2} - \vec{p}^2. 
 \end{align}
 
 <br>
 
 Система не взаимодействует $\Rightarrow$
 
 <br>
 
\begin{align}
    &\Rightarrow E = \sum E_i, \; \text{то есть не взаимодейтв.}
    \\
    & \vec{p} = \sum \vec{p}, \; \text{это всегда так}
    \\
    &\Rightarrow M^2 c^2 = \dfrac{(\sum E_i)^2}{c^2} - (\sum \vec{p})^2.
\end{align}

<br>
 
 Давайте перейдём в СО, где $\sum \vec{p} = 0 $
 
 <br>
 
 \begin{align}
     &\Rightarrow M^2 c^2 = \dfrac{(\sum E_i)^2}{c^2}
     \\
     &\Rightarrow M = \dfrac{\sum E_i}{c^2}
     \\
     &\Rightarrow = \dfrac{\sum m_i c^2 + \sum T_i}{c^2}
     \\
     &\Rightarrow M = \sum m_i + \sum \dfrac{T_i}{c^2}, 
 \end{align}
 
 <br>
 
 то есть если мы разбиваем частицу на её составляющие, то масса системы будет больше массы её компонент. [на величину полной кинетической энергии, где $\sum \vec{p} = 0$]
 
 Пусть есть взаимодействие. 
 
 <br>
 
 \begin{align}
     &E = \underset{i}{\sum}\epsilon_i +U , \; \text{U<0 - характеристика устойчивой системы. Энергия связи.}
     \\
     &\Rightarrow M^2 c^2 = \dfrac{(\sum\epsilon_i +U)^2}{c^2}
     \\
     & M = \dfrac{\sum m_i c^2}{c^2} + \dfrac{\underset{i}{\sum}T_i}{c^2} + \dfrac{U}{c^2}. 
 \end{align}
 
 <br>
 
Пусть кинетическая энергия невелика:

<br>

\begin{align}
    & M = \sum m_i + \dfrac{U}{c^2}
    \\
    & \nonumber \text{так как U<0, то}
    \\
    & M = \sum m_i - \dfrac{|U|}{c^2},
    \\
    & \text{то есть} \sum m_i - M = \dfrac{|U|}{c^2} = \triangle M - \text{дефект массы}. 
\end{align}

<br>

Энергия связи должна быть значительной, чтобы использовать формулу:  $\triangle M \cdot c^2 = |U|$. 

# Уравнение движения релятивистской частицы

<br>

\begin{align}
    &\dfrac{\mathrm{d}\vec{P}}{\mathrm{d}t_c} = \vec{\Phi};\\
    &\mathrm{d}t_c = \dfrac{\mathrm{d}t}{\gamma};\\
    &\vec{P} = (\gamma mc, \gamma m \vec{v});\\
    &\Phi = \left(\dfrac{\gamma\vec{v}\vec{F}}{c}, \gamma \vec{F}\right).\\ \hline
    & \dfrac{\mathrm{d}}{\mathrm{d}t}\gamma m c = \dfrac{\vec{v}\vec{F}}{c}:\\
    &1) \dfrac{\mathrm{d}E}{\mathrm{d}t} = \vec{v}\vec{F}\nonumber\\
    &E = \gamma mc^2\nonumber\\
    &2) \dfrac{\mathrm{d}}{\mathrm{d}t} \gamma m \vec{v} = \dfrac{\mathrm{d}}{\mathrm{d}t} \vec{p} = \vec{F}.\nonumber\\\hline
    &\textbf{Вывод:}\nonumber\\
    &\dfrac{\mathrm{d}}{\mathrm{d}t}(\gamma m\vec{v}) = \dfrac{\mathrm{d}}{\mathrm{d}t}\dfrac{E\vec{v}}{c^2} =\nonumber\\
    &=\dfrac{\vec{v}}{c^2} \dfrac{\mathrm{d}E}{\mathrm{d}t} + \dfrac{E}{c^2}\dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = \dfrac{\vec{v}\left(\vec{v},\vec{F}\right)}{c^2} + \gamma m \dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = \vec{F}.\\\hline
    &m\dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t} = \dfrac{1}{\gamma}\left[\vec{F}-\dfrac{\vec{v}}{c^2}\left(\vec{v},\vec{F}\right)\right].
\end{align}

<br>

Получается, что обычное ускорение $\dfrac{\mathrm{d}\vec{v}}{\mathrm{d}t}$ не сонаправлено с силой.
